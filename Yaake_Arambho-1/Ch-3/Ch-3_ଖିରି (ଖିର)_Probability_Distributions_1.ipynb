## Bernoulli Distribution

The Bernoulli distribution is a discrete distribution that models binary trials, with two possible outcomes: $x \in \{0, 1\}$, referred to as "failure" and "success." The Bernoulli distribution is governed by a single parameter $\lambda \in [0, 1]$, which defines the probability of observing a success $x = 1$.

The probability mass function of the Bernoulli distribution is given by:

$$
P(x) = \begin{cases}
1 - \lambda, & \text{if } x = 0 \\
\lambda, & \text{if } x = 1
\end{cases}
$$

Alternatively, it can be expressed as:

$$
P(x) = \lambda^x (1 - \lambda)^{1-x}
$$

Or using the equivalent notation:

$$
P(x) = \text{Bern}_x[\lambda]
$$

## Fig.1: Bernoulli Distribution

The Bernoulli distribution is a discrete distribution with two possible outcomes, $x \in \{0, 1\}$, representing failure and success. It is governed by a single parameter $\lambda$ that determines the probability of success.

## Conclusion

The Bernoulli distribution is a fundamental probability distribution used to model binary trials. It is characterized by a single parameter $\lambda$ that defines the probability of success. This distribution is commonly used in machine vision to model binary data or binary states of the world.


````python
import random
import matplotlib.pyplot as plt

# Bernoulli distribution
def bernoulli_distribution(lambda_param):
    return lambda_param if random.random() < lambda_param else 0

# Generate samples from the Bernoulli distribution
lambda_param = 0.7  # Probability of success
num_samples = 1000
samples = [bernoulli_distribution(lambda_param) for _ in range(num_samples)]

# Plot the histogram of samples
plt.hist(samples, bins=2, density=True, alpha=0.7, label="Bernoulli Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
````

## Beta Distribution

### Definition

The beta distribution is a continuous distribution defined on a single variable $\lambda$ where $\lambda \in [0, 1]$. It is suitable for representing uncertainty in the parameter $\lambda$ of the Bernoulli distribution.

## Parameters

The beta distribution has two parameters $\alpha, \beta \in [0, \infty)$, which both take positive values and affect the shape of the distribution. The expected value of $\lambda$ is given by $E[\lambda] = \frac{\alpha}{\alpha + \beta}$.

## Probability Density Function

The probability density function (PDF) of the beta distribution is defined as:

$$
P(\lambda) = \frac{\lambda^{\alpha - 1} (1 - \lambda)^{\beta - 1}}{\text{B}(\alpha, \beta)}
$$

where $\text{B}(\alpha, \beta)$ is the beta function, defined as:

$$
\text{B}(\alpha, \beta) = \frac{\Gamma(\alpha) \Gamma(\beta)}{\Gamma(\alpha + \beta)}
$$

and $\Gamma(\cdot)$ is the gamma function.

## Fig.2: Beta Distribution

The beta distribution is visualized in Figure 3.2, showing how the shape of the distribution changes with different values of $\alpha$ and $\beta$. The expected value $E[\lambda]$ is indicated for each curve, and the concentration around the expected value increases as the absolute values of $\alpha$ and $\beta$ increase.

## Categorical Distribution

## Definition

The categorical distribution is a discrete distribution that determines the probability of observing one of $K$ possible outcomes. It is a generalization of the Bernoulli distribution, which has only two outcomes.

## Parameters

The categorical distribution is parameterized by a $K \times 1$ vector $\lambda = [\lambda_1, \lambda_2, ..., \lambda_K]$, where $\lambda_k \in [0, 1]$ and $\sum_{k=1}^K \lambda_k = 1$. Each $\lambda_k$ represents the probability of observing the $k$-th outcome.

## Probability Mass Function

The probability mass function (PMF) of the categorical distribution is given by:

$$
P(x = k) = \lambda_k
$$

where $x$ is the random variable representing the outcome.

## Fig.3: Categorical Distribution

The categorical distribution can be visualized as a normalized histogram with $K$ bins, as shown in Figure 3.3. Each bin represents one of the possible outcomes, and the height of the bin corresponds to the probability of that outcome.

## Conclusion

The beta distribution is a continuous distribution used to model uncertainty in parameters, while the categorical distribution is a discrete distribution used to model the probabilities of multiple outcomes. These distributions are commonly used in machine vision to model various aspects of data and states of the world.

## Dirichlet Distribution

## Definition

The Dirichlet distribution is defined over $K$ continuous values $\lambda_1, \lambda_2, ..., \lambda_K$ such that $\sum_{k=1}^K \lambda_k = 1$ and $\lambda_k \in [0, 1]$ for all $k \in \{1, 2, ..., K\}$. It is suitable for defining a distribution over the parameters of the categorical distribution.

## Parameters

In $K$ dimensions, the Dirichlet distribution has $K$ parameters $\alpha_1, \alpha_2, ..., \alpha_K$, each of which can take any positive value. The relative values of the parameters determine the expected values $E[\lambda_1], ..., E[\lambda_k]$. The absolute values determine the concentration around the expected value.

## Probability Density Function

The probability density function (PDF) of the Dirichlet distribution is given by:

$$
P(\lambda_1, ..., \lambda_K) = \frac{\prod_{k=1}^K \lambda_k^{\alpha_k - 1}}{\text{B}(\alpha_1, ..., \alpha_K)}
$$

where $\text{B}(\alpha_1, ..., \alpha_K)$ is the multivariate beta function.

##  Univariate Normal Distribution

## Definition

The univariate normal or Gaussian distribution is defined on continuous values $x \in [-\infty, \infty]$. It is commonly used in vision to model the intensity of a pixel or the distance to an object.

## Parameters

The normal distribution has two parameters: the mean $\mu$ and the variance $\sigma^2$. The mean parameter $\mu$ determines the position of the peak, and the variance $\sigma^2$ determines the width of the distribution.

## Probability Density Function

The probability density function (PDF) of the univariate normal distribution is given by:

$$
P(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$

## Parameters

The normal distribution has two parameters: the mean $\mu$ and the variance $\sigma^2$. The mean parameter $\mu$ can take any value and determines the position of the peak. The variance $\sigma^2$ takes only positive values and determines the width of the distribution.

## Probability Density Function

The probability density function (PDF) of the univariate normal distribution is given by:

## 3.6 Normal-Scaled Inverse Gamma Distribution

## Definition

The normal-scaled inverse gamma distribution is defined over a pair of continuous values $\mu$ and $\sigma^2$, where $\mu$ can take any value and $\sigma^2$ is constrained to be positive. It can define a distribution over the mean and variance parameters of the normal distribution.

## Parameters

The normal-scaled inverse gamma distribution has four parameters $\alpha, \beta, \gamma, \delta$, where $\alpha, \beta, \gamma$ are positive real numbers, and $\delta$ can take any value.

## Probability Density Function

The probability density function (PDF) of the normal-scaled inverse gamma distribution is given by:

$$
P(\mu, \sigma^2) = \frac{\sqrt{\alpha + 1} \sqrt{\gamma} \beta^{\alpha - 1} \exp\left(-\frac{2\beta + \gamma(\delta - \mu)^2}{2\sigma^2}\right)}{\sigma \sqrt{2\pi} \Gamma(\alpha) \sigma^2}
$$

## Definition

The multivariate normal or Gaussian distribution models $D$-dimensional variables $x = [x_1, ..., x_D]^T$, where each dimension $x_d$ is continuous and lies in the range $[-\infty, +\infty]$. It is defined by a $D \times 1$ mean vector $\mu$ and a $D \times D$ covariance matrix $\Sigma$.

## Parameters

The multivariate normal distribution has two parameters: the mean $\mu$ and the covariance $\Sigma$. The mean $\mu$ is a $D \times 1$ vector that describes the mean of the distribution. The covariance $\Sigma$ is a symmetric $D \times D$ positive definite matrix.

## Probability Density Function

The probability density function (PDF) of the multivariate normal distribution is given by:

$$
P(x) = \frac{1}{(2\pi)^{D/2} |\Sigma|^{1/2}} \exp\left(-\frac{1}{2}(x - \mu)^T \Sigma^{-1} (x - \mu)\right)
$$

## Fig.7: Multivariate Normal Distribution

The multivariate normal distribution is visualized in Figure 3.7, showing how the distribution is defined in $D$-dimensional space. The mean vector $\mu$ determines the center of the distribution, and the covariance matrix $\Sigma$ determines the shape and orientation of the ellipsoidal contours.

## Conclusion

The univariate normal distribution is a continuous distribution commonly used in vision to model various continuous variables. The normal-scaled inverse gamma distribution is used to define a distribution over the mean and variance parameters of the normal distribution. The multivariate normal distribution extends the univariate normal to model $D$-dimensional variables, and it is widely used in machine vision for modeling joint distributions and representing uncertainty in the state of the world.

````python
import random
import matplotlib.pyplot as plt

# Univariate Normal Distribution
def univariate_normal_distribution(mu, sigma):
    return random.gauss(mu, sigma)

# Generate samples from the univariate normal distribution
mu = 0
sigma = 1
num_samples = 1000
samples = [univariate_normal_distribution(mu, sigma) for _ in range(num_samples)]

# Plot the histogram of samples
plt.hist(samples, bins=20, density=True, alpha=0.7, label="Univariate Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

# Normal-Scaled Inverse Gamma Distribution
def normal_scaled_inverse_gamma_distribution(alpha, beta, gamma, delta):
    sigma = random.gammavariate(alpha, 1 / beta)
    mu = delta + random.gauss(0, sigma / gamma)
    return mu, sigma

# Generate samples from the normal-scaled inverse gamma distribution
alpha = 2
beta = 3
gamma = 4
delta = 0
num_samples = 1000
samples = [normal_scaled_inverse_gamma_distribution(alpha, beta, gamma, delta) for _ in range(num_samples)]

# Plot the histogram of samples
plt.hist([sample[0] for sample in samples], bins=20, density=True, alpha=0.7, label="Normal-Scaled Inverse Gamma Distribution (mu)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

# Multivariate Normal Distribution
def multivariate_normal_distribution(mu, sigma):
    return [random.gauss(mu[i], sigma[i]) for i in range(len(mu))]

# Generate samples from the multivariate normal distribution
mu = [0, 0]
sigma = [1, 1]
num_samples = 1000
samples = [multivariate_normal_distribution(mu, sigma) for _ in range(num_samples)]

# Plot the histogram of samples for each dimension
for i in range(len(mu)):
    plt.hist([sample[i] for sample in samples], bins=20, density=True, alpha=0.7, label=f"Multivariate Normal Distribution (x{i+1})")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
````

# 3.8 Normal Inverse Wishart Distribution

## Definition

The normal inverse Wishart distribution defines a distribution over a $D \times 1$ vector $\mu$ and a $D \times D$ positive definite matrix $\Sigma$. It is suitable for describing uncertainty in the parameters of a multivariate normal distribution.

## Parameters

The normal inverse Wishart distribution has four parameters $\alpha, \Psi, \gamma, \delta$, where $\alpha$ and $\gamma$ are positive scalars, $\delta$ is a $D \times 1$ vector, and $\Psi$ is a positive definite $D \times D$ matrix.

## Probability Density Function

The probability density function (PDF) of the normal inverse Wishart distribution is given by:

$$
P(\mu, \Sigma) = \frac{\gamma^{D/2} |\Psi|^{\alpha/2} \exp\left(-\frac{1}{2} \text{Tr}[\Psi\Sigma^{-1}] - \gamma(\mu - \delta)^T \Sigma^{-1} (\mu - \delta)\right)}{2^{\alpha D/2} (2\pi)^{D/2} |\Sigma|^{(\alpha + D + 2)/2} \Gamma_D[\alpha/2]}
$$

where $\Gamma_D[\cdot]$ is the multivariate gamma function and $\text{Tr}[\Psi]$ returns the trace of the matrix $\Psi$.

## Fig.8: Sampling from Normal Inverse Wishart Distribution

The normal inverse Wishart distribution is visualized in Figure 3.8, showing how each sample consists of a mean vector and a covariance matrix. The samples are visualized using 2D ellipses, illustrating the iso-contour of the associated Gaussian at a fixed Mahalanobis distance.

## Conjugacy

Conjugacy is a special relationship between a pair of distributions. When a distribution is multiplied by its conjugate, the result is proportional to a new distribution that has the same form as the conjugate. This property is useful in both learning (fitting distributions) and evaluating the model (assessing the probability of new data).

## Conjugate Pairs

- Beta distribution is conjugate to the Bernoulli distribution.
- Dirichlet distribution is conjugate to the categorical distribution.
- Normal-scaled inverse gamma distribution is conjugate to the univariate normal distribution.
- Normal inverse Wishart distribution is conjugate to the multivariate normal distribution.

## Example: Beta and Bernoulli Conjugacy

For example, if we have the product of a Bernoulli distribution and a Beta distribution:

$$
\text{Bern}_x[\lambda] \cdot \text{Beta}_\lambda[\alpha, \beta] = \kappa(x, \alpha, \beta) \cdot \text{Beta}_\lambda[\tilde{\alpha}, \tilde{\beta}]
$$

where $\kappa$ is a scaling factor that is constant with respect to the variable of interest, $\lambda$.

## Conclusion

Conjugate pairs of distributions allow for efficient computation of products during learning and evaluation. The normal inverse Wishart distribution is a key distribution in Bayesian inference for multivariate normal models, providing a way to represent uncertainty in the parameters of the multivariate normal distribution.

````python
import random
import matplotlib.pyplot as plt

# Normal Inverse Wishart Distribution
def normal_inverse_wishart_distribution(alpha, psi, gamma, delta):
    sigma = random.gammavariate(alpha, 1 / psi)
    mu = delta + random.gauss(0, sigma / gamma)
    return mu, sigma

# Generate samples from the normal inverse Wishart distribution
alpha = 2
psi = 3
gamma = 4
delta = [0, 0]
num_samples = 1000
samples = [normal_inverse_wishart_distribution(alpha, psi, gamma, delta) for _ in range(num_samples)]

# Plot the histogram of samples for each dimension
for i in range(len(delta)):
    plt.hist([sample[i] for sample in samples], bins=20, density=True, alpha=0.7, label=f"Normal Inverse Wishart Distribution (mu{i+1})")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

# Multivariate Normal Distribution
def multivariate_normal_distribution(mu, sigma):
    return [random.gauss(mu[i], sigma[i]) for i in range(len(mu))]

# Generate samples from the multivariate normal distribution
mu = [0, 0]
sigma = [1, 1]
num_samples = 1000
samples = [multivariate_normal_distribution(mu, sigma) for _ in range(num_samples)]

# Plot the histogram of samples for each dimension
for i in range(len(mu)):
    plt.hist([sample[i] for sample in samples], bins=20, density=True, alpha=0.7, label=f"Multivariate Normal Distribution (x{i+1})")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
````
