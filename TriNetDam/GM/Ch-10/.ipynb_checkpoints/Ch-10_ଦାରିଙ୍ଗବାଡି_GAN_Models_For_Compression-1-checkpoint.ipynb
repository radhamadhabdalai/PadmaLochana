## Deep Generative Modeling for Neural Compression

### Introduction

In December 2020, Facebook reported having around 1.8 billion daily active users and around 2.8 billion monthly active users [1]. Assuming that users uploaded, on average, a single photo each day, the resulting volume of data would give a very rough (let me stress it, a very rough) estimate of around 3000 TB of new images per day. This single case of Facebook alone already shows us the potential great costs associated with storing and transmitting data. 

In the digital era, we can simply say this: efficient and effective manner of handling data (i.e., faster and smaller) means more money in the pocket. The most straightforward way of dealing with these issues (i.e., smaller and faster) is based on applying compression, and, in particular, image compression algorithms (codecs) that allow us to decrease the size of an image. Instead of changing infrastructure, we can efficiently and effectively store and transmit images by making them simply smaller! 

Let us be honest, the more we compress an image, the more and faster we can send it and the less disk memory we need!

If we think of image compression, probably the first association is JPEG or PNG, standards used on a daily basis by everyone. I will not go into details of these standards (e.g., see [2, 3] for an introduction), but what is important to know is that they use some predefined math like discrete cosine transform. The main advantage of standard codecs like JPEG is that they are interpretable, i.e., all steps are hand-designed, and their behavior can be predicted. However, this comes at the cost of insufficient flexibility which could drastically decrease their performance.

So how can we increase the flexibility of transformations? Any idea? Anyone? Do I hear deep learning [4, 5]? Indeed! Many of today’s image compression algorithms are enhanced by neural networks. The emerging field of compression algorithms using neural networks is called **neural compression**. Neural compression becomes a leading trend in developing new codecs where neural networks replace parts of the standard codecs [6] or neural-based codecs are trained [7] together with quantization [8] and entropy coding [9–12]. 

We will discuss the general compression scheme in detail in the next subsection, but here it is important to understand why **deep generative modeling** is important in the context of neural compression. The answer was given a long time ago by Claude Shannon who showed in [13] that (informally):

> _The length of a message representing source data is proportional to the entropy of this data._

We do not know the entropy of data because we do not know the probability distribution of data, $ p(x) $, but we can estimate it using one of the deep generative models we have discussed so far! Because of that, recently, there has been an increasing interest in using deep generative modeling to improve neural compression. We can use deep generative models for modeling probability distribution for entropy coders [9–12] but also to significantly increase the final reconstruction and compression quality by incorporating new schemes for inference [14] and reconstruction [15].

## General Compression Scheme

Before we jump into neural compression, it is beneficial (and educational, don’t be afraid of refreshing the basics!) to remind ourselves what is image (or data, in general) compression. We can distinguish two image compression approaches [16], namely:

- **Lossless compression**: A method that preserves all information and reconstructions are error-free.
- **Lossy compression**: Information is not preserved completely by a compression method.

A general recipe for designing a compression algorithm relies on devising a uniquely decodable code whose expected length is as close as possible to the entropy of the data [13]. The general compression system consists of two components [16, 17]: an encoder and a decoder.

Please do not think of it as a deterministic VAE because it is not the same. There are some similarities, but in the compression task, we are really interested in sending the bitstream, while in VAEs we typically do not care at all. We play with floats and say about codes just to make it more intuitive, but it requires a few extra steps to turn it into a “real” compression scheme. We are going to explain these extra steps in the next sections.

###  Encoder

In the encoder, the goal is to transform an image into a discrete signal. It is important to understand that the signal does not necessarily need to be binary. The transformation we use could be invertible; however, it is not a requirement. If the transformation is invertible, then we can use its inverse in the decoder, and, in principle, we can talk about **lossless compression**. Why? Take a look at the flow-based models where we discussed invertibility. However, if the transformation is not invertible, then some information is lost, and we land in the group of lossy compression methods.

Next, after applying the transformation to the input image, the discrete signal is encoded into a bitstream in a lossless manner. In other words, discrete symbols are mapped to binary variables (bits). Typically, entropy coders utilize information about the probability of symbol occurrence, e.g., Huffman coders or arithmetic coders [16]. This is important to understand that for many entropy coders we need to know \( p(x) \), and here we can use deep generative models.

###  Decoder

Once the message (i.e., the bits) is sent and received, the bitstream is decoded to the discrete signal by the entropy decoder. The entropy decoder is the inversion of the entropy encoder. Entropy coding methods allow us to receive the original symbols from the bitstream. Eventually, an inverse transformation (not necessarily the inversion of the encoder transformation) is applied to reconstruct the original image.

### The Objective

The final performance of a codec is evaluated in terms of reconstruction error and compression ratio. The **reconstruction error** is called the distortion measure and is calculated as a difference between the original image and the reconstructed image using the mean square error (MSE) (typically, the peak signal-to-noise ratio (PSNR) $ 255^2 $ expressed as $ 10 \log_{10} \text{MSE} $ is reported) or perceptual metrics like the multi-scale structure similarity index (MS-SSIM) [19]. 

The **compression ratio**, called **rate**, is usually expressed by the bits per pixel (bpp), i.e., the total size in bits of the encoder output divided by the total size in pixels of the encoder input [17].

Typically, the performance of codecs is compared by inspecting the **rate-distortion plane** (i.e., plotting curves on a plane with the rate on the x-axis and the distortion on the y-axis).

Formally, we assume an autoencoder architecture (see Fig. 10.1 again) with an encoding transformation, $ f_e: X \rightarrow Y $, that takes an input $ x $ and returns a discrete signal $ y $ (a code). After sending the message, a reconstruction $ \hat{x} $ is given by a decoder, $ f_d: Y \rightarrow X $. Moreover, there is an (adaptive) entropy coding model that learns the distribution $ p(y) $ and is further used to turn the discrete signal $ y $ into a bitstream by an entropy coder (e.g., Huffman coding, arithmetic coding).

If a compression method has any adaptive (hyper)parameters, it could be learned by optimizing the following objective function:

$$
L(x) = d(x, \hat{x}) + \beta r(y)
$$

where:
- $ d(\cdot, \cdot) $ is the distortion measure (e.g., PSNR, MS-SSIM),
- $ r(\cdot) $ is the rate measure (e.g., $ r(y) = -\ln p(y) $),
- $ \beta > 0 $ is a weighting factor controlling the balance between rate and distortion.

Notice that distortion requires both the encoder and the decoder, while rate requires the encoder and the entropy model.


````python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_probability as tfp

# Step 1: Build the Autoencoder Model
def build_autoencoder(input_shape=(256, 256, 3)):
    # Encoder
    input_img = layers.Input(shape=input_shape)
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    
    # Bottleneck (latent space representation)
    latent_space = layers.Flatten()(x)
    latent_space = layers.Dense(1024, activation='relu')(latent_space)

    # Decoder
    x = layers.Dense(128 * 32 * 32, activation='relu')(latent_space)
    x = layers.Reshape((32, 32, 128))(x)
    x = layers.Conv2DTranspose(128, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2DTranspose(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2DTranspose(32, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    decoded = layers.Conv2DTranspose(3, (3, 3), activation='sigmoid', padding='same')(x)

    # Autoencoder model
    autoencoder = models.Model(input_img, decoded)
    return autoencoder

# Step 2: Compile the Model
autoencoder = build_autoencoder()
autoencoder.compile(optimizer='adam', loss='mse')

# Step 3: Prepare the Dataset
# Load and prepare the CIFAR-10 dataset
(x_train, _), (x_test, _) = tf.keras.datasets.cifar10.load_data()

# Normalize images to the range [0, 1]
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# Resize the images to match the input shape of the model
x_train = tf.image.resize(x_train, (256, 256))
x_test = tf.image.resize(x_test, (256, 256))

# Step 4: Train the Autoencoder
history = autoencoder.fit(x_train, x_train,
                          epochs=10,
                          batch_size=64,
                          validation_data=(x_test, x_test))

# Step 5: Visualize the Results
def visualize_results(x_test, decoded_imgs):
    n = 10  # Number of images to display
    plt.figure(figsize=(20, 4))
    for i in range(n):
        # Display original
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(x_test[i])
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

        # Display reconstruction
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(decoded_imgs[i])
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.show()

# Predict some images from the test set
decoded_imgs = autoencoder.predict(x_test[:10])

# Visualize the original and reconstructed images
visualize_results(x_test, decoded_imgs)

# Step 6: Implement Entropy Coding (Using Normal Distribution for Latent Space)
latent_distribution = tfp.distributions.Normal(loc=0., scale=1.)
entropy = latent_distribution.entropy()

print(f"Entropy of the latent distribution: {entropy}")

# Step 7: Calculate PSNR (Peak Signal to Noise Ratio)
def calculate_psnr(original, reconstructed):
    mse = np.mean((original - reconstructed) ** 2)
    if mse == 0:
        return 100
    max_pixel = 1.0
    return 20 * np.log10(max_pixel / np.sqrt(mse))

# Example: Calculate PSNR for the first test image
psnr = calculate_psnr(x_test[0], decoded_imgs[0])
print(f"PSNR for the first test image: {psnr} dB")
````

````python
import numpy as np
import matplotlib.pyplot as plt
import random

# Generate random "image" data for testing (e.g., 256x256 image with 3 channels RGB)
def generate_random_image(height=256, width=256):
    # Simulate an image with random RGB values (values between 0 and 255)
    return np.random.randint(0, 256, size=(height, width, 3), dtype=np.uint8)

# A simplified autoencoder with basic manual "compression" mechanism
class SimpleAutoencoder:
    def __init__(self, compression_factor=16):
        self.compression_factor = compression_factor  # The compression factor (downscale ratio)
    
    def encode(self, image):
        # Simulate "compression" by reducing resolution (downsampling) and encoding the image
        height, width, _ = image.shape
        new_height = height // self.compression_factor
        new_width = width // self.compression_factor
        compressed_image = image[::self.compression_factor, ::self.compression_factor]
        return compressed_image
    
    def decode(self, compressed_image):
        # "Decompression" by repeating the compressed image (upsampling)
        new_height, new_width, _ = compressed_image.shape
        decompressed_image = np.repeat(np.repeat(compressed_image, self.compression_factor, axis=0), self.compression_factor, axis=1)
        return decompressed_image[:new_height*self.compression_factor, :new_width*self.compression_factor, :]

# Example: Image Compression
def visualize_image_compression(original_image, compressed_image, decompressed_image):
    # Visualize the original, compressed, and decompressed images side by side
    plt.figure(figsize=(12, 4))

    # Original image
    plt.subplot(1, 3, 1)
    plt.imshow(original_image)
    plt.title("Original Image")
    plt.axis("off")

    # Compressed image
    plt.subplot(1, 3, 2)
    plt.imshow(compressed_image)
    plt.title("Compressed Image")
    plt.axis("off")

    # Decompressed image
    plt.subplot(1, 3, 3)
    plt.imshow(decompressed_image)
    plt.title("Decompressed Image")
    plt.axis("off")

    plt.show()

# Simulating the Compression Process
original_image = generate_random_image()  # Create a random image to simulate the process
autoencoder = SimpleAutoencoder(compression_factor=16)

# Perform encoding (compression)
compressed_image = autoencoder.encode(original_image)

# Perform decoding (decompression)
decompressed_image = autoencoder.decode(compressed_image)

# Visualize the results
visualize_image_compression(original_image, compressed_image, decompressed_image)
````

````python
import random
import matplotlib.pyplot as plt

# Simulate an "image" as a 2D list of pixel values (height x width x channels)
def generate_random_image(height=256, width=256):
    """Generate a random image with RGB values"""
    return [[[random.randint(0, 255) for _ in range(3)] for _ in range(width)] for _ in range(height)]

class SimpleAutoencoder:
    def __init__(self, compression_factor=16):
        """Initialize the autoencoder with a compression factor"""
        self.compression_factor = compression_factor

    def encode(self, image):
        """Simulate compression by downscaling (subsampling) the image"""
        height = len(image)
        width = len(image[0])
        
        # Downscale the image (subsampling)
        compressed_image = []
        for i in range(0, height, self.compression_factor):
            row = []
            for j in range(0, width, self.compression_factor):
                row.append(image[i][j])
            compressed_image.append(row)
        return compressed_image
    
    def decode(self, compressed_image):
        """Simulate decompression by upscaling the image (nearest neighbor replication)"""
        compressed_height = len(compressed_image)
        compressed_width = len(compressed_image[0])
        
        # Upscale the compressed image (nearest neighbor interpolation)
        decompressed_image = []
        for i in range(compressed_height):
            row = []
            for j in range(compressed_width):
                # Replicate the compressed image pixels by the compression factor
                for _ in range(self.compression_factor):
                    row.append(compressed_image[i][j])
            # Replicate the rows by the compression factor
            for _ in range(self.compression_factor):
                decompressed_image.append(row)
        return decompressed_image

def visualize_image_compression(original_image, compressed_image, decompressed_image):
    """Display the original, compressed, and decompressed images using matplotlib"""
    plt.figure(figsize=(12, 4))
    
    # Original image
    plt.subplot(1, 3, 1)
    plt.imshow(original_image)
    plt.title("Original Image")
    plt.axis("off")
    
    # Compressed image (downsampled)
    plt.subplot(1, 3, 2)
    plt.imshow(compressed_image)
    plt.title("Compressed Image")
    plt.axis("off")
    
    # Decompressed image (upsampled)
    plt.subplot(1, 3, 3)
    plt.imshow(decompressed_image)
    plt.title("Decompressed Image")
    plt.axis("off")
    
    plt.show()

# Example usage of the SimpleAutoencoder
original_image = generate_random_image()  # Random 256x256 RGB image
autoencoder = SimpleAutoencoder(compression_factor=16)

# Encoding (compressing)
compressed_image = autoencoder.encode(original_image)

# Decoding (decompressing)
decompressed_image = autoencoder.decode(compressed_image)

# Visualizing the results
visualize_image_compression(original_image, compressed_image, decompressed_image)
````

## JPEG Compression: A Comprehensive Overview

###  JPEG Compression Basics

#### Color Space Transformation

The JPEG compression process begins with transforming RGB color space to YCbCr:

$$
\begin{bmatrix} Y \\ Cb \\ Cr \end{bmatrix} = \begin{bmatrix} 0.299 & 0.587 & 0.114 \\ -0.168736 & -0.331264 & 0.5 \\ 0.5 & -0.48688 & -0.081312 \end{bmatrix} \begin{bmatrix} R \\ G \\ B \end{bmatrix} + \begin{bmatrix} 0 \\ 128 \\ 128 \end{bmatrix}
$$

### Compression Stages

1. **Chroma Subsampling**: Cb and Cr channels are downscaled
2. **Block Segmentation**: Split channels into 8 × 8 blocks
3. **Discrete Cosine Transform (DCT)**: Apply frequency domain transformation
4. **Quantization**: Reduce precision of DCT coefficients
5. **Entropy Coding**: Typically Huffman coding

## Neural Compression Components

### Key Challenges
- Flexibility beyond predefined transformations
- Differentiable quantization
- Adaptive entropy coding

### Mathematical Representation of Neural Compression

Let $x$ be the input image, $f_{\theta}$ be the neural compression function:

$$
\hat{x} = \text{Decode}\left(\text{Entropy\_Coding}\left(\text{Quantize}\left(f_{\theta}(x)\right)\right)\right)
$$

Where:
- $f_{\theta}$ is a learnable transformation
- $\text{Quantize}$ provides discrete representation
- $\text{Entropy\_Coding}$ optimizes compression ratio

# Neural Compression: Encoder and Decoder Implementations

## Mathematical Framework

The neural compression process can be represented as:

$$
\begin{aligned}
\text{Encoder: } h_e &= E_{\theta_e}(x) \\
\text{Decoder: } x_{\text{rec}} &= D_{\theta_d}(h_e)
\end{aligned}
$$

Where:
- $x$ is the input image
- $h_e$ is the encoded representation
- $x_{\text{rec}}$ is the reconstructed image
- $\theta_e, \theta_d$ are encoder and decoder parameters

## PyTorch Implementation

### Encoder Class

```python
class Encoder(nn.Module):
    def __init__(self, encoder_net):
        super(Encoder, self).__init__()
        self.encoder = encoder_net
    
    def encode(self, x):
        h_e = self.encoder(x)
        return h_e
    
    def forward(self, x):
        return self.encode(x)
```

### Decoder Class

```python
class Decoder(nn.Module):
    def __init__(self, decoder_net):
        super(Decoder, self).__init__()
        self.decoder = decoder_net
    
    def decode(self, z):
        h_d = self.decoder(z)
        return h_d
    
    def forward(self, z, x=None):
        x_rec = self.decode(z)
        return x_rec
```

```python
e_net = nn.Sequential(
```

#### Decoder Network

```python
d_net = nn.Sequential(

    nn.Linear(M//2, M), nn.BatchNorm1d(M), nn.ReLU(),

```

### Key Characteristics
- Continuous code output
- No distribution parameterization
- Focus on architectural design
- Symmetric encoder-decoder structure

## Hyperparameters
- $D$: Input dimension
- $M$: Intermediate dimension
- $C$: Compressed representation size

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

class Encoder(nn.Module):
    def __init__(self, encoder_net):
        super(Encoder, self).__init__()
        self.encoder = encoder_net
    
    def encode(self, x):
        return self.encoder(x)
    
    def forward(self, x):
        return self.encode(x)

class Decoder(nn.Module):
    def __init__(self, decoder_net):
        super(Decoder, self).__init__()
        self.decoder = decoder_net
    
    def decode(self, z):
        return self.decoder(z)
    
    def forward(self, z, x=None):
        return self.decode(z)

class NeuralCompressor(nn.Module):
    def __init__(self, D, M, C):
        """
        Args:
        D: Input dimension
        M: Intermediate dimension
        C: Compressed representation size
        """
        super(NeuralCompressor, self).__init__()
        
        # Encoder Network
        encoder_net = nn.Sequential(
            nn.Linear(D, M*2), 
            nn.BatchNorm1d(M*2), 
            nn.ReLU(),
            nn.Linear(M*2, M), 
            nn.BatchNorm1d(M), 
            nn.ReLU(),
            nn.Linear(M, M//2), 
            nn.BatchNorm1d(M//2), 
            nn.ReLU(),
            nn.Linear(M//2, C)
        )
        
        # Decoder Network
        decoder_net = nn.Sequential(
            nn.Linear(C, M//2), 
            nn.BatchNorm1d(M//2), 
            nn.ReLU(),
            nn.Linear(M//2, M), 
            nn.BatchNorm1d(M), 
            nn.ReLU(),
            nn.Linear(M, M*2), 
            nn.BatchNorm1d(M*2), 
            nn.ReLU(),
            nn.Linear(M*2, D)
        )
        
        self.encoder = Encoder(encoder_net)
        self.decoder = Decoder(decoder_net)
    
    def forward(self, x):
        # Encode input
        compressed = self.encoder(x)
        
        # Decode compressed representation
        reconstructed = self.decoder(compressed)
        
        return compressed, reconstructed
    
    def compute_loss(self, x, reconstructed):
        """
        Compute reconstruction loss
        
        Args:
        x: Original input
        reconstructed: Reconstructed input
        
        Returns:
        Reconstruction loss
        """
        return F.mse_loss(x, reconstructed)

# Example usage
def main():
    # Hyperparameters
    D = 784  # Input dimension (e.g., MNIST)
    M = 256  # Intermediate dimension
    C = 32   # Compressed representation size
    
    # Create model
    compressor = NeuralCompressor(D, M, C)
    
    # Simulate input
    x = torch.randn(64, D)  # Batch of 64 samples
    
    # Forward pass
    compressed, reconstructed = compressor(x)
    
    # Compute loss
    loss = compressor.compute_loss(x, reconstructed)
    print(f"Reconstruction Loss: {loss.item()}")

if __name__ == "__main__":
    main()
````

import torch

class NeuralCompressor(nn.Module):
    def __init__(self, D, M, C):
        super(NeuralCompressor, self).__init__()
        
        encoder_net = nn.Sequential(
            nn.Linear(D, M*2), 
            nn.BatchNorm1d(M*2), 
            nn.ReLU(),
            nn.Linear(M*2, M), 
            nn.BatchNorm1d(M), 
            nn.ReLU(),

            nn.BatchNorm1d(M//2), 
            nn.ReLU(),
            nn.Linear(M//2, C)

        decoder_net = nn.Sequential(

            nn.ReLU(),

            nn.ReLU(),
            nn.Linear(M*2, D)

        self.encoder = nn.Sequential(*list(encoder_net.children()))
        self.decoder = nn.Sequential(*list(decoder_net.children()))
    
    def forward(self, x):
        compressed = self.encoder(x)
        reconstructed = self.decoder(compressed)
        return compressed, reconstructed

# Differentiable Quantization in Neural Compression

## Mathematical Formulation

### Quantization Process

Given:
- $y \in \mathbb{R}^M$: Continuous code
- $c \in \mathbb{R}^K$: Codebook

Quantization Steps:
1. Similarity Matrix Calculation:
   
   $$S_{ij} = \exp\left(-\frac{(y_i - c_j)^2}{\tau}\right)$$

2. Softmax with Temperature:
   
   $$\hat{S}_{ij} = \frac{\exp(S_{ij}/\tau)}{\sum_k \exp(S_{ik}/\tau)}$$

3. Quantized Code:
   
   $$\hat{y} = \hat{S} \cdot c$$

## Python Implementation

```python
import torch
import torch.nn.functional as F

class DifferentiableQuantization:
    def __init__(self, codebook_size=256, code_dim=64):
        self.codebook = torch.randn(codebook_size, code_dim)
        self.temperature = 1.0
    
    def quantize(self, y):
        # Repeat y and codebook for broadcasting
        Y = y.unsqueeze(1).repeat(1, self.codebook.shape[0])
        C = self.codebook.unsqueeze(0).repeat(y.shape[0], 1, 1)
        
        # Compute similarity matrix
        similarity = torch.exp(-torch.pow(Y - C, 2))
        
        # Apply softmax with temperature
        quantization_indices = F.softmax(similarity / self.temperature, dim=1)
        
        # Quantized code
        quantized_code = torch.matmul(quantization_indices, self.codebook)
        
        return quantized_code, quantization_indices
```

### Key Properties
- Differentiable quantization
- Learnable codebook
- Discretization via softmax
- Gradient-compatible encoding

## Mathematical Insights

The quantization process ensures:
1. Discrete representation
2. Differentiable gradient flow
3. Finite codebook values

# Differentiable Quantizer in Neural Compression

## Mathematical Formulation

### Quantization Process

$$
\begin{aligned}
\text{Distances} &= \exp\left(-\sqrt{(x - c)^2}\right) \\
\text{Soft Indices} &= \text{softmax}(\tau \cdot \text{Distances}) \\
\text{Quantized Code} &= \text{Soft Indices} \cdot \text{Codebook}
\end{aligned}
$$

## Python Implementation

```python
import torch
import torch.nn as nn

class Quantizer(nn.Module):
    def __init__(self, input_dim, codebook_dim, temp=1e7):
        super(Quantizer, self).__init__()

        self.input_dim = input_dim
        self.codebook_dim = codebook_dim
        
        # Learnable codebook initialization
        self.codebook = nn.Parameter(
            torch.FloatTensor(1, codebook_dim).uniform_(-1/codebook_dim, 1/codebook_dim)
        )
    
    def indices2codebook(self, indices_onehot):
        return torch.matmul(indices_onehot, self.codebook.t()).squeeze()
    
    def indices_to_onehot(self, inputs_shape, indices):
        indices_hard = torch.zeros(inputs_shape[0], inputs_shape[1], self.codebook_dim)
        indices_hard.scatter_(2, indices, 1)
        return indices_hard
    
    def forward(self, inputs):

        inputs_repeat = inputs.unsqueeze(2).repeat(1, 1, self.codebook_dim)
        
        # Distance calculation
        distances = torch.exp(-torch.sqrt(torch.pow(inputs_repeat - self.codebook.unsqueeze(1), 2)))
        
        # Hard indices (non-differentiable)
        indices = torch.argmax(distances, dim=2).unsqueeze(2)
        indices_hard = self.indices_to_onehot(inputs_shape, indices)
        
        # Soft indices (differentiable)
        indices_soft = torch.softmax(self.temp * distances, -1)

        return indices_soft, indices_hard, quantized
```

### Key Features
- Differentiable quantization
- Learnable codebook
- Soft and hard indexing
- Temperature-based softmax

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

class NeuralQuantizer(nn.Module):
    def __init__(self, input_dim, codebook_dim, temp=1e7):
        super().__init__()
        
        # Codebook parameters
        self.temp = temp
        self.input_dim = input_dim
        self.codebook_dim = codebook_dim
        
        # Learnable codebook initialization
        self.codebook = nn.Parameter(
            torch.FloatTensor(1, codebook_dim).uniform_(-1/codebook_dim, 1/codebook_dim)
        )
    
    def indices2codebook(self, indices_onehot):
        """Map one-hot indices to codebook values"""
        return torch.matmul(indices_onehot, self.codebook.t()).squeeze()
    
    def indices_to_onehot(self, inputs_shape, indices):
        """Convert indices to one-hot representation"""
        indices_hard = torch.zeros(inputs_shape[0], inputs_shape[1], self.codebook_dim)
        indices_hard.scatter_(2, indices, 1)
        return indices_hard
    
    def forward(self, inputs):
        """Quantization forward pass"""
        inputs_shape = inputs.shape
        
        # Repeat inputs for broadcasting
        inputs_repeat = inputs.unsqueeze(2).repeat(1, 1, self.codebook_dim)
        
        # Calculate distances between inputs and codebook
        distances = torch.exp(-torch.sqrt(torch.pow(inputs_repeat - self.codebook.unsqueeze(1), 2)))
        
        # Hard indices (non-differentiable)
        indices = torch.argmax(distances, dim=2).unsqueeze(2)
        indices_hard = self.indices_to_onehot(inputs_shape, indices)
        
        # Soft indices (differentiable)
        indices_soft = F.softmax(self.temp * distances, dim=-1)
        
        # Quantized values using soft indices
        quantized = self.indices2codebook(indices_onehot=indices_soft)
        
        return indices_soft, indices_hard, quantized

def test_quantizer():
    # Example usage
    input_dim = 64
    codebook_dim = 256
    batch_size = 32
    
    quantizer = NeuralQuantizer(input_dim, codebook_dim)
    
    # Random input tensor
    inputs = torch.randn(batch_size, input_dim)
    
    # Quantization
    soft_indices, hard_indices, quantized = quantizer(inputs)
    
    print(f"Input shape: {inputs.shape}")
    print(f"Soft Indices shape: {soft_indices.shape}")
    print(f"Hard Indices shape: {hard_indices.shape}")
    print(f"Quantized shape: {quantized.shape}")

if __name__ == "__main__":
    test_quantizer()
````

````python
import random
import math

class NeuralQuantizer:
    def __init__(self, input_dim=64, codebook_dim=256, temp=10):
        self.input_dim = input_dim
        self.codebook_dim = codebook_dim
        self.temp = temp
        
        # Initialize codebook with random values
        self.codebook = [random.uniform(-1/codebook_dim, 1/codebook_dim) 
                         for _ in range(codebook_dim)]
    
    def calculate_distances(self, inputs):
        """Calculate distances between inputs and codebook"""
        distances = []
        for input_val in inputs:
            input_distances = [
                math.exp(-abs(input_val - code_val)) 
                for code_val in self.codebook
            ]
            distances.append(input_distances)
        return distances
    
    def softmax(self, distances):
        """Apply softmax with temperature"""
        softmax_results = []
        for input_distances in distances:
            # Find max to prevent overflow
            max_dist = max(input_distances)
            
            # Compute scaled exponentials
            exp_values = [math.exp((dist - max_dist) * self.temp) for dist in input_distances]
            
            # Compute sum for normalization
            sum_exp = sum(exp_values)
            
            # Normalize
            soft_indices = [val / sum_exp for val in exp_values]
            softmax_results.append(soft_indices)
        
        return softmax_results
    
    def quantize(self, inputs):
        """Quantization process"""
        # Calculate distances
        distances = self.calculate_distances(inputs)
        
        # Soft indices (differentiable)
        soft_indices = self.softmax(distances)
        
        # Quantized values
        quantized = [
            sum(soft_index * code_val for soft_index, code_val in zip(input_soft_indices, self.codebook))
            for input_soft_indices in soft_indices
        ]
        
        return soft_indices, quantized

def test_quantizer():
    # Create quantizer
    quantizer = NeuralQuantizer(input_dim=4, codebook_dim=8)
    
    # Test inputs
    inputs = [random.uniform(-1, 1) for _ in range(4)]
    
    # Quantize
    soft_indices, quantized = quantizer.quantize(inputs)
    
    # Print results
    print("Inputs:", inputs)
    print("Soft Indices:", soft_indices)
    print("Quantized Values:", quantized)

# Run test
test_quantizer()
````

## A Neural Compression System

Neural networks can be utilized for data compression through a specialized system.

### Objective Function

Given:
- Training data $\mathcal{D} = \{x_1, \ldots, x_N\}$
- Empirical distribution $p_\text{data}(x)$

Components:
- Encoder network $f_{e,\phi}(x)$ with weights $\phi$
- Quantizer $Q(y;c)$ with codebook $c$
- Decoder network $f_{d,\theta}(\hat{y})$ with weights $\theta$
- Entropy coding model $p_\lambda(\hat{y})$ with weights $\lambda$

Training objective:

$$L(\theta, \phi, \lambda, c) = \mathbb{E}_{x \sim p_\text{data}(x)} (x - \hat{x})^2 + \beta \mathbb{E}_{\hat{y} \sim p_\text{data}(x)} \left[-\ln p_\lambda(\hat{y})\right]$$

### Bits per Pixel Calculation

$$\text{bpp} = -\log_2(p(\hat{y}))/D$$

Where:
- $M$: code size
- $K = 2^\kappa$: possible values
- $D$: image size

The objective balances reconstruction error (distortion) and rate, allowing end-to-end training of the compression system.

## Neural Compressor Implementation

### Bits per Pixel Calculation

The bits per pixel (bpp) is calculated as:

$$\text{bpp} = \begin{cases} 
-\log_2(1/(\kappa \times M))/D & \text{(Uniform Distribution)} \\
-\log_2(p(\hat{y}))/D & \text{(Entropy Coding)}
\end{cases}$$

Where:
- $M$: code size
- $\kappa$: number of bits per value
- $K = 2^\kappa$: possible values
- $D$: image size
- $p(\hat{y})$: probability from entropy coding

### Neural Compressor Class

```python
class NeuralCompressor(nn.Module):
    def __init__(self, encoder, decoder, entropy_coding, quantizer, beta=1., detaching=False):
        super(NeuralCompressor, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.entropy_coding = entropy_coding
        self.quantizer = quantizer
        self.beta = beta
        self.detaching = detaching

    def forward(self, x, reduction='avg'):
        # Encoding
        z = self.encoder(x)
        
        # Quantizing
        quantizer_out = self.quantizer(z)
        
        # Decoding
        x_rec = self.decoder(quantizer_out[2])
        
        # Distortion (MSE)
        Distortion = torch.mean(torch.pow(x - x_rec, 2), 1)
        
        # Rate via entropy coding
        Rate = torch.mean(self.entropy_coding(quantizer_out[0], quantizer_out[2]), 1)
        
        # Objective
        objective = Distortion + self.beta * Rate
        
        return (objective.sum(), Distortion.sum(), Rate.sum()) if reduction == 'sum' \
               else (objective.mean(), Distortion.mean(), Rate.mean())
```

### Compression Procedure

1. Encode input image: $y = f_{e,\phi}(x)$
2. Quantize code: $\hat{y} = Q(\hat{y}; c)$
3. Convert quantized code to bitstream using $p_\lambda(\hat{y})$
4. Transmit bits
5. Decode bits using $p_\lambda(\hat{y})$
6. Reconstruct image: $\hat{x} = f_{d,\theta}(\hat{y})$

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

class Quantizer(nn.Module):
    def __init__(self, num_embeddings, embedding_dim):
        super().__init__()
        self.embedding = nn.Embedding(num_embeddings, embedding_dim)
        self.num_embeddings = num_embeddings
        
    def forward(self, x):
        # Flatten input
        x_flat = x.view(-1, x.shape[-1])
        
        # Calculate distances to embedding vectors
        distances = (x_flat.pow(2).sum(1, keepdim=True) 
                     - 2 * x_flat @ self.embedding.weight.t() 
                     + self.embedding.weight.pow(2).sum(1))
        
        # Find nearest embedding
        indices = distances.argmin(1)
        quantized = self.embedding(indices).view(x.shape)
        
        return quantized, indices, x

class Encoder(nn.Module):
    def __init__(self, input_dim, hidden_dims, latent_dim):
        super().__init__()
        layers = []
        prev_dim = input_dim
        for hidden_dim in hidden_dims:
            layers.append(nn.Conv2d(prev_dim, hidden_dim, kernel_size=3, stride=2, padding=1))
            layers.append(nn.ReLU())
            prev_dim = hidden_dim
        
        self.encoder = nn.Sequential(*layers)
        self.fc = nn.Linear(prev_dim * (input_dim // (2 ** len(hidden_dims))), latent_dim)
    
    def forward(self, x):
        x = self.encoder(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)

class Decoder(nn.Module):
    def __init__(self, latent_dim, hidden_dims, output_dim):
        super().__init__()
        self.fc = nn.Linear(latent_dim, hidden_dims[0] * (output_dim // (2 ** len(hidden_dims))))
        
        layers = []
        prev_dim = hidden_dims[0]
        for hidden_dim in hidden_dims[1:]:
            layers.append(nn.ConvTranspose2d(prev_dim, hidden_dim, kernel_size=3, stride=2, padding=1, output_padding=1))
            layers.append(nn.ReLU())
            prev_dim = hidden_dim
        
        layers.append(nn.ConvTranspose2d(prev_dim, output_dim, kernel_size=3, stride=1, padding=1))
        layers.append(nn.Tanh())
        
        self.decoder = nn.Sequential(*layers)
    
    def forward(self, x):
        x = self.fc(x)
        x = x.view(x.size(0), -1, output_dim // (2 ** len(hidden_dims)), output_dim // (2 ** len(hidden_dims)))
        return self.decoder(x)

class EntropyEncoder(nn.Module):
    def __init__(self, latent_dim):
        super().__init__()
        self.fc1 = nn.Linear(latent_dim, latent_dim // 2)
        self.fc2 = nn.Linear(latent_dim // 2, 1)
    
    def forward(self, indices, quantized):
        # Simple entropy estimation
        entropy = F.softmax(self.fc2(F.relu(self.fc1(quantized))), dim=0)
        return -torch.log(entropy + 1e-7)

class NeuralCompressor(nn.Module):
    def __init__(self, input_dim, hidden_dims, latent_dim, num_embeddings, beta=1.0):
        super().__init__()
        self.encoder = Encoder(input_dim, hidden_dims, latent_dim)
        self.decoder = Decoder(latent_dim, hidden_dims[::-1], input_dim)
        self.quantizer = Quantizer(num_embeddings, latent_dim)
        self.entropy_encoder = EntropyEncoder(latent_dim)
        self.beta = beta
    
    def forward(self, x):
        # Encode
        z = self.encoder(x)
        
        # Quantize
        quantized, indices, _ = self.quantizer(z)
        
        # Decode
        x_rec = self.decoder(quantized)
        
        # Calculate distortion (MSE)
        distortion = F.mse_loss(x, x_rec, reduction='none').mean((1,2,3))
        
        # Calculate rate (entropy)
        rate = self.entropy_encoder(indices, quantized)
        
        # Total loss
        loss = distortion + self.beta * rate
        
        return loss, distortion, rate

# Example usage
if __name__ == "__main__":
    # Hyperparameters
    input_dim = 1  # grayscale image
    hidden_dims = [16, 32, 64]
    latent_dim = 128
    num_embeddings = 512
    beta = 1.0

    # Create model
    compressor = NeuralCompressor(
        input_dim=input_dim, 
        hidden_dims=hidden_dims, 
        latent_dim=latent_dim, 
        num_embeddings=num_embeddings, 
        beta=beta
    )

    # Example training loop
    optimizer = torch.optim.Adam(compressor.parameters())
    
    def train(model, dataloader, epochs):
        for epoch in range(epochs):
            total_loss = 0
            for batch in dataloader:
                optimizer.zero_grad()
                loss, distortion, rate = model(batch)
                loss.mean().backward()
                optimizer.step()
                total_loss += loss.mean().item()
            print(f"Epoch {epoch+1}, Total Loss: {total_loss}")
````

````python
import math
import random

class Quantizer:
    def __init__(self, num_embeddings, embedding_dim):
        self.num_embeddings = num_embeddings
        self.embedding_dim = embedding_dim
        self.embeddings = [
            [random.uniform(-1, 1) for _ in range(embedding_dim)] 
            for _ in range(num_embeddings)
        ]
    
    def forward(self, x):
        # Find nearest embedding
        distances = [
            sum((x_val - emb_val)**2 for x_val, emb_val in zip(x, emb))
            for emb in self.embeddings
        ]
        indices = distances.index(min(distances))
        quantized = self.embeddings[indices]
        
        return quantized, indices, x

class Encoder:
    def __init__(self, input_dim, hidden_dims, latent_dim):
        self.input_dim = input_dim
        self.hidden_dims = hidden_dims
        self.latent_dim = latent_dim
        
        # Initialize weights for each layer
        self.weights = []
        prev_dim = input_dim
        for hidden_dim in hidden_dims:
            layer_weights = [
                [random.uniform(-1, 1) for _ in range(hidden_dim)]
                for _ in range(prev_dim)
            ]
            self.weights.append(layer_weights)
            prev_dim = hidden_dim
        
        # Final layer to latent dim
        self.final_weights = [
            [random.uniform(-1, 1) for _ in range(latent_dim)]
            for _ in range(prev_dim)
        ]
    
    def forward(self, x):
        # Simple linear transformation
        current = x
        for layer_weights in self.weights:
            # Compute new layer
            next_layer = [
                sum(current[j] * layer_weights[j][i] for j in range(len(current)))
                for i in range(len(layer_weights[0]))
            ]
            current = next_layer
        
        # Final layer to latent dim
        latent = [
            sum(current[j] * self.final_weights[j][i] for j in range(len(current)))
            for i in range(len(self.final_weights[0]))
        ]
        
        return latent

class Decoder:
    def __init__(self, latent_dim, hidden_dims, output_dim):
        self.latent_dim = latent_dim
        self.hidden_dims = hidden_dims
        self.output_dim = output_dim
        
        # Initialize weights for each layer
        self.weights = []
        prev_dim = latent_dim
        for hidden_dim in hidden_dims:
            layer_weights = [
                [random.uniform(-1, 1) for _ in range(hidden_dim)]
                for _ in range(prev_dim)
            ]
            self.weights.append(layer_weights)
            prev_dim = hidden_dim
        
        # Final layer to output dim
        self.final_weights = [
            [random.uniform(-1, 1) for _ in range(output_dim)]
            for _ in range(prev_dim)
        ]
    
    def forward(self, x):
        # Reconstruction transformation
        current = x
        for layer_weights in self.weights:
            # Compute new layer
            next_layer = [
                sum(current[j] * layer_weights[j][i] for j in range(len(current)))
                for i in range(len(layer_weights[0]))
            ]
            current = next_layer
        
        # Final layer to output dim
        output = [
            sum(current[j] * self.final_weights[j][i] for j in range(len(current)))
            for i in range(len(self.final_weights[0]))
        ]
        
        return output

class EntropyEncoder:
    def __init__(self, latent_dim):
        self.weights1 = [random.uniform(-1, 1) for _ in range(latent_dim // 2)]
        self.weights2 = [random.uniform(-1, 1) for _ in range(1)]
    
    def forward(self, indices, quantized):
        # Simple entropy estimation
        transformed = [
            sum(q * w for q, w in zip(quantized, self.weights1))
            for _ in range(len(quantized))
        ]
        entropy = sum(w * t for w, t in zip(self.weights2, transformed))
        return -math.log(abs(entropy) + 1e-7)

class NeuralCompressor:
    def __init__(self, input_dim, hidden_dims, latent_dim, num_embeddings, beta=1.0):
        self.encoder = Encoder(input_dim, hidden_dims, latent_dim)
        self.decoder = Decoder(latent_dim, hidden_dims[::-1], input_dim)
        self.quantizer = Quantizer(num_embeddings, latent_dim)
        self.entropy_encoder = EntropyEncoder(latent_dim)
        self.beta = beta
    
    def forward(self, x):
        # Encode
        z = self.encoder.forward(x)
        
        # Quantize
        quantized, indices, _ = self.quantizer.forward(z)
        
        # Decode
        x_rec = self.decoder.forward(quantized)
        
        # Calculate distortion (MSE)
        distortion = sum((x_val - rec_val)**2 for x_val, rec_val in zip(x, x_rec)) / len(x)
        
        # Calculate rate (entropy)
        rate = self.entropy_encoder.forward(indices, quantized)
        
        # Total loss
        loss = distortion + self.beta * rate
        
        return loss, distortion, rate

def main():
    # Hyperparameters
    input_dim = 10
    hidden_dims = [16, 32, 64]
    latent_dim = 8
    num_embeddings = 512
    beta = 1.0
    
    # Create compressor
    compressor = NeuralCompressor(
        input_dim=input_dim, 
        hidden_dims=hidden_dims, 
        latent_dim=latent_dim, 
        num_embeddings=num_embeddings, 
        beta=beta
    )
    
    # Generate sample data
    sample_data = [
        [random.uniform(-1, 1) for _ in range(input_dim)] 
        for _ in range(100)
    ]
    
    # Training simulation
    for epoch in range(5):
        total_loss = 0
        for data_point in sample_data:
            loss, distortion, rate = compressor.forward(data_point)
            total_loss += loss
        print(f"Epoch {epoch+1}, Total Loss: {total_loss}")

if __name__ == "__main__":
    main()
````

````python

import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models

# Load and preprocess your image data
# For simplicity, let's assume you have a dataset of images and their corresponding labels
# You can replace this with your own dataset or generate synthetic data

# Example: Load images and labels
images = np.random.rand(1000, 28, 28, 1)  # Replace with your image data
labels = np.random.randint(0, 10, 1000)  # Replace with your label data

# Split the data into training and validation sets
train_images = images[:800]
train_labels = labels[:800]
val_images = images[800:]
val_labels = labels[800:]

# Define the neural network architecture
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')  # Assuming 10 classes
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(train_images, train_labels, epochs=10, validation_data=(val_images, val_labels))

# Save the trained model
model.save('image_compression_model.h5')

# Load the trained model for inference
loaded_model = tf.keras.models.load_model('image_compression_model.h5')

# Example usage: Compress an image
test_image = np.random.rand(28, 28, 1)  # Replace with your test image
compressed_image = loaded_model.predict(test_image)

# The compressed_image now contains the compressed representation of the test image
````
