##  Lower Bound and EM Algorithm Analysis

###  Lower Bound for EM Algorithm

We define the lower bound $B[\{q_i(h_i)\}, \theta]$ to be:

$$B[\{q_i(h_i)\}, \theta] = \sum_{i=1}^I \int q_i(h_i) \log \frac{Pr(x_i, h_i|\theta)}{q_i(h_i)} dh_i$$

$$\leq \log \sum_{i=1}^I \int q_i(h_i) \frac{Pr(x_i, h_i|\theta)}{q_i(h_i)} dh_i$$

$$= \log \sum_{i=1}^I \int Pr(x_i, h_i|\theta) dh_i$$

where we have used Jensen's inequality between the first and second lines. This states that because the logarithm is a concave function, we can write:

$$\int Pr(y) \log[y]dy \leq \log \int yPr(y)dy$$

or equivalently:

$$E[\log[y]] \leq \log(E[y])$$

## The E-step

In the E-step, we update the bound $B[\{q_i(h_i)\}, \theta]$ with respect to the distributions $q_i(h_i)$:

$$B[\{q_i(h_i)\}, \theta] = \sum_{i=1}^I \int q_i(h_i) \log \frac{Pr(x_i, h_i|\theta)}{q_i(h_i)} dh_i$$

$$= \sum_{i=1}^I \int q_i(h_i) \log \frac{Pr(h_i|x_i, \theta)Pr(x_i|\theta)}{q_i(h_i)} dh_i$$

$$= \sum_{i=1}^I \log[Pr(x_i|\theta)] - \sum_{i=1}^I \int q_i(h_i) \log \frac{q_i(h_i)}{Pr(h_i|x_i, \theta)} dh_i$$

The optimal distributions $\hat{q}_i(h_i)$ satisfy:

$$\hat{q}_i(h_i) = \text{argmin} \int -q_i(h_i) \log \frac{Pr(h_i|x_i, \theta)}{q_i(h_i)} dh_i$$

## The M-step

In the M-step we maximize:

$$\theta^{[t]} = \text{argmax}_\theta B[\{q_i^{[t]}(h_i)\}, \theta]$$

$$= \text{argmax}_\theta \sum_{i=1}^I \int q_i^{[t]}(h_i) \log \frac{Pr(x_i, h_i|\theta)}{q_i^{[t]}(h_i)} dh_i$$

$$= \text{argmax}_\theta \sum_{i=1}^I \int q_i^{[t]}(h_i) \log[Pr(x_i, h_i|\theta)] dh_i$$

## Lower Bound for EM Algorithm and Jensen's Inequality

### Theory and Mathematical Foundation

#### Lower Bound Definition



Fig.25 Jensenâ€™s inequality for the logarithmic function (discrete case). a) Taking a weighted average of examples E[y] and passing them through the log function. b) Passing the samples through the log function and taking a weighted average E[log[y]]. The latter case always produces a smaller value than the former (E[log[y]] â‰¤ log(E[y])): higher valued examples are relatively compressed by the concave log function.
The EM algorithm's lower bound, denoted as $B[\{q_i(h_i)\}, \theta]$, is defined as:

$$B[\{q_i(h_i)\}, \theta] = \sum_{i=1}^I \int q_i(h_i) \log \frac{Pr(x_i, h_i|\theta)}{q_i(h_i)} dh_i$$

This can be manipulated using Jensen's inequality to show:

$$\leq \log \sum_{i=1}^I \int q_i(h_i) \frac{Pr(x_i, h_i|\theta)}{q_i(h_i)} dh_i$$

$$= \log \sum_{i=1}^I \int Pr(x_i, h_i|\theta) dh_i$$

### Detailed Explanation

1. **Initial Expression**:
   - The bound is a sum over all data points $i$ from 1 to $I$
   - For each point, we integrate over the hidden variables $h_i$
   - $q_i(h_i)$ represents our current estimate of the distribution of hidden variables
   - $Pr(x_i, h_i|\theta)$ is the joint probability of observed and hidden variables

2. **Jensen's Inequality Application**:
   The key mathematical principle here is Jensen's inequality, which states:

   $$\int Pr(y) \log[y]dy \leq \log \int yPr(y)dy$$

   Or in expectation notation:
   
   $$E[\log[y]] \leq \log(E[y])$$

### Visual Interpretation

The inequality is illustrated for the discrete case in two scenarios:

1. **Scenario A**: 
   - Take weighted average of examples $E[y]$
   - Pass the result through log function
   - This gives the larger value

2. **Scenario B**:
   - Pass individual samples through log function first
   - Take weighted average of results $E[\log[y]]$
   - This gives the smaller value

### Key Properties

1. **Concavity**:
   - The logarithm function is concave
   - This means it curves downward
   - Higher values are relatively compressed compared to lower values

2. **Inequality Direction**:
   - Always: $E[\log[y]] \leq \log(E[y])$
   - The difference between these values is the "gap" in Jensen's inequality

3. **Practical Implications**:
   - This inequality is crucial for the EM algorithm
   - It provides a guaranteed lower bound on the log-likelihood
   - The bound is used to iteratively improve parameter estimates

### Mathematical Note

The inequality holds because of the logarithm's concavity, which means:
- For any two points on the curve
- Any point on the line connecting them
- The curve always lies above this line

This property is fundamental to why the EM algorithm is guaranteed to converge to at least a local maximum.

````python
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt

class EMAlgorithm:
    def __init__(self, n_components=2, max_iter=100, tol=1e-4):
        """
        Initialize EM Algorithm for Gaussian Mixture Model
        
        Parameters:
        -----------
        n_components: int
            Number of Gaussian components
        max_iter: int
            Maximum number of iterations
        tol: float
            Convergence tolerance
        """
        self.n_components = n_components
        self.max_iter = max_iter
        self.tol = tol
        
    def initialize_parameters(self, X):
        """
        Initialize model parameters
        
        Parameters:
        -----------
        X: array-like of shape (n_samples, n_features)
            Training data
        """
        n_samples, n_features = X.shape
        
        # Initialize mixing coefficients
        self.weights_ = np.ones(self.n_components) / self.n_components
        
        # Initialize means randomly from data points
        random_idx = np.random.choice(n_samples, self.n_components, replace=False)
        self.means_ = X[random_idx]
        
        # Initialize covariances as identity matrices
        self.covs_ = np.array([np.eye(n_features) for _ in range(self.n_components)])
        
    def e_step(self, X):
        """
        E-step: Compute responsibilities (posterior probabilities)
        
        Parameters:
        -----------
        X: array-like of shape (n_samples, n_features)
            Training data
            
        Returns:
        --------
        responsibilities: array-like of shape (n_samples, n_components)
            Posterior probabilities of each component for each sample
        """
        n_samples = X.shape[0]
        responsibilities = np.zeros((n_samples, self.n_components))
        
        # Compute likelihoods for each component
        for k in range(self.n_components):
            responsibilities[:, k] = self.weights_[k] * multivariate_normal.pdf(
                X, mean=self.means_[k], cov=self.covs_[k]
            )
            
        # Normalize responsibilities
        responsibilities /= responsibilities.sum(axis=1, keepdims=True)
        return responsibilities
    
    def m_step(self, X, responsibilities):
        """
        M-step: Update parameters
        
        Parameters:
        -----------
        X: array-like of shape (n_samples, n_features)
            Training data
        responsibilities: array-like of shape (n_samples, n_components)
            Posterior probabilities from E-step
        """
        n_samples = X.shape[0]
        
        # Update weights (mixing coefficients)
        Nk = responsibilities.sum(axis=0)
        self.weights_ = Nk / n_samples
        
        # Update means
        self.means_ = np.dot(responsibilities.T, X) / Nk[:, np.newaxis]
        
        # Update covariances
        for k in range(self.n_components):
            diff = X - self.means_[k]
            self.covs_[k] = np.dot(responsibilities[:, k] * diff.T, diff) / Nk[k]
            
    def compute_lower_bound(self, X, responsibilities):
        """
        Compute the lower bound (log-likelihood)
        
        Parameters:
        -----------
        X: array-like of shape (n_samples, n_features)
            Training data
        responsibilities: array-like of shape (n_samples, n_components)
            Posterior probabilities
            
        Returns:
        --------
        float: Lower bound value
        """
        n_samples = X.shape[0]
        lower_bound = 0
        
        for k in range(self.n_components):
            # Compute log probability for each component
            log_prob = multivariate_normal.logpdf(
                X, mean=self.means_[k], cov=self.covs_[k]
            )
            lower_bound += np.sum(responsibilities[:, k] * (
                np.log(self.weights_[k]) + log_prob
            ))
            
        return lower_bound
    
    def fit(self, X):
        """
        Fit the model to the data
        
        Parameters:
        -----------
        X: array-like of shape (n_samples, n_features)
            Training data
        """
        self.initialize_parameters(X)
        
        prev_lower_bound = -np.inf
        
        for iteration in range(self.max_iter):
            # E-step
            responsibilities = self.e_step(X)
            
            # M-step
            self.m_step(X, responsibilities)
            
            # Compute lower bound
            lower_bound = self.compute_lower_bound(X, responsibilities)
            
            # Check convergence
            change = lower_bound - prev_lower_bound
            if iteration > 0 and abs(change) < self.tol:
                print(f"Converged at iteration {iteration}")
                break
                
            prev_lower_bound = lower_bound
            
    def predict(self, X):
        """
        Predict cluster labels for new data
        
        Parameters:
        -----------
        X: array-like of shape (n_samples, n_features)
            New data points
            
        Returns:
        --------
        array-like of shape (n_samples,)
            Predicted cluster labels
        """
        responsibilities = self.e_step(X)
        return np.argmax(responsibilities, axis=1)

# Example usage and visualization
def generate_sample_data(n_samples=300):
    """Generate sample data from two Gaussians"""
    np.random.seed(42)
    
    # First Gaussian
    X1 = np.random.multivariate_normal(
        mean=[-2, -2],
        cov=[[1, 0.5], [0.5, 1]],
        size=n_samples//2
    )
    
    # Second Gaussian
    X2 = np.random.multivariate_normal(
        mean=[2, 2],
        cov=[[1.5, -0.5], [-0.5, 1.5]],
        size=n_samples//2
    )
    
    # Combine data
    X = np.vstack([X1, X2])
    return X

def plot_results(X, labels, means):
    """Plot clustering results"""
    plt.figure(figsize=(10, 6))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
    plt.scatter(means[:, 0], means[:, 1], c='red', marker='x', s=200, linewidths=3)
    plt.title('EM Algorithm Clustering Results')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()
````

````python
import math
import random
import matplotlib.pyplot as plt

class EMAlgorithm:
    def __init__(self, n_components=2, max_iter=100, tol=1e-4):
        self.n_components = n_components
        self.max_iter = max_iter
        self.tol = tol

    def initialize_parameters(self, X):
        n_samples, n_features = len(X), len(X[0])  # Get dimensions without NumPy

        self.weights_ = [1.0 / self.n_components] * self.n_components  # List for weights

        # Initialize means randomly from data points (without NumPy choice)
        random_idx = random.sample(range(n_samples), self.n_components)
        self.means_ = [X[i] for i in random_idx]

        # Initialize covariances as identity matrices (lists of lists)
        self.covs_ = [[ [1.0 if i == j else 0.0 for j in range(n_features)] for i in range(n_features)] for _ in range(self.n_components)]

    def multivariate_normal_pdf(self, x, mean, cov): # Manual multivariate normal calculation
      n_features = len(x)
      det_cov = 1.0 # Determinant calculation without numpy (for 2x2, can be generalized)
      if n_features==2:
        det_cov = cov[0][0]*cov[1][1]-cov[0][1]*cov[1][0]
      elif n_features==1:
        det_cov = cov[0][0]

      inv_cov = [[0.0]*n_features for _ in range(n_features)] # Inverse calculation without numpy (for 2x2, can be generalized)
      if n_features==2:
        inv_cov = [[cov[1][1]/det_cov, -cov[0][1]/det_cov],[-cov[0][1]/det_cov, cov[0][0]/det_cov]]
      elif n_features==1:
        inv_cov = [[1.0/det_cov]]

      diff = [x[i] - mean[i] for i in range(n_features)]
      exponent = 0.0
      for i in range(n_features):
        for j in range(n_features):
          exponent += diff[i] * inv_cov[i][j] * diff[j]
      return (1.0 / math.sqrt((2 * math.pi)**n_features * det_cov)) * math.exp(-0.5 * exponent)

    def e_step(self, X):
        n_samples = len(X)
        responsibilities = [[0.0] * self.n_components for _ in range(n_samples)]

        for k in range(self.n_components):
            for i in range(n_samples):
                responsibilities[i][k] = self.weights_[k] * self.multivariate_normal_pdf(X[i], self.means_[k], self.covs_[k])

        # Normalize responsibilities (manually)
        for i in range(n_samples):
            row_sum = sum(responsibilities[i])
            for k in range(self.n_components):
                responsibilities[i][k] /= row_sum

        return responsibilities

    def m_step(self, X, responsibilities):
        n_samples = len(X)

        # Update weights
        Nk = [sum(responsibilities[i][k] for i in range(n_samples)) for k in range(self.n_components)]
        self.weights_ = [Nk[k] / n_samples for k in range(self.n_components)]

        # Update means
        n_features = len(X[0])
        self.means_ = [[0.0] * n_features for _ in range(self.n_components)]
        for k in range(self.n_components):
            for j in range(n_features):
                for i in range(n_samples):
                    self.means_[k][j] += responsibilities[i][k] * X[i][j]
                self.means_[k][j] /= Nk[k]

        # Update covariances
        self.covs_ = [[[0.0] * n_features for _ in range(n_features)] for _ in range(self.n_components)]
        for k in range(self.n_components):
            for i in range(n_samples):
                diff = [X[i][j] - self.means_[k][j] for j in range(n_features)]
                for r in range(n_features):
                  for c in range(n_features):
                    self.covs_[k][r][c] += responsibilities[i][k]*diff[r]*diff[c]
                
            for r in range(n_features):
              for c in range(n_features):
                self.covs_[k][r][c] /= Nk[k]
            

    def compute_lower_bound(self, X, responsibilities):
        n_samples = len(X)
        lower_bound = 0.0

        for k in range(self.n_components):
            for i in range(n_samples):
              log_prob = math.log(self.multivariate_normal_pdf(X[i], self.means_[k], self.covs_[k]))
              lower_bound += responsibilities[i][k] * (math.log(self.weights_[k]) + log_prob)

        return lower_bound

    def fit(self, X):
        self.initialize_parameters(X)

        prev_lower_bound = float('-inf')

        for iteration in range(self.max_iter):
            responsibilities = self.e_step(X)
            self.m_step(X, responsibilities)
            lower_bound = self.compute_lower_bound(X, responsibilities)

            change = lower_bound - prev_lower_bound
            if iteration > 0 and abs(change) < self.tol:
                print(f"Converged at iteration {iteration}")
                break

            prev_lower_bound = lower_bound

    def predict(self, X):
        responsibilities = self.e_step(X)
        return [row.index(max(row)) for row in responsibilities]  # Manual argmax


# Example usage and visualization (without NumPy)
def generate_sample_data(n_samples=300):
    random.seed(42)

    X1 = [[0.0]*2 for _ in range(n_samples//2)]
    for i in range(n_samples//2):
        X1[i] = [random.gauss(-2, 1), random.gauss(-2, 1)]

    X2 = [[0.0]*2 for _ in range(n_samples//2)]
    for i in range(n_samples//2):
        X2[i] = [random.gauss(2, 1.5), random.gauss(2, 1.5)]

    X = X1 + X2
    return X

def plot_results(X, labels, means):
    plt.figure(figsize=(10, 6))
    plt.scatter([x[0] for x in X], [x[1] for x in X], c=labels, cmap='viridis', alpha=0.6)
    plt.scatter([m[0] for m in means], [m[1] for m in means], c='red', marker='x', s=200, linewidths=3)
    plt.title('EM Algorithm Clustering Results (No NumPy)')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()


X = generate_sample_data()
em_algorithm = EMAlgorithm(n_components=2)
em_algorithm.fit(X)
labels = em_algorithm.predict(X)
plot_results(X, labels, em_algorithm.means_)
````

## The E-step

In the E-step, we update the bound $B[\{q_i(h_i)\}, \theta]$ with respect to the distributions $q_i(h_i)$. To see how to do this, we manipulate the expression for the bound as follows:
## Derivation of Optimal $q_i$$(h_i)$ in the E-step

This section details the derivation of the optimal distribution $q_i(h_i)$ in the E-step of the Expectation-Maximization (EM) algorithm. The goal is to maximize the lower bound $B[\{q_i(h_i)\}, \theta]$ with respect to the distributions $q_i(h_i)$.

**1. Starting with the Lower Bound:**

The lower bound is defined as:

$$
B[\{q_i(h_i)\}, \theta] = \sum_{i=1}^I \int q_i(h_i) \log \frac{P(x_i, h_i | \theta)}{q_i(h_i)} dh_i \quad (7.46)
$$

**2. Applying Conditional Probability:**

Using the definition of conditional probability, $P(x_i, h_i | \theta) = P(h_i | x_i, \theta)P(x_i | \theta)$, we can rewrite the bound as:

$$
B[\{q_i(h_i)\}, \theta] = \sum_{i=1}^I \int q_i(h_i) \log \frac{P(h_i | x_i, \theta) P(x_i | \theta)}{q_i(h_i)} dh_i
$$

**3. Separating the Logarithm:**

The logarithm inside the integral can be separated:

$$
B[\{q_i(h_i)\}, \theta] = \sum_{i=1}^I \int q_i(h_i) \left( \log P(x_i | \theta) + \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} \right) dh_i
$$

**4. Distributing the Integral:**

The integral can be distributed across the terms:

$$
B[\{q_i(h_i)\}, \theta] = \sum_{i=1}^I \int q_i(h_i) \log P(x_i | \theta) dh_i + \sum_{i=1}^I \int q_i(h_i) \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} dh_i
$$

**5. Simplifying the First Term:**

Since $\log P(x_i | \theta)$ is independent of $h_i$, it can be factored out of the integral:

$$
B[\{q_i(h_i)\}, \theta] = \sum_{i=1}^I \log P(x_i | \theta) \int q_i(h_i) dh_i + \sum_{i=1}^I \int q_i(h_i) \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} dh_i
$$

Because $q_i(h_i)$ is a probability distribution, its integral over $h_i$ is equal to 1:

$$
B[\{q_i(h_i)\}, \theta] = \sum_{i=1}^I \log P(x_i | \theta) + \sum_{i=1}^I \int q_i(h_i) \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} dh_i
$$

**6. Optimization Focus:**

The term $\sum_{i=1}^I \log P(x_i | \theta)$ is constant with respect to $q_i(h_i)$.  Therefore, maximizing the lower bound with respect to $q_i(h_i)$ is equivalent to maximizing the second term (or minimizing its negative):

$$
\underset{q_i(h_i)}{\text{argmax}} \, B[\{q_i(h_i)\}, \theta] \Leftrightarrow \underset{q_i(h_i)}{\text{argmin}} \sum_{i=1}^I \int q_i(h_i) \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} dh_i
$$

**7. Kullback-Leibler Divergence:**

Recognizing the Kullback-Leibler (KL) divergence:

$$
KL(q_i(h_i) || P(h_i | x_i, \theta)) = \int q_i(h_i) \log \frac{q_i(h_i)}{P(h_i | x_i, \theta)} dh_i
$$

Minimizing the KL divergence between $q_i(h_i)$ and $P(h_i | x_i, \theta)$ makes $q_i(h_i)$ as similar as possible to $P(h_i | x_i, \theta)$.

**8. Optimal Distribution:**

The KL divergence is minimized (and is equal to zero) when the two distributions are identical.  Therefore, the optimal $q_i(h_i)$ is:

$$
\hat{q}_i(h_i) = P(h_i | x_i, \theta)
$$

This is the core result of the E-step.  $q_i(h_i)$ is set to the posterior distribution of the hidden variable $h_i$ given the observed data $x_i$ and the current parameters $\theta$. This posterior is calculated using Bayes' rule:

$$
P(h_i | x_i, \theta) = \frac{P(x_i | h_i, \theta) P(h_i)}{P(x_i | \theta)}
$$

In summary, the E-step computes the posterior distribution of the hidden variables given the data and current parameters. This posterior becomes the optimal $q_i(h_i)$ that maximizes the lower bound.
$$
\begin{aligned}
B[\{q_i(h_i)\}, \theta] &= \sum_{i=1}^I \int q_i(h_i) \log \frac{P(x_i, h_i | \theta)}{q_i(h_i)} dh_i \\
&= \sum_{i=1}^I \int q_i(h_i) \log \frac{P(h_i | x_i, \theta) P(x_i | \theta)}{q_i(h_i)} dh_i \\
&= \sum_{i=1}^I \int q_i(h_i) \log [P(x_i | \theta)] dh_i - \sum_{i=1}^I \int q_i(h_i) \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} dh_i \\
&= \sum_{i=1}^I \log [P(x_i | \theta)] - \sum_{i=1}^I \int q_i(h_i) \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} dh_i \quad (7.46)
\end{aligned}
$$

where the hidden variables from the first term are integrated out between the last two lines. The first term in this expression is constant with respect to the distributions $q_i(h_i)$, and so to optimize the bound, we must find the distributions $\hat{q}_i(h_i)$ that satisfy

$$
\begin{aligned}
\hat{q}_i(h_i) &= \underset{q_i(h_i)}{\text{argmax}} \left( - \int q_i(h_i) \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} dh_i \right) \\
&= \underset{q_i(h_i)}{\text{argmax}} \int q_i(h_i) \log \frac{q_i(h_i)}{P(h_i | x_i, \theta)} dh_i \\
&= \underset{q_i(h_i)}{\text{argmin}} \int q_i(h_i) \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} dh_i \quad (7.47)
\end{aligned}
$$

This expression is known as the Kullback-Leibler (KL) divergence between $q_i(h_i)$ and $P(h_i | x_i, \theta)$. It is a measure of "distance" between probability distributions. We can use the inequality $\log[y] \le y - 1$ to show that this cost function (including the minus sign) is always positive:

$$
\begin{aligned}
\int q_i(h_i) \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} dh_i &\le \int q_i(h_i) \left( \frac{P(h_i | x_i, \theta)}{q_i(h_i)} - 1 \right) dh_i \\
&= \int P(h_i | x_i, \theta) dh_i - \int q_i(h_i) dh_i \\
&= 1 - 1 = 0 \quad (7.48)
\end{aligned}
$$

implying that when we reintroduce the minus sign, the cost function must be positive. So the criteria in Equation 7.46 will be maximized when the KL divergence is zero. This value is reached when $q_i(h_i) = P(h_i | x_i, \theta)$, so that

$$
\begin{aligned}
\int q_i(h_i) \log \frac{P(h_i | x_i, \theta)}{q_i(h_i)} dh_i &= \int P(h_i | x_i, \theta) \log \frac{P(h_i | x_i, \theta)}{P(h_i | x_i, \theta)} dh_i \\
&= \int P(h_i | x_i, \theta) \log [1] dh_i = 0 \quad (7.49)
\end{aligned}
$$

In other words, to maximize the bound with respect to $q_i(h_i)$, we set this to be the posterior distribution $P(h_i | x_i, \theta)$ over the hidden variables $h_i$ given the current set of parameters. In practice, this is computed using Bayes' rule:

$$
P(h_i | x_i, \theta) = \frac{P(x_i | h_i, \theta) P(h_i)}{P(x_i)} \quad (7.50)
$$

So the E-step consists of computing the posterior distribution for each hidden variable using Bayes' rule.

````python
import math
import random
import matplotlib.pyplot as plt

# Data and parameters
y = 4.3  # Observed data
sigma = 1.46  # Langevin diffusion step size
n_samples = 10000  # Number of MCMC samples
n_burnin = 1000  # Burn-in period

# Initialize chains
theta = [[0.0, 0.0] for _ in range(n_samples)]
theta[0] = [random.gauss(0, 1), random.gauss(0, 1)]  # Initial values from prior

# Langevin Metropolis-Hastings algorithm
for t in range(1, n_samples):
    theta_current = theta[t - 1]

    # Langevin proposal (manual calculation without numpy)
    mean_proposal = [
        (sigma**2 / 2) * ((2*y - 2*theta_current[1] - 5*theta_current[0]) / 8) + theta_current[0],
        (sigma**2 / 2) * ((4*y - 2*theta_current[0] - 5*theta_current[1]) / 8) + theta_current[1]
    ]

    theta_proposal = [random.gauss(mean_proposal[0], sigma), random.gauss(mean_proposal[1], sigma)]

    # Calculate acceptance probability (using log probabilities for numerical stability)
    log_pi_current = -0.5 * (5/4 * theta_current[0]**2 + 5/4 * theta_current[1]**2 + 
                            theta_current[0]*theta_current[1]/2 - (theta_current[0] + theta_current[1])*y)
    log_pi_proposal = -0.5 * (5/4 * theta_proposal[0]**2 + 5/4 * theta_proposal[1]**2 + 
                            theta_proposal[0]*theta_proposal[1]/2 - (theta_proposal[0] + theta_proposal[1])*y)

    log_alpha = log_pi_proposal - log_pi_current
    alpha = math.exp(log_alpha)

    # Accept or reject
    if random.random() < alpha:
        theta[t] = theta_proposal[:]  # Create a copy
    else:
        theta[t] = theta_current[:]  # Create a copy

# Burn-in removal
theta_after_burnin = theta[n_burnin:]

# Calculate empirical averages (manual calculation without numpy)
theta_means = []
cumulative_sum_0 = 0.0
for i in range(len(theta_after_burnin)):
    cumulative_sum_0 += theta_after_burnin[i][0]
    theta_means.append(cumulative_sum_0 / (i + 1))

# Calculate quantiles for confidence intervals (manual calculation without numpy)
theta_values_for_quantile = [val[0] for val in theta_after_burnin]
theta_values_for_quantile.sort()  # Sort to find quantiles
lower_index = int(0.05 * len(theta_values_for_quantile))
upper_index = int(0.95 * len(theta_values_for_quantile))

quantiles = [theta_values_for_quantile[lower_index], theta_values_for_quantile[upper_index]]

# --- IID Simulation (without numpy) ---
iid_samples = [[random.gauss(0, 1), random.gauss(0, 1)] for _ in range(len(theta_after_burnin))]
iid_means = []
cumulative_sum_iid = 0.0
for i in range(len(iid_samples)):
    cumulative_sum_iid += iid_samples[i][0]
    iid_means.append(cumulative_sum_iid / (i + 1))

# Plotting (using lists with matplotlib)
plt.figure(figsize=(10, 6))
plt.plot(theta_means, label='Langevin Metropolis-Hastings')
plt.plot(iid_means, label='iid Simulation', linestyle='--')
plt.axhline(quantiles[0], color='gray', linestyle='--', alpha=0.7)
plt.axhline(quantiles[1], color='gray', linestyle='--', alpha=0.7)

plt.xlabel('Iteration')
plt.ylabel('Theta 1')
plt.title('Convergence of Empirical Averages')
plt.legend()
plt.grid(True)
plt.show()

print(f"90% Confidence Interval for Theta 1 (Langevin): [{quantiles[0]:.3f}, {quantiles[1]:.3f}]")
````

````python
import random
import math

def gaussian_pdf(x, mean, std_dev):
    return (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))

def e_step(data, means, std_devs, weights):
    responsibilities = []
    for i in range(len(data)):
        responsibility = []
        for j in range(len(means)):
            responsibility.append(weights[j] * gaussian_pdf(data[i], means[j], std_devs[j]))
        responsibility_sum = sum(responsibility)
        if responsibility_sum == 0:
            responsibility_sum = 1e-10  # Add a small epsilon to avoid division by zero
        responsibility = [r / responsibility_sum for r in responsibility]
        responsibilities.append(responsibility)
    return responsibilities

def m_step(data, responsibilities):
    new_means = []
    new_std_devs = []
    new_weights = []
    for j in range(len(means)):
        numerator = 0
        denominator = 0
        for i in range(len(data)):
            numerator += responsibilities[i][j] * data[i]
            denominator += responsibilities[i][j]
        new_means.append(numerator / denominator)
        new_std_devs.append(math.sqrt(sum([responsibilities[i][j] * (data[i] - new_means[j]) ** 2 for i in range(len(data))]) / denominator))
        new_weights.append(sum([responsibilities[i][j] for i in range(len(data))]) / len(data))
    return new_means, new_std_devs, new_weights

# Example data
data = [random.gauss(0, 1) for _ in range(1000)]

# Initial parameters
means = [0, 1]
std_devs = [1, 1]
weights = [0.5, 0.5]

# EM algorithm
converged = False
while not converged:
    responsibilities = e_step(data, means, std_devs, weights)
    new_means, new_std_devs, new_weights = m_step(data, responsibilities)
    if all(abs(new_means[i] - means[i]) < 1e-6 and abs(new_std_devs[i] - std_devs[i]) < 1e-6 and abs(new_weights[i] - weights[i]) < 1e-6 for i in range(len(means))):
        converged = True
    means = new_means
    std_devs = new_std_devs
    weights = new_weights

print("Final means:", means)
print("Final standard deviations:", std_devs)
print("Final weights:", weights)
````

# 7.8.3 The M-step

In the M-step, we maximize the bound with respect to the parameters $\theta$ to find the optimal parameters. The optimization problem can be written as:

$$
\hat{\theta} = \arg\max_{\theta} B[\{q_i(h_i)\}, \theta]
$$

where $B[\{q_i(h_i)\}, \theta]$ is the lower bound derived in the previous section.

By simplifying the lower bound and omitting terms that do not depend on $\theta$, we obtain:

$$
\hat{\theta} = \arg\max_{\theta} \sum_{i=1}^I \int q_i(h_i) \log P(x_i, h_i | \theta) dh_i
$$

This optimization problem is equivalent to maximizing the expected complete-data log-likelihood, which is a common objective in the M-step of the EM algorithm.

$$

````python
# dode for face detetio
import random
import math
import numpy as np
from scipy.stats import multivariate_normal

def generate_data(num_samples, num_classes, means, covariances):
    data = []
    labels = []
    for i in range(num_classes):
        for _ in range(num_samples // num_classes):
            x = np.random.multivariate_normal(means[i], covariances[i])
            data.append(x)
            labels.append(i)
    return np.array(data), np.array(labels)

def fit_mixture_of_gaussians(data, labels, num_components):
    means = []
    covariances = []
    weights = []
    for i in range(num_classes):
        class_data = data[labels == i]
        class_mean = np.mean(class_data, axis=0)
        class_covariance = np.cov(class_data.T)
        class_weight = len(class_data) / len(data)
        means.append(class_mean)
        covariances.append(class_covariance)
        weights.append(class_weight)
    return means, covariances, weights

def classify_face(test_data, means, covariances, weights):
    posterior_probabilities = []
    for i in range(num_classes):
        posterior_prob = weights[i] * multivariate_normal.pdf(test_data, mean=means[i], cov=covariances[i])
        posterior_probabilities.append(posterior_prob)
    posterior_probabilities = np.array(posterior_probabilities)
    posterior_probabilities = posterior_probabilities / np.sum(posterior_probabilities)
    return np.argmax(posterior_probabilities)

# Example parameters
num_samples = 1000
num_classes = 2
num_components = 10

# Generate synthetic data
means = [[0, 0], [1, 1]]
covariances = [[[1, 0], [0, 1]], [[1, 0], [0, 1]]]
data, labels = generate_data(num_samples, num_classes, means, covariances)

# Fit mixture of Gaussians model
means, covariances, weights = fit_mixture_of_gaussians(data, labels, num_components)

# Test data
test_data = np.array([[0.5, 0.5], [1.5, 1.5]])

# Classify faces
for i, test_point in enumerate(test_data):
    predicted_class = classify_face(test_point, means, covariances, weights)
    print(f"Test point {i+1}: Predicted class = {predicted_class}")
````
