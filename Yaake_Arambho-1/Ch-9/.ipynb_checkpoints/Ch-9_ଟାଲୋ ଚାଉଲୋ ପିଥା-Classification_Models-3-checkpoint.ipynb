$$
\textbf{Relevance Vector Classification}
$$

The Bayesian version of the kernel logistic regression model is powerful, but computationally expensive as it requires us to compute dot products between the new data example and all of the training examples (in the kernel function in equation 9.31). It would be more efficient if the model depended only sparsely on the training data. To achieve this, we impose a penalty for every non-zero weighted training example.

As in the relevance regression model (section 8.8), we replace the normal prior over the dual parameters $ \psi $ (equation 9.25) with a product of one-dimensional t-distributions so that:

$$
P r(\psi) = \prod_{i=1}^{I} \text{Stud}_{\psi_i} [0, 1, \nu].
$$

Applying the Bayesian approach to this model with respect to the parameters $ \Psi $ is known as \textbf{relevance vector classification}. Following the argument of section 8.6, we rewrite each Student t-distribution as a marginalization of a joint distribution:

$$
P r(\psi_i , h_i ) = \text{Norm}_{\psi_i} (0, h_i^{-1}) \cdot \text{Gam}_{h_i} \left(\frac{\nu}{2}, \frac{\nu}{2} \right).
$$

Thus, the prior can be written as:

$$
P r(\psi) = \int \prod_{i=1}^{I} \text{Norm}_{\psi} [0, H^{-1}] \prod_{d=1}^{D} \text{Gam}_{h_d} [\nu/2, \nu/2] \ dH.
$$

Now, we express the model likelihood:

$$
P r(w|X) = \int P r(w|X, \psi) P r(\psi) d\psi.
$$

Expanding further:

$$
P r(w|X) = \int \prod_{i=1}^{I} \text{Bern}_{w_i} [\text{sig}(\psi^T K[X, x_i])] \cdot \text{Norm}_{\psi} [0, H^{-1}] \cdot \prod_{d=1}^{D} \text{Gam}_{h_d} [\nu/2, \nu/2] \ dH \ d\psi.
$$

We now make two approximations. 

1. **Laplace Approximation**: We approximate the posterior distribution over $ \psi $ as a normal distribution centered at the MAP estimate $ \mu $, with covariance $ \Sigma $. Using the result:

$$
\int q(\psi) d\psi \approx q(\mu) (2\pi)^{D/2} |\Sigma|^{1/2},
$$

we obtain:

$$
P r(w|X) \approx (2\pi)^{I/2} |\Sigma|^{1/2} \int \prod_{i=1}^{I} \text{Bern}_{w_i} [\text{sig}(\mu^T K[X, x_i])] \cdot \text{Norm}_{\mu} [0, H^{-1}] \cdot \prod_{i=1}^{I} \text{Gam}_{h_i} [\nu/2, \nu/2] dH.
$$

2. **Optimization Over Hidden Variables**: Instead of integrating over $ H $, we maximize over them:

$$
P r(w|X) \approx \max_{H} (2\pi)^{I/2} |\Sigma|^{1/2} \prod_{i=1}^{I} \text{Bern}_{w_i} [\text{sig}(\mu^T K[X, x_i])] \cdot \text{Norm}_{\mu} [0, H^{-1}] \cdot \prod_{i=1}^{I} \text{Gam}_{h_i} [\nu/2, \nu/2].
$$

To learn the model, we now alternate between:

- Updating the mean and variance $ \mu $ and $ \Sigma $ of the posterior distribution.
- Updating the hidden variables $ h_i $.

This process results in a sparse solution where only a subset of training points are used in classification, making relevance vector classification highly efficient.

```python
import random
import math
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic 2D data
random.seed(42)
num_samples = 100
x_data = [random.uniform(-2, 2) for _ in range(num_samples)]
y_data = [random.uniform(-2, 2) for _ in range(num_samples)]
labels = [1 if x**2 + y**2 + random.uniform(-0.5, 0.5) < 2 else 0 for x, y in zip(x_data, y_data)]

def rbf_kernel(x1, y1, x2, y2, gamma=0.5):
    return math.exp(-gamma * ((x1 - x2)**2 + (y1 - y2)**2))

def kernel_matrix(x_data, y_data, gamma=0.5):
    """ Compute RBF Kernel Matrix """
    n = len(x_data)
    K = [[rbf_kernel(x_data[i], y_data[i], x_data[j], y_data[j], gamma) for j in range(n)] for i in range(n)]
    return K

def train_klr(K, labels, lr=0.1, epochs=100):
    """ Kernel Logistic Regression using gradient ascent """
    n = len(labels)
    weights = [random.uniform(-0.1, 0.1) for _ in range(n)]
    
    for _ in range(epochs):
        for i in range(n):
            weighted_sum = sum(K[i][j] * weights[j] for j in range(n))
            prediction = 1 / (1 + math.exp(-weighted_sum))
            gradient = (labels[i] - prediction) * K[i][i]
            weights[i] += lr * gradient
    
    return weights

def relevance_vector_selection(weights, threshold=0.05):
    """ Select relevant vectors by setting small weights to zero """
    return [w if abs(w) > threshold else 0 for w in weights]

def plot_decision_boundary(x_data, y_data, labels, weights, K, title):
    """ Plot decision boundary """
    plt.figure(figsize=(6, 6))
    sns.scatterplot(x=x_data, y=y_data, hue=labels, palette={0: 'blue', 1: 'red'}, edgecolor='black')
    
    x_min, x_max = min(x_data) - 0.5, max(x_data) + 0.5
    y_min, y_max = min(y_data) - 0.5, max(y_data) + 0.5
    
    resolution = 50
    x_range = [x_min + (x_max - x_min) * i / resolution for i in range(resolution)]
    y_range = [y_min + (y_max - y_min) * i / resolution for i in range(resolution)]
    
    grid_predictions = []
    for x in x_range:
        row = []
        for y in y_range:
            weighted_sum = sum(K[i][i] * weights[i] for i in range(len(weights)))
            prob = 1 / (1 + math.exp(-weighted_sum))
            row.append(prob)
        grid_predictions.append(row)
    
    plt.contourf(x_range, y_range, grid_predictions, levels=10, cmap='coolwarm', alpha=0.5)
    plt.title(title)
    plt.show()

# Compute kernel matrix
K = kernel_matrix(x_data, y_data, gamma=0.5)

# Train Kernel Logistic Regression
klr_weights = train_klr(K, labels)

# Train Relevance Vector Classification (by pruning weights)
rvc_weights = relevance_vector_selection(klr_weights, threshold=0.05)

# Plot comparison
plot_decision_boundary(x_data, y_data, labels, klr_weights, K, "Kernel Logistic Regression")
plot_decision_boundary(x_data, y_data, labels, rvc_weights, K, "Relevance Vector Classification")
```

# Relevance Vector Classification

The Bayesian version of the kernel logistic regression model is powerful, but computationally expensive as it requires us to compute dot products between the new data example and all of the training examples (in the kernel function in equation 9.31). It would be more efficient if the model depended only sparsely on the training data. 

To achieve this, we impose a penalty for every non-zero weighted training example. As in the relevance regression model (section 8.8), we replace the normal prior over the dual parameters $ \psi $ (equation 9.25) with a product of one-dimensional t-distributions:

$$
P r(\psi) = \prod_{i=1}^{I} \text{Stud}_{\psi_i} [0, 1, \nu] .
$$

Applying the Bayesian approach to this model with respect to the parameters \( \Psi \) is known as relevance vector classification. Following the argument of section 8.6, we re-write each Student's t-distribution as a marginalization of a joint distribution:

$$
P r(\psi_i , h_i ) = \text{Norm}_{\psi_i} \left( 0, \frac{1}{h_i} \right) \text{Gam}_{h_i} \left( \frac{\nu}{2}, \frac{\nu}{2} \right)
$$

which gives:

$$
P r(\psi) = \int \prod_{d=1}^{D} \text{Norm}_{\psi} [0, H^{-1}] \text{Gam}_{h_d} [\nu/2, \nu/2] dH.
$$

Now we can write the model likelihood as:

$$
P r(w|X) = \int \left( \prod_{i=1}^{I} \text{Bern}_{w_i} [\sigma(a_i)] \right) \text{Norm}_{\psi} [0, H^{-1}] \text{Gam}_{h_d} [\nu/2, \nu/2] dH d\psi.
$$

### Approximations

To simplify the integral, we make two approximations:

1. **Laplace Approximation:** We approximate the first two terms in the integral as a normal distribution with mean $ \mu $ and covariance $ \Sigma $ centered at the MAP parameters:

   $$
   \int q(\psi) d\psi \approx q(\mu) (2\pi)^{D/2} |\Sigma|^{1/2}
   $$

   yielding:

   $$
   P r(w|X) \approx (2\pi)^{I/2} |\Sigma|^{0.5} \prod_{i=1}^{I} \text{Bern}_{w_i} [\sigma(\mu^T K[X, x_i ])] \text{Norm}_{\mu} [0, H^{-1}] \text{Gam}_{h_i} [\nu/2, \nu/2] dH.
   $$

2. **Maximization over Hidden Variables:** Instead of integrating over $ h_i $, we maximize:

   $$
   P r(w|X) \approx \max_H \left( (2\pi)^{I/2} |\Sigma|^{0.5} \prod_{i=1}^{I} \text{Bern}_{w_i} [\sigma(\mu^T K[X, x_i ])] \text{Norm}_{\mu} [0, H^{-1}] \text{Gam}_{h_i} [\nu/2, \nu/2] \right).
   $$

### Learning

To learn the model, we alternate between updating:

1. **Mean and Variance Parameters**  
   We maximize:

   $$
   L = \sum_{i=1}^{I} \log \text{Bern}_{w_i} [\sigma(\psi^T K[X, x_i ])] + \log \text{Norm}_{\psi} [0, H^{-1}]
   $$

   using the derivatives:

   $$
   \frac{\partial L}{\partial \psi} = - \sum_{i=1}^{I} (\sigma(a_i) - w_i ) K[X, x_i ] - H \psi
   $$

   $$
   \frac{\partial^2 L}{\partial \psi^2} = - \sum_{i=1}^{I} \sigma(a_i) (1 - \sigma(a_i)) K[X, x_i ]K[x_i, X] - H.
   $$

   and set:

   $$
   \mu = \psi
   $$

   $$
   \Sigma = \left(- \frac{\partial^2 L}{\partial \psi^2} \right)^{-1} \Bigg|_{\psi = \psi}.
   $$

2. **Hidden Variables Update**  
   The hidden variables $ h_i $ are updated as:

   $$
   h_i^{\text{new}} = \frac{1 - h_i \Sigma_{ii} + \nu}{\mu_i^2 + \nu}.
   $$

   As optimization proceeds, some $ h_i $ become very large, making the prior over $ \psi_i $ concentrated around zero. This results in a sparse kernelized classifier that depends only on a subset of the data.

### Inference

For a new data example $ x^* $, we compute the posterior over activation:

$$
P r(a) = P r(\psi^T K[X^T, x^* ]) = \text{Norm}_a [\mu_a, \sigma_a^2]
$$

where:

$$
\mu_a = \mu^T K[X, x^*]
$$

$$
\sigma_a^2 = K[x^*, X] \Sigma K[X, x^*].
$$

We then approximate the predictive distribution using equation 9.19.

### Results

Fig.14 shows an example of relevance vector classification, demonstrating that the dataset can be discriminated based on just 6 out of 40 training points. This results in a considerable computational saving and provides a simpler solution that helps prevent overfitting.


```python
import math
import random
from functools import reduce

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def gaussian_kernel(x1, x2, length_scale=1.0):
    squared_diff = sum((a - b) ** 2 for a, b in zip(x1, x2))
    return math.exp(-0.5 * squared_diff / length_scale ** 2)

def compute_kernel_matrix(X):
    I = len(X)
    return [[gaussian_kernel(X[i], X[j]) for j in range(I)] for i in range(I)]

def initialize_weights(I):
    return [random.uniform(-0.1, 0.1) for _ in range(I)]

def log_likelihood(K, w, psi, H):
    likelihood = 0
    I = len(w)
    for i in range(I):
        ai = sum(psi[j] * K[i][j] for j in range(I))
        likelihood += w[i] * math.log(sigmoid(ai)) + (1 - w[i]) * math.log(1 - sigmoid(ai))
    prior = -0.5 * sum(H[i] * psi[i] ** 2 for i in range(I))
    return likelihood + prior

def compute_gradients(K, w, psi, H):
    I = len(w)
    gradient = [0] * I
    for i in range(I):
        ai = sum(psi[j] * K[i][j] for j in range(I))
        gradient[i] = -(sigmoid(ai) - w[i]) * sum(K[i][j] for j in range(I)) - H[i] * psi[i]
    return gradient

def optimize_psi(K, w, H, max_iter=100, learning_rate=0.01):
    I = len(w)
    psi = initialize_weights(I)
    for _ in range(max_iter):
        gradients = compute_gradients(K, w, psi, H)
        psi = [psi[i] + learning_rate * gradients[i] for i in range(I)]
    return psi

def update_H(psi, Sigma, nu):
    I = len(psi)
    return [(1 - Sigma[i][i] + nu) / (psi[i] ** 2 + nu) for i in range(I)]

def relevance_vector_classification(X, w, nu=1.0, max_iter=10):
    K = compute_kernel_matrix(X)
    I = len(w)
    H = [1.0] * I
    psi = optimize_psi(K, w, H)
    Sigma = [[0.1 if i == j else 0 for j in range(I)] for i in range(I)]
    for _ in range(max_iter):
        H = update_H(psi, Sigma, nu)
        psi = optimize_psi(K, w, H)
    return psi, H

def predict(X_train, psi, x_new):
    Kx = [gaussian_kernel(x_new, x_train) for x_train in X_train]
    activation = sum(psi[i] * Kx[i] for i in range(len(X_train)))
    return sigmoid(activation)

def plot_decision_boundary(X, w, psi):
    import matplotlib.pyplot as plt
    x_min, x_max = min(x[0] for x in X) - 1, max(x[0] for x in X) + 1
    y_min, y_max = min(x[1] for x in X) - 1, max(x[1] for x in X) + 1
    
    plt.figure(figsize=(8, 6))
    plt.scatter([X[i][0] for i in range(len(X)) if w[i] == 1],
                [X[i][1] for i in range(len(X)) if w[i] == 1], color='blue', label='Class 1')
    plt.scatter([X[i][0] for i in range(len(X)) if w[i] == 0],
                [X[i][1] for i in range(len(X)) if w[i] == 0], color='red', label='Class 0')
    
    for i, p in enumerate(psi):
        if abs(p) > 0.01:
            plt.scatter(X[i][0], X[i][1], color='black', s=100, edgecolors='yellow')
    
    plt.legend()
    plt.show()

if __name__ == "__main__":
    X_train = [[random.uniform(0, 1), random.uniform(0, 1)] for _ in range(40)]
    w_train = [1 if x[0] + x[1] > 1 else 0 for x in X_train]
    
    psi, H = relevance_vector_classification(X_train, w_train)
    plot_decision_boundary(X_train, w_train, psi)
    
    x_test = [0.5, 0.5]
    print("Prediction for new point:", predict(X_train, psi, x_test))
```


## Incremental Fitting and Boosting for Logistic Regression

## Introduction

Incremental fitting is a greedy optimization approach that builds a sparse model by sequentially adding the most informative basis functions. The approach constructs the activation function step by step while optimizing parameters at each stage.

## Logistic Regression Model

The probability of class label $ w_i $ given input $ x_i $ follows:

$$ P(w_i | x_i) = \text{Bern}(w_i | \sigma(a_i)) $$

where $ \sigma(\cdot) $ is the logistic sigmoid function:

$$ \sigma(a) = \frac{1}{1 + e^{-a}} $$

The activation function $ a_i $ is expressed as:

$$ a_i = \phi_0 + \sum_{k=1}^{K} \phi_k f(x_i, \xi_k) $$

where:
- $ \phi_k $ are the model parameters.
- $ f(x_i, \xi_k) $ is a nonlinear transformation function (e.g., RBF or arctan).

### Example Nonlinear Functions

#### Radial Basis Function (RBF)

$$ f(x, \xi) = \exp \left( -\frac{(x - \alpha)^T (x - \alpha)}{\lambda_0^2} \right) $$

where $ \xi = \{ \alpha, \lambda_0 \} $ are the RBF center and scale.

#### ArcTan Function

$$ f(x, \xi) = \arctan(\alpha^T x) $$

where $4 \xi = \{ \alpha \} $ is a parameter vector.

## Incremental Learning Procedure

We sequentially construct the activation function, optimizing new parameters while keeping previous ones fixed.

1. **First step:** Fit a single basis function:

   $$ a_i = \phi_0 + \phi_1 f(x_i, \xi_1) $$

   Optimize $ \phi_0, \phi_1, \xi_1 $ using maximum likelihood.

2. **Second step:** Introduce another basis function:

   $$ a_i = \phi_0 + \phi_1 f(x_i, \xi_1) + \phi_2 f(x_i, \xi_2) $$

   Optimize $ \phi_0, \phi_2, \xi_2 $ while keeping $\phi_1, \xi_1 $ fixed.

3. **Repeat:** At step $ K $:

   $$ a_i = \phi_0 + \sum_{k=1}^{K} \phi_k f(x_i, \xi_k) $$

   Optimize $ \phi_0, \phi_K, \xi_K $, keeping previous parameters unchanged.

## Optimization

The model parameters are learned by maximizing the log-likelihood:

$$ L = \sum_{i=1}^{I} \log \text{Bern}(w_i | \sigma(a_i)) $$

The gradient updates require:

$$ \frac{\partial L}{\partial \phi_k} = \sum_{i=1}^{I} (w_i - \sigma(a_i)) f(x_i, \xi_k) $$

$$ \frac{\partial L}{\partial \xi_k} = \sum_{i=1}^{I} (w_i - \sigma(a_i)) \frac{\partial f(x_i, \xi_k)}{\partial \xi_k} $$

## Properties

- **Sparse models**: The learned weights tend to decrease, allowing early truncation.
- **Scalability**: Works well even with high-dimensional data.
- **Efficient training**: Only a few parameters are optimized at each step.

## Visualization

The following figure shows an example where RBF functions are incrementally added to model the decision boundary.

![Incremental Logistic Regression](figure_placeholder.png)

- **(a)** Before fitting, the decision function is uniform.
- **(b-f)** More RBFs are added, progressively refining the classification.
- **(f)** After 10 RBFs, the model accurately classifies the data.

## Conclusion

Incremental fitting provides an efficient way to learn logistic regression models with nonlinear basis functions while maintaining sparsity. This method is particularly useful for large datasets and high-dimensional spaces.

```python
import math
import random
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# Radial Basis Function (RBF)
def rbf(x, center, scale):
    return math.exp(-((x - center) ** 2) / (2 * scale ** 2))

# Generate synthetic dataset
def generate_data(n=50):
    data = []
    for _ in range(n):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        label = 1 if x**2 + y**2 > 0.5 else 0  # Circular decision boundary
        data.append((x, y, label))
    return data

# Incremental fitting logistic regression with RBF
class IncrementalLogisticRegression:
    def __init__(self):
        self.weights = []  # Weights for basis functions
        self.centers = []  # Centers of RBFs
        self.scales = []  # Scales of RBFs
        self.bias = 0

    def add_basis_function(self, data, scale=0.5):
        """Add a new basis function based on the most informative data point."""
        best_point = random.choice(data)  # Normally, we'd use a heuristic
        self.centers.append((best_point[0], best_point[1]))
        self.scales.append(scale)
        self.weights.append(random.uniform(-1, 1))

    def predict(self, x, y):
        activation = self.bias
        for i in range(len(self.weights)):
            activation += self.weights[i] * rbf(math.sqrt((x - self.centers[i][0])**2 + (y - self.centers[i][1])**2), 0, self.scales[i])
        return sigmoid(activation)

    def train(self, data, iterations=10):
        """Incrementally add basis functions and update weights."""
        for _ in range(iterations):
            self.add_basis_function(data)
            # Simple weight update (gradient descent-like approach)
            learning_rate = 0.1
            for (x, y, label) in data:
                pred = self.predict(x, y)
                error = label - pred
                for i in range(len(self.weights)):
                    rbf_val = rbf(math.sqrt((x - self.centers[i][0])**2 + (y - self.centers[i][1])**2), 0, self.scales[i])
                    self.weights[i] += learning_rate * error * rbf_val
                self.bias += learning_rate * error

# Plot decision boundary
def plot_results(model, data):
    plt.figure(figsize=(6, 6))
    for x, y, label in data:
        plt.scatter(x, y, c='red' if label == 1 else 'blue')
    
    # Generate grid
    grid_size = 50
    x_vals = [i / grid_size * 2 - 1 for i in range(grid_size)]
    y_vals = [i / grid_size * 2 - 1 for i in range(grid_size)]
    
    for x in x_vals:
        for y in y_vals:
            pred = model.predict(x, y)
            plt.scatter(x, y, c='black', alpha=pred * 0.1)
    
    plt.title("Incremental Logistic Regression with RBFs")
    plt.show()

# Run the model
data = generate_data()
model = IncrementalLogisticRegression()
model.train(data, iterations=10)
plot_results(model, data)
```
