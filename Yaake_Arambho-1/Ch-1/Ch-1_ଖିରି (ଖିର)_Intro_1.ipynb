## Random Variables

A random variable $x$ denotes an uncertain quantity. It can represent the result of an experiment (e.g., flipping a coin) or a real-world measurement of a fluctuating property (e.g., temperature). The probability distribution $P(x)$ of a random variable captures the likelihood of different values it can take.

Random variables can be discrete or continuous:

- **Discrete Variables**: Take values from a predefined set, which can be ordered or unordered, finite or infinite. The probability distribution can be visualized as a histogram or Hinton diagram.
- **Continuous Variables**: Take values that are real numbers, which can be finite or infinite. The probability distribution is represented by a probability density function (pdf), where the probability density represents the relative propensity of the variable to take a particular value.

## Visualizing Probability Distributions

- **Discrete Variables**: Visualized as a histogram or Hinton diagram, where the height of the bar or area of the square represents the probability of each outcome.
- **Continuous Variables**: Plotted as a probability density function (pdf), where the height of the curve represents the probability density at each point.

##  Joint Probability

## Definition

Consider two random variables, $x$ and $y$. The joint probability distribution of $x$ and $y$, denoted as $P(x, y)$, represents the probability of observing specific values of both $x$ and $y$ together. The comma in $P(x, y)$ can be read as the English word "and," so $P(x, y)$ is the probability of $x$ and $y$.

Joint probability distributions can relate variables that are all discrete, all continuous, or a combination of discrete and continuous variables. Regardless of the type of variables, the total probability of all outcomes is always one.

## Multidimensional Variables

In general, we may be interested in the joint probability distribution of more than two variables. For example, $P(x, y, z)$ represents the joint probability distribution of scalar variables $x$, $y$, and $z$. We can also write $P(x)$ to represent the joint probability of all elements of a multidimensional variable $x = [x_1, x_2, ..., x_K]^T$. Similarly, $P(x, y)$ represents the joint distribution of all elements from multidimensional variables $x$ and $y$.

## Example: Joint Probability Distribution

Consider two continuous variables $x$ and $y$ with a joint probability distribution represented as a surface, contour plot, and image in Figure 2.3a-c. The joint distribution captures the relationship between the two variables, showing how the probability of observing specific values of $x$ and $y$ varies across the range of possible values.

## Marginalization

Marginalization is the process of recovering the probability distribution of a single variable from a joint distribution by summing (discrete case) or integrating (continuous case) over all other variables. This allows us to find the probability distribution of a variable regardless of the values of other variables.

For example, if $x$ and $y$ are both continuous and we know $P(x, y)$, we can recover the marginal distributions $P(x)$ and $P(y)$ using the following relations:

$$
\begin{align*}
P(x) &= \int P(x, y) \, dy \\
P(y) &= \int P(x, y) \, dx
\end{align*}
$$

The recovered distributions $P(x)$ and $P(y)$ are referred to as marginal distributions.

## Example: Marginal Distributions

Figure 2.4 illustrates the concept of marginalization for different types of variables. In each case, the marginal distribution is obtained by summing or integrating over the other variable(s) in the joint distribution.

## Conditional Probability

Conditional probability, denoted as $P(x|y = y^*)$, represents the relative propensity of the random variable $x$ to take different outcomes given that the random variable $y$ is fixed to the value $y^*$. The vertical line "|" can be read as the English word "given."

To calculate the conditional probability distribution, we examine the appropriate slice $P(x, y = y^*)$ of the joint distribution (Figure 2.5). The values in the slice provide information about the relative probability of $x$ taking various values when $y$ is fixed to $y^*$. However, these values do not form a valid probability distribution as they only represent a subset of the joint distribution.

To obtain the conditional probability distribution, we normalize the values in the slice by the total probability in the slice:

$$
P(x|y = y^*) = \frac{P(x, y = y^*)}{\int P(x, y = y^*) \, dx}
$$

## Example: Conditional Probability

Consider the joint distribution $P(x, y)$ and a specific value $y^*$. The conditional probability $P(x|y = y^*)$ represents the probability distribution of $x$ given that $y$ is fixed to $y^*$. By normalizing the values in the slice $P(x, y = y^*)$, we obtain the conditional probability distribution.

## Bayes' Rule

Bayes' rule is a fundamental theorem in probability theory that relates the conditional and marginal probabilities. It allows us to calculate the conditional probability of $y$ given $x$ using the conditional probability of $x$ given $y$ and the marginal probability of $x$:

$$
P(y|x) = \frac{P(x|y) \, P(y)}{P(x)}
$$

Bayes' rule is a powerful tool for updating our beliefs about the probability of an event based on new information or evidence.

## Example: Bayes' Rule

Suppose we have the following probabilities:

- $P(x) = 0.6$
- $P(y) = 0.4$
- $P(x|y) = 0.8$

Using Bayes' rule, we can calculate $P(y|x)$:

$$
P(y|x) = \frac{P(x|y) \, P(y)}{P(x)} = \frac{0.8 \times 0.4}{0.6} = \frac{2}{3}
$$

So, the conditional probability of $y$ given $x$ is $\frac{2}{3}$.

## Conclusion

Joint probability, marginalization, and conditional probability are fundamental concepts in probability theory. Understanding these concepts is essential for building probabilistic models and reasoning about uncertainty in computer vision applications. Bayes' rule provides a powerful tool for updating probabilities based on new information, which is crucial for many machine learning and inference tasks.

# 4. Conditional Probability

## Definition

Conditional probability, denoted as $P(x|y = y^*)$, represents the relative propensity of the random variable $x$ to take different outcomes given that the random variable $y$ is fixed to the value $y^*$. The vertical line "|" can be read as the English word "given."

## Calculating Conditional Probability

The conditional probability $P(x|y = y^*)$ can be calculated from the joint distribution $P(x, y)$ by examining the appropriate slice $P(x, y = y^*)$ (Figure 2.5). The values in the slice provide information about the relative probability of $x$ taking various values when $y$ is fixed to $y^*$. However, these values do not form a valid probability distribution as they only represent a subset of the joint distribution.

To obtain the conditional probability distribution, we normalize the values in the slice by the total probability in the slice:

$$
P(x|y = y^*) = \frac{P(x, y = y^*)}{\int P(x, y = y^*) \, dx}
$$

## Marginal Probability Relation

The marginal probability relation, given by Equation 2.1, can be used to simplify the denominator in the normalization process:

$$
P(x|y = y^*) = \frac{P(x, y = y^*)}{P(y = y^*)}
$$

## Compact Notation

It is common to use a more compact notation for conditional probability, writing $P(x, y)$ as $P(x|y)$:

$$
P(x, y) = P(x|y)P(y)
$$

## Symmetry

By symmetry, we also have:

$$
P(x, y) = P(y|x)P(x)
$$

## Conditional Probability with Multiple Variables

When we have more than two variables, we can repeatedly take conditional probabilities to divide up the joint probability distribution into a product of terms:

$$
P(w, x, y, z) = P(w, x, y|z)P(z) = P(w, x|y, z)P(y|z)P(z) = P(w|x, y, z)P(x|y, z)P(y|z)P(z)
$$

## Independence

Two random variables $x$ and $y$ are said to be independent if their joint probability distribution can be written as the product of their marginal distributions:

$$
P(x, y) = P(x)P(y)
$$

In this case, knowing the value of one variable does not provide any additional information about the other variable. Figure 2.6a illustrates the joint probability distribution of independent continuous variables, while Figure 2.6b shows the joint distribution of independent discrete variables.

## Bayes' Rule

Bayes' rule is a fundamental theorem in probability theory that relates the conditional and marginal probabilities. It allows us to calculate the conditional probability of $y$ given $x$ using the conditional probability of $x$ given $y$ and the marginal probability of $x$:

$$
P(y|x) = \frac{P(x|y)P(y)}{P(x)}
$$

Bayes' rule is often used to update our beliefs about the probability of an event based on new information or evidence.

## Terms in Bayes' Rule

The terms in Bayes' rule have specific names:

- $P(y|x)$ is the posterior, representing what we know about $y$ given $x$.
- $P(y)$ is the prior, representing what is known about $y$ before considering $x$.
- $P(x|y)$ is the likelihood, representing the probability of observing $x$ given $y$.
- $P(x)$ is the evidence, representing the marginal probability of $x$.

## Conclusion

Conditional probability is a fundamental concept in probability theory, allowing us to reason about the probability of an event given specific conditions or information. Bayes' rule provides a powerful tool for updating probabilities based on new evidence, which is essential for many machine learning and inference tasks in computer vision.

## Rules for Manipulating Expectations

The expected value of a function $f[x]$ of a random variable $x$ is defined as:

$$
\begin{align*}
E[f[x]] &= \sum_x f[x]P(x) \quad \text{(for discrete variables)} \\
E[f[x]] &= \int f[x]P(x) \, dx \quad \text{(for continuous variables)}
\end{align*}
$$

The expectation is a way to summarize the properties of a complex probability distribution. For some functions $f[x]$, the expectation is given a special name, as shown in Table 2.1.

There are four rules for manipulating expectations:

1. The expected value of a constant $\kappa$ with respect to the random variable $x$ is just the constant itself: $E[\kappa] = \kappa$.
2. The expected value of a constant $\kappa$ times a function $f[x]$ of the random variable $x$ is $\kappa$ times the expected value of the function: $E[\kappa f[x]] = \kappa E[f[x]]$.
3. The expected value of the sum of two functions $f[x]$ and $g[x]$ of a random variable $x$ is the sum of the individual expected values: $E[f[x] + g[x]] = E[f[x]] + E[g[x]]$.
4. The expected value of the product of two functions $f[x]$ and $g[y]$ of random variables $x$ and $y$ is equal to the product of the individual expected values if the variables $x$ and $y$ are independent: $E[f[x]g[y]] = E[f[x]]E[g[y]]$, if $x, y$ are independent.

Bayes' rule and the concept of independence are fundamental to understanding probability theory and its applications in computer vision.
