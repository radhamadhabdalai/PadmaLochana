## Introduction

Once we discussed latent variable models, we claimed that they naturally define a generative process by first sampling latents $ \mathbf{z} \sim p(\mathbf{z}) $ and then generating observables $ \mathbf{x} \sim p_{\theta} (\mathbf{x}|\mathbf{z}) $. That is nice! However, the problem appears when we start thinking about training. To be more precise, the training objective is an issue. Why? Well, the probability theory tells us to get rid of all unobserved random variables by marginalizing them out. In the case of latent variable models, this is equivalent to calculating the (marginal) log-likelihood function in the following form:

$$
\log p_{\theta} (\mathbf{x}) = \int \log p_{\theta} (\mathbf{x}|\mathbf{z}) p(\mathbf{z}) d\mathbf{z} \tag{8.1}
$$

As we mentioned already in the section about VAEs (see Sect. 5.3), the problematic part is calculating the integral because it is not analytically tractable unless all distributions are Gaussian and the dependency between $ \mathbf{x} $ and $ \mathbf{z} $ is linear. However, let us forget for a moment about all these issues and take a look at what we can do here.

First, we can approximate the integral using **Monte Carlo samples** from the prior $ p(\mathbf{z}) $ that yields:

$$
\log p_{\theta} (\mathbf{x}) = \int \log p_{\theta} (\mathbf{x}|\mathbf{z}) p(\mathbf{z}) d\mathbf{z} \approx \frac{1}{S} \sum_{s=1}^S \log p_{\theta} (\mathbf{x} | \mathbf{z}_s) \tag{8.2}
$$

$$
\approx \log \left( \sum_{s=1}^S \exp \left( \log p_{\theta} (\mathbf{x}|\mathbf{z}_s) \right) \right) - \log S \tag{8.3}
$$

$$
= \text{LogSumExp} \left( \left\{ p_{\theta} (\mathbf{x} | \mathbf{z}_s) \right\} \right) - \log S, \tag{8.4}
$$

where $ \text{LogSumExp} \left( f(s) \right) = \log \left( \sum_{s=1}^S \exp(f(s)) \right) $ is the log-sum-exp function.

Assuming for a second that this is a good (i.e., a tight) approximation, we turn the problem of calculating the integral into a problem of sampling from the prior. For simplicity, we can assume a prior that is relatively easy to sample from, e.g., the **standard Gaussian**, $ p(\mathbf{z}) = \mathcal{N} (\mathbf{z} | 0, I) $. In other words, we need to model $ p_{\theta} (\mathbf{x}|\mathbf{z}) $ only, i.e., pick a parameterization for it. Guess what, we will use a **neural network** again!

If we model images, then we can use the **categorical distribution** for the conditional likelihood, and then a neural network parameterizes the probabilities. Or if we use a **Gaussian distribution** like in the case of energy-based models or diffusion-based deep generative models, then $ p_{\theta} (\mathbf{x}|\mathbf{z}) $ could be Gaussian as well, and the neural network outputs the variance and/or the mean.

Since the log-sum-exp function is differentiable (and the application of the log-sum-exp trick makes it even numerically stable), there is no problem learning this model end to end! This approach is a precursor of many deep generative models and was dubbed **density networks** [1]; see Fig. 8.1 for a schematic representation of density networks.

### Why Density Networks Matter

Density networks are important and, unfortunately, underappreciated deep generative models. It is worth knowing them at least for three reasons:
1. Understanding how they work helps a lot in comprehending other latent variable models and how to improve them.
2. They serve as a great starting point for understanding the difference between prescribed models and implicit models.
3. They allow us to formulate a **nonlinear latent variable model** and train it using backpropagation (or gradient descent, in general).

Alright, so now you may have some questions because we made a few assumptions on the way that might have been pretty confusing. The main assumptions made here are the following:

- We need to specify the **prior distribution** $ p(\mathbf{z}) $, e.g., the standard Gaussian.
- We need to specify the form of the **conditional likelihood** $ p(\mathbf{x}|\mathbf{z}) $. Typically, people use the **Gaussian distribution** or a **mixture of Gaussians**. Hence, density nets are the **prescribed models** because we need to analytically formulate all distributions in advance. As a result, we get the following:
  - The **objective function** is the (approximated) log-likelihood function.
  - We can optimize the objective using **gradient-based optimization methods** and the autograd tools.
  - We can parameterize the conditional likelihood using **deep neural networks**.

However, we pay a great price for all the goodies coming from the formulation of the density networks:
- There is no analytical solution (except the case equivalent to the probabilistic PCA).
- We get an approximation of the log-likelihood function.
- We need a lot of samples from the prior to get a reliable approximation of the log-likelihood function.
- It suffers from the **curse of dimensionality**.

As you can see, the issue with dimensionality is especially limiting. What can we do with a model if it cannot be efficient for higher-dimensional problems? All interesting applications like image or audio analysis/synthesis are gone! So what can we do then?

One possible direction is to stick to the **prescribed models** and apply **variational inference** (see Sect. 5.3). However, the other direction is to **abandon the likelihood-based approach**. I know it sounds ridiculous, but it is possible...


```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

# Define a simple neural network for the likelihood model p(x|z)
class LikelihoodNetwork(nn.Module):
    def __init__(self, z_dim, x_dim):
        super(LikelihoodNetwork, self).__init__()
        self.fc1 = nn.Linear(z_dim, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, x_dim)  # Output the mean (and variance in case of Gaussian)
        
    def forward(self, z):
        x_hat = torch.relu(self.fc1(z))
        x_hat = torch.relu(self.fc2(x_hat))
        x_hat = self.fc3(x_hat)  # Mean (can also output variance if needed)
        return x_hat

# Function to calculate the log-likelihood approximation using Monte Carlo
def monte_carlo_log_likelihood(model, prior, x, num_samples=100):
    # Sample from the prior p(z) = N(0, I)
    z_samples = prior.sample((num_samples,))  # (num_samples, z_dim)
    
    # Approximate log p(x) using the Monte Carlo samples
    log_likelihood = 0
    for z in z_samples:
        x_hat = model(z)  # p(x|z) from the neural network
        log_likelihood += torch.log(torch.exp(-0.5 * torch.sum((x - x_hat) ** 2)))
    
    # Return the mean of the log likelihood estimates
    return log_likelihood / num_samples

# Define a simple Gaussian prior p(z) = N(0, I)
class GaussianPrior:
    def __init__(self, z_dim):
        self.z_dim = z_dim
    
    def sample(self, shape):
        return torch.randn(shape[0], self.z_dim)

# Set random seed for reproducibility
torch.manual_seed(42)

# Parameters
z_dim = 10  # Latent space dimension
x_dim = 5   # Observable space dimension (e.g., image pixels)
num_samples = 100  # Number of Monte Carlo samples for the approximation

# Initialize the model and prior
model = LikelihoodNetwork(z_dim, x_dim)
prior = GaussianPrior(z_dim)

# Example data (for simplicity, assume x is a vector of 5 dimensions)
x_data = torch.randn(x_dim)  # Simulated example data

# Define the optimizer
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
epochs = 100
for epoch in range(epochs):
    optimizer.zero_grad()
    
    # Calculate the log likelihood approximation
    log_likelihood = monte_carlo_log_likelihood(model, prior, x_data, num_samples)
    
    # Define the loss as the negative log likelihood (maximize likelihood by minimizing negative log likelihood)
    loss = -log_likelihood
    
    # Backpropagate and optimize
    loss.backward()
    optimizer.step()
    
    if epoch % 100 == 0:
        print(f"Epoch [{epoch}/{epochs}], Loss: {loss.item()}")

# After training, model parameters should have learned a good approximation to p(x|z)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import random

# Define a simple neural network for the likelihood model p(x|z)
class LikelihoodNetwork(keras.Model):
    def __init__(self, z_dim, x_dim):
        super(LikelihoodNetwork, self).__init__()
        self.dense1 = layers.Dense(128, activation='relu', input_dim=z_dim)
        self.dense2 = layers.Dense(128, activation='relu')
        self.dense3 = layers.Dense(x_dim)  # Output the mean (can also output variance if needed)
        
    def call(self, z):
        x_hat = self.dense1(z)
        x_hat = self.dense2(x_hat)
        x_hat = self.dense3(x_hat)
        return x_hat

# Function to calculate the log-likelihood approximation using Monte Carlo
def monte_carlo_log_likelihood(model, prior, x, num_samples=100):
    log_likelihood = 0
    for _ in range(num_samples):
        # Sample from the prior p(z) = N(0, I)
        z = prior()
        # Approximate log p(x) using the Monte Carlo samples
        x_hat = model(z)  # p(x|z) from the neural network
        # Calculate the log-likelihood for the Gaussian likelihood
        log_likelihood += -tf.reduce_sum((x - x_hat) ** 2)  # Simple squared error
    return log_likelihood / num_samples

# Define a simple Gaussian prior p(z) = N(0, I)
class GaussianPrior:
    def __init__(self, z_dim):
        self.z_dim = z_dim
    
    def __call__(self):
        return tf.random.normal((self.z_dim,))  # Sample a random vector from the standard normal distribution

# Parameters
z_dim = 10  # Latent space dimension
x_dim = 5   # Observable space dimension (e.g., image pixels)
num_samples = 100  # Number of Monte Carlo samples for the approximation

# Initialize the model and prior
model = LikelihoodNetwork(z_dim, x_dim)
prior = GaussianPrior(z_dim)

# Example data (for simplicity, assume x is a vector of 5 dimensions)
x_data = tf.random.normal((x_dim,))  # Simulated example data

# Define the optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

# Training loop
epochs = 1000
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        # Calculate the log likelihood approximation
        log_likelihood = monte_carlo_log_likelihood(model, prior, x_data, num_samples)
        
        # Define the loss as the negative log likelihood (maximize likelihood by minimizing negative log likelihood)
        loss = -log_likelihood
    
    # Get the gradients
    gradients = tape.gradient(loss, model.trainable_variables)
    
    # Apply the gradients
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    
    if epoch % 100 == 0:
        print(f"Epoch [{epoch}/{epochs}], Loss: {loss.numpy()}")

# After training, model parameters should have learned a good approximation to p(x|z)
```

```python
import random
import math

# Simple feedforward neural network for the likelihood model p(x|z)
class LikelihoodNetwork:
    def __init__(self, z_dim, x_dim):
        self.z_dim = z_dim
        self.x_dim = x_dim
        self.weights1 = self.init_weights(z_dim, 128)
        self.biases1 = self.init_weights(128, 1)
        self.weights2 = self.init_weights(128, 128)
        self.biases2 = self.init_weights(128, 1)
        self.weights3 = self.init_weights(128, x_dim)
        self.biases3 = self.init_weights(x_dim, 1)

    def init_weights(self, input_dim, output_dim):
        return [[random.random() * 0.1 for _ in range(output_dim)] for _ in range(input_dim)]

    def relu(self, x):
        return max(0, x)

    def forward(self, z):
        # Layer 1: Dense layer with ReLU activation
        x1 = [self.relu(sum(z[j] * self.weights1[j][i] for j in range(self.z_dim)) + self.biases1[i][0]) for i in range(128)]
        # Layer 2: Dense layer with ReLU activation
        x2 = [self.relu(sum(x1[j] * self.weights2[j][i] for j in range(128)) + self.biases2[i][0]) for i in range(128)]
        # Layer 3: Dense layer to output the mean of the likelihood (x_hat)
        x_hat = [sum(x2[j] * self.weights3[j][i] for j in range(128)) + self.biases3[i][0] for i in range(self.x_dim)]
        return x_hat

# Simple Gaussian prior p(z) = N(0, I)
class GaussianPrior:
    def __init__(self, z_dim):
        self.z_dim = z_dim
    
    def sample(self):
        return [random.gauss(0, 1) for _ in range(self.z_dim)]

# Function to calculate the log-likelihood approximation using Monte Carlo
def monte_carlo_log_likelihood(model, prior, x, num_samples=100):
    log_likelihood = 0
    for _ in range(num_samples):
        # Sample from the prior p(z) = N(0, I)
        z = prior.sample()
        # Approximate log p(x) using the Monte Carlo samples
        x_hat = model.forward(z)  # p(x|z) from the neural network
        # Calculate the log-likelihood for the Gaussian likelihood
        log_likelihood += -sum((x[i] - x_hat[i]) ** 2 for i in range(len(x)))  # Simple squared error
    return log_likelihood / num_samples

# Training loop for Monte Carlo Approximation
def train(model, prior, x_data, num_samples=100, epochs=1000, learning_rate=0.001):
    for epoch in range(epochs):
        # Compute the log likelihood approximation
        log_likelihood = monte_carlo_log_likelihood(model, prior, x_data, num_samples)
        
        # Calculate the loss (negative log-likelihood to minimize)
        loss = -log_likelihood

        # Print out the loss occasionally
        if epoch % 100 == 0:
            print(f"Epoch {epoch}/{epochs}, Loss: {loss}")

        # Gradient descent (naive version, no libraries)
        # Here we would normally compute gradients and update weights, but for simplicity,
        # we will perform a simple update step (this is a highly simplified version).
        # Proper backpropagation with manual gradient calculation would be much more involved.
        model.weights1 = [[w - learning_rate * random.gauss(0, 1) for w in row] for row in model.weights1]
        model.weights2 = [[w - learning_rate * random.gauss(0, 1) for w in row] for row in model.weights2]
        model.weights3 = [[w - learning_rate * random.gauss(0, 1) for w in row] for row in model.weights3]
        model.biases1 = [[b - learning_rate * random.gauss(0, 1)] for b in model.biases1]
        model.biases2 = [[b - learning_rate * random.gauss(0, 1)] for b in model.biases2]
        model.biases3 = [[b - learning_rate * random.gauss(0, 1)] for b in model.biases3]

# Example parameters
z_dim = 10  # Latent space dimension
x_dim = 5   # Observable space dimension
num_samples = 100  # Number of Monte Carlo samples for the approximation

# Initialize the model and prior
model = LikelihoodNetwork(z_dim, x_dim)
prior = GaussianPrior(z_dim)

# Example data (for simplicity, assume x is a vector of 5 dimensions)
x_data = [random.gauss(0, 1) for _ in range(x_dim)]  # Simulated example data

# Train the model using Monte Carlo log-likelihood approximation
train(model, prior, x_data, num_samples=num_samples, epochs=1000, learning_rate=0.001)
```

## Implicit Modeling with Generative Adversarial Networks (GANs)

### Getting Rid of Kullback-Leibler

Let us think again about what density networks tell us. First of all, they define a nice generative process: first, sample latents, and then generate observables. Clear! Then, for training, they use the (marginal) log-likelihood function. In other words, the log-likelihood function assesses the difference between a training datum and a generated object. 

To be even more precise, we first pick the specific probability distribution for the conditional likelihood $ p_{\theta}(x|z) $ that defines how to calculate the difference between the training point and the generated observables. One may ask here whether there is a different fashion of calculating the difference between real data and generated objects. If we recall our considerations about hierarchical VAEs, learning of the likelihood-based models is equivalent to optimizing the Kullback-Leibler (KL) divergence between the empirical distribution and the model:

$$
KL[p_{\text{data}}(x) \| p_{\theta}(x)].
$$

The KL-based approach requires a well-behaved distribution because of the logarithms. Moreover, we can think of it as a local way of comparing the empirical distribution (i.e., given data) and the generated data (i.e., data generated by our prescribed model). By "local," we mean considering one point at a time and then summing all individual errors instead of comparing samples (i.e., collections of individuals), which we can refer to as a "global" comparison. However, we do not need to stick to the KL divergence! Instead, we can use other metrics that look at a set of points (i.e., distributions represented by a set of points), such as integral probability metrics (e.g., the maximum mean discrepancy, MMD), or use other divergences.

Still, all of the mentioned metrics rely on defining explicitly how we measure the error. The question is whether we can parameterize our loss function and learn it alongside our model. Since we talk all the time about neural networks, can we go even further and utilize a neural network to calculate differences?

###  Getting Rid of Prescribed Distributions

Alright, we agreed on the fact that the KL divergence is only one of many possible loss functions. Moreover, we asked ourselves whether we could use a learnable loss function. However, there is also one question floating in the air: do we need to use the prescribed models in the first place?

The reason is the following: Since we know that density networks take noise and turn it into a distribution in the observable space, do we really need to output a full distribution? What if we return a single point? In other words, what if we define the conditional likelihood as Dirac’s delta:

$$
p_{\theta}(x|z) = \delta(x - G_{\theta}(z)),
$$

This is equivalent to saying that instead of a Gaussian (i.e., a mean and a variance), $ G_{\theta}(z) $ outputs only the mean. Interestingly, if we consider the marginal distribution over $ x $'s, we get a nicely behaved distribution.

To see that, let us first calculate the marginal distribution:

$$
p_{\theta}(x) = \int \delta(x - G_{\theta}(z)) p(z) dz.
$$

Now, let us understand what is going on! The marginal distribution is an infinite mixture of delta peaks. In other words, we take a single $ z $ and plot a peak (or a point in 2D, as it is easier to imagine) in the observable space. We proceed to infinity, and once we do that, the observable space will be covered by more and more points, and some regions will be denser than others. This kind of modeling of a distribution is also known as **implicit modeling**.

So where is the problem then? Well, the problem in the prescribed modeling setting is that the term $ \log \delta(x - G_{\theta}(z)) $ is ill-defined and cannot be used in many probability measures, including the KL term because we cannot calculate the loss function. Therefore, we can ask ourselves whether we can define our own loss function, perhaps? And even more, parameterize it with neural networks! You must admit it sounds appealing!

###  Adversarial Loss

Let us start with the following story: There is a con artist (a fraud) and a friend of the fraud (an expert) who knows little about art. Moreover, there is a real artist who has passed away (e.g., Pablo Picasso). The fraud tries to mimic the style of Pablo Picasso as well as possible. The friend expert browses for paintings of Picasso and compares them to the paintings provided by the fraud. Hence, the fraud tries to fool his friend, while the expert tries to distinguish real paintings of Picasso from fakes. Over time, the fraud becomes better and better, and the expert also learns how to decide whether a given painting is a fake.

Eventually and unfortunately to the world of art, the work of the fraud may become indistinguishable from Picasso, and the expert may be completely uncertain about the paintings and whether they are fakes. 

Now, let us formalize this wicked game. We call the expert a **discriminator** that takes an object $ x $ and returns a probability whether it is real (i.e., coming from the empirical distribution):

$$
D_{\alpha}: X \to [0, 1].
$$

We refer to the fraud as a **generator** that takes noise and turns it into an object \( x \):

$$
G_{\beta}: Z \to X.
$$

All $ x $'s coming from the empirical distribution $ p_{\text{data}}(x) $ are called **real**, and all $ x $'s generated by $ G_{\beta}(z) $ are dubbed **fake**. Then, we construct the objective function as follows:

- We have two sources of data: $ x \sim p_{\theta}(x) = G_{\beta}(z) p(z) dz $ and $ x \sim p_{\text{data}}(x) $.
- The discriminator solves the classification task by assigning 0 to all fake datapoints and 1 to all real datapoints.
- Since the discriminator can be seen as a classifier, we can use the **binary cross-entropy loss function** in the following form:

$$
\ell(\alpha, \beta) = \mathbb{E}_{x \sim \text{real}}[\log D_{\alpha}(x)] + \mathbb{E}_{z \sim p(z)}[\log(1 - D_{\alpha}(G_{\beta}(z)))].
$$

The left part corresponds to the real data source, and the right part contains the fake data source. We try to maximize $ \ell(\alpha, \beta) $ with respect to $ \alpha $ (i.e., the discriminator). In plain words, we want the discriminator to be as good as possible. The generator tries to fool the discriminator; thus, it tries to minimize $ \ell(\alpha, \beta) $ with respect to $ \beta $ (i.e., the generator).

Eventually, we face the following learning objective:

$$
\min_{\beta} \max_{\alpha} \mathbb{E}_{x \sim \text{real}}[\log D_{\alpha}(x)] + \mathbb{E}_{z \sim p(z)}[\log(1 - D_{\alpha}(G_{\beta}(z)))].
$$

We refer to $ \ell(\alpha, \beta) $ as the **adversarial loss** since there are two actors trying to achieve two opposite goals.


```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Helper function to generate random noise
def generate_noise(batch_size, z_dim):
    return np.random.normal(0, 1, size=(batch_size, z_dim))

# Helper function to generate fake data from noise using the generator
def generator(z, weights_g, bias_g):
    return np.tanh(np.dot(z, weights_g) + bias_g)

# Helper function to classify real/fake data using the discriminator
def discriminator(x, weights_d, bias_d):
    return 1 / (1 + np.exp(-(np.dot(x, weights_d) + bias_d)))  # Sigmoid

# Define the GAN training process
def train_gan(real_data, z_dim, epochs, batch_size, learning_rate=0.0002):
    # Dimensions of the generator and discriminator
    n_features = real_data.shape[1]
    
    # Initialize weights and biases for generator and discriminator
    weights_g = np.random.randn(z_dim, n_features) * 0.01
    bias_g = np.zeros(n_features)
    weights_d = np.random.randn(n_features, 1) * 0.01
    bias_d = np.zeros(1)
    
    # Loss tracking
    gen_losses = []
    disc_losses = []

    for epoch in range(epochs):
        # 1. Sample real data batch (from the real data distribution)
        real_batch = real_data[np.random.choice(real_data.shape[0], batch_size)]
        
        # 2. Generate fake data batch (from the noise)
        noise = generate_noise(batch_size, z_dim)
        fake_batch = generator(noise, weights_g, bias_g)
        
        # 3. Train discriminator (maximize log(D(x)) + log(1 - D(G(z))))
        real_predictions = discriminator(real_batch, weights_d, bias_d)
        fake_predictions = discriminator(fake_batch, weights_d, bias_d)
        
        disc_loss = -np.mean(np.log(real_predictions) + np.log(1 - fake_predictions))
        disc_gradients = np.dot(real_batch.T, (real_predictions - 1)) + np.dot(fake_batch.T, fake_predictions)
        weights_d -= learning_rate * disc_gradients  # Gradient descent
        bias_d -= learning_rate * np.sum(real_predictions - 1) + np.sum(fake_predictions)  # Bias update

        # 4. Train generator (minimize log(1 - D(G(z))) but equivalently we can minimize log(D(G(z))) for training)
        fake_predictions_gen = discriminator(fake_batch, weights_d, bias_d)
        gen_loss = -np.mean(np.log(fake_predictions_gen))
        gen_gradients = np.dot(noise.T, fake_predictions_gen - 1)  # Backpropagate
        weights_g -= learning_rate * gen_gradients  # Update generator weights
        bias_g -= learning_rate * np.sum(fake_predictions_gen - 1)  # Bias update

        # Record losses for plotting
        gen_losses.append(gen_loss)
        disc_losses.append(disc_loss)

        # Print progress every 1000 iterations
        if epoch % 1000 == 0:
            print(f"Epoch {epoch}/{epochs} | Generator Loss: {gen_loss:.4f} | Discriminator Loss: {disc_loss:.4f}")
    
    return weights_g, bias_g, weights_d, bias_d, gen_losses, disc_losses

# Generate synthetic real data (e.g., 2D points)
real_data = np.random.uniform(-1, 1, size=(1000, 2))

# Set GAN hyperparameters
z_dim = 5  # Dimension of the noise vector
epochs = 10000
batch_size = 64

# Train GAN
weights_g, bias_g, weights_d, bias_d, gen_losses, disc_losses = train_gan(real_data, z_dim, epochs, batch_size)

# Plot the loss curves
plt.figure(figsize=(10, 5))
plt.plot(gen_losses, label="Generator Loss")
plt.plot(disc_losses, label="Discriminator Loss")
plt.legend()
plt.title("Training Losses of Generator and Discriminator")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
