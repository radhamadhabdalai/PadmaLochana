
## Fitting Probability Models

## Maximum Likelihood

Maximum likelihood (ML) is a method for fitting probability models to data. It finds the set of parameters $\hat{\theta}$ that maximize the likelihood of the data. The likelihood function $P(x_i | \theta)$ is calculated by evaluating the probability density function at each data point $x_i$. Assuming independent data points, the likelihood function for a set of points is the product of individual likelihoods.

The ML estimate of the parameters is given by:

$$
\hat{\theta} = \arg\max_{\theta} [P(x_1, ..., x_I | \theta)]
$$

or equivalently:

$$
\hat{\theta} = \arg\max_{\theta} \prod_{i=1}^I P(x_i | \theta)
$$

To evaluate the predictive distribution for a new data point $x^*$, we simply evaluate the probability density function $P(x^* | \hat{\theta})$ using the ML-fitted parameters $\hat{\theta}$.

## Maximum a Posteriori

Maximum a posteriori (MAP) fitting introduces prior information about the parameters $\theta$. It maximizes the posterior probability $P(\theta | x_1, ..., x_I)$, taking into account both the likelihood of the data and the prior distribution of the parameters.

The MAP estimate of the parameters is given by:

$$
\hat{\theta} = \arg\max_{\theta} [P(\theta | x_1, ..., x_I)]
$$

or equivalently:

$$
\hat{\theta} = \arg\max_{\theta} \frac{P(x_1, ..., x_I | \theta) P(\theta)}{P(x_1, ..., x_I)}
$$

where $P(x_1, ..., x_I)$ is the marginal likelihood.

The predictive density for a new data point $x^*$ is calculated by evaluating the probability density function $P(x^* | \hat{\theta})$ using the MAP-fitted parameters $\hat{\theta}$.

## Bayesian Approach

In the Bayesian approach, instead of estimating single fixed values of the parameters, we compute a probability distribution $P(\theta | x_1, ..., x_I)$ over the parameters based on the data. This distribution represents our uncertainty about the parameter values.

The Bayesian predictive distribution for a new data point $x^*$ is given by:

$$
P(x^* | x_1, ..., x_I) = \int P(x^* | \theta) P(\theta | x_1, ..., x_I) d\theta
$$

This can be interpreted as a weighted sum of predictions made by different parameter values, where the weights are determined by the posterior probability distribution over the parameters.

## Unifying Predictive Distributions

The predictive density calculations for the Bayesian, MAP, and ML cases can be unified by considering the ML and MAP estimates as special probability distributions over the parameters. Specifically, we can view them as delta functions centered at $\hat{\theta}$.

A delta function $\delta[z]$ is a function that integrates to one and returns zero everywhere except at $z = 0$. Using this concept, we can write:

$$
P(x^* | x_1, ..., x_I) = P(x^* | \theta) \delta[\theta - \hat{\theta}] d\theta = P(x^* | \hat{\theta})
$$

This shows that the predictive density calculation for all three cases reduces to evaluating the probability of the new data point under the model with the estimated parameters.

## Conclusion

Fitting probability models involves finding the parameters that best describe the data. Maximum likelihood, maximum a posteriori, and the Bayesian approach are three methods for achieving this. Each method has its own advantages and considerations, and the choice depends on the specific problem and available information.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Maximum Likelihood
def max_likelihood(data):
    mean = np.mean(data)
    std = np.std(data)
    return mean, std

# Generate data for ML
data_ml = [random.gauss(0, 1) for _ in range(1000)]
mean_ml, std_ml = max_likelihood(data_ml)

# Predictive distribution for ML
x_values = np.linspace(-3, 3, 100)
pdf_ml = norm.pdf(x_values, mean_ml, std_ml)

# Maximum a Posteriori
def max_a_posteriori(data, prior_mean, prior_std):
    mean = np.mean(data)
    std = np.std(data)
    return prior_mean + (mean - prior_mean) / (1 + 1 / prior_std**2), np.sqrt(std**2 + prior_std**2)

# Generate data for MAP
data_map = [random.gauss(1, 1) for _ in range(1000)]
prior_mean_map = 0
prior_std_map = 1
mean_map, std_map = max_a_posteriori(data_map, prior_mean_map, prior_std_map)

# Predictive distribution for MAP
pdf_map = norm.pdf(x_values, mean_map, std_map)

# Bayesian Approach
def bayesian_predictive(data, prior_mean, prior_std):
    mean = np.mean(data)
    std = np.std(data)
    posterior_mean = prior_mean + (mean - prior_mean) / (1 + 1 / prior_std**2)
    posterior_std = np.sqrt(std**2 + prior_std**2)
    return posterior_mean, posterior_std

# Generate data for Bayesian
data_bayesian = [random.gauss(2, 1) for _ in range(1000)]
prior_mean_bayesian = 0
prior_std_bayesian = 1
mean_bayesian, std_bayesian = bayesian_predictive(data_bayesian, prior_mean_bayesian, prior_std_bayesian)

# Predictive distribution for Bayesian
pdf_bayesian = norm.pdf(x_values, mean_bayesian, std_bayesian)

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(data_ml, bins=20, density=True, alpha=0.7, label="Data")
plt.plot(x_values, pdf_ml, color='red', label="ML Predictive Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Maximum Likelihood")

plt.subplot(1, 3, 2)
plt.hist(data_map, bins=20, density=True, alpha=0.7, label="Data")
plt.plot(x_values, pdf_map, color='blue', label="MAP Predictive Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Maximum a Posteriori")

plt.subplot(1, 3, 3)
plt.hist(data_bayesian, bins=20, density=True, alpha=0.7, label="Data")
plt.plot(x_values, pdf_bayesian, color='green', label="Bayesian Predictive Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Bayesian Approach")

plt.tight_layout()
plt.show()
```


## Worked Example 1: Univariate Normal

## Maximum Likelihood Estimation

We will illustrate the concepts of fitting probability models using the example of fitting a univariate normal model to scalar data $\{x_i\}_{i=1}^I$. The univariate normal model has the probability density function (PDF):

$$
P(x | \mu, \sigma^2) = \text{Norm}_x[\mu, \sigma^2] = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$

where $\mu$ is the mean and $\sigma^2$ is the variance.

## Generating Data


Fig.1 Maximum likelihood ﬁtting. The likelihood of the parameters for a single datapoint is the height of the pdf evaluated at that point (blue vertical lines). The likelihood of a set of independently sampled data is the product of the individual likelihoods. a) The likelihood for this normal distribution is low because the large variance means the height of the pdf is low everywhere. b) The likelihood for this normal distribution is even lower as the left-most datum is very unlikely under the model. c) The maximum likelihood solution is the set of parameters for which the likelihood is maximized.


We generate $I$ independent data points $x_1, ..., x_I$ from a univariate normal distribution with $\mu = 1$ and $\sigma^2 = 1$. Our goal is to re-estimate these parameters from the data.

## Likelihood Computation

The likelihood $P(x_1, ..., x_I | \mu, \sigma^2)$ of the parameters $\{\mu, \sigma^2\}$ for the observed data is computed by evaluating the PDF for each data point separately and taking the product:

$$
P(x_1, ..., x_I | \mu, \sigma^2) = \prod_{i=1}^I P(x_i | \mu, \sigma^2)
$$

## Maximum Likelihood Solution

The maximum likelihood solution $\hat{\mu}, \hat{\sigma}^2$ occurs at the peak of the likelihood surface, which can be found by maximizing the likelihood function:

$$
\hat{\mu}, \hat{\sigma}^2 = \arg\max_{\mu, \sigma^2} P(x_1, ..., x_I | \mu, \sigma^2)
$$

In practice, it is often more convenient to work with the logarithm of the likelihood function, as it simplifies the optimization process. The log-likelihood $L$ is given by:

$$
L = \log P(x_1, ..., x_I | \mu, \sigma^2)
$$



Fig.2 The likelihood function for a ﬁxed set of observed data is a func- tion of the mean µ and variance σ 2 parameters. The plot shows that there are many parameter settings which might plausibly be responsible for the ten data points from ﬁg.1. A sensible choice for the “best” parameter setting is the maximum likelihood solution (green cross), which corre- sponds to the maximum of this func- tion.

## Least Squares Fitting

As an aside, we note that least squares fitting is equivalent to fitting the mean parameter of a normal distribution using the maximum likelihood method. The cost function for least squares fitting is:

$$
\hat{\mu} = \arg\min_{\mu} \sum_{i=1}^I (x_i - \mu)^2
$$

## Conclusion

In this worked example, we illustrated the process of fitting a univariate normal model to data using maximum likelihood estimation. We generated data from a known distribution, computed the likelihood of the parameters, and found the maximum likelihood solution. The example also highlighted the connection between maximum likelihood estimation and least squares fitting for the mean parameter of a normal distribution.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data from a univariate normal distribution
mu_true = 1
sigma_true = 1
num_samples = 1000
data = [random.gauss(mu_true, sigma_true) for _ in range(num_samples)]

# Maximum Likelihood Estimation
def max_likelihood(data):
    mean = np.mean(data)
    std = np.std(data)
    return mean, std

# Estimate parameters using maximum likelihood
mu_ml, sigma_ml = max_likelihood(data)

# Predictive distribution for ML
x_values = np.linspace(-3, 3, 100)
pdf_ml = norm.pdf(x_values, mu_ml, sigma_ml)

# Plotting
plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.hist(data, bins=20, density=True, alpha=0.7, label="Data")
plt.plot(x_values, pdf_ml, color='red', label="ML Predictive Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Maximum Likelihood Estimation")

plt.subplot(2, 1, 2)
plt.plot(x_values, norm.pdf(x_values, mu_true, sigma_true), color='blue', label="True Distribution")
plt.plot(x_values, pdf_ml, color='red', label="ML Predictive Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Comparison with True Distribution")

plt.tight_layout()
plt.show()
```

## Maximum a Posteriori Estimation

## Cost Function

For maximum a posteriori (MAP) estimation, the cost function becomes:

$$
\hat{\mu}, \hat{\sigma}^2 = \arg\max_{\mu, \sigma^2} P(x_i | \mu, \sigma^2) P(\mu, \sigma^2)
$$

where $P(\mu, \sigma^2)$ is the prior distribution over the parameters.

## Normal Inverse Gamma Prior

We choose a normal inverse gamma prior with parameters $\alpha, \beta, \gamma, \delta$ as it is conjugate to the normal distribution. The prior distribution is given by:

$$
P(\mu, \sigma^2) = \frac{\sqrt{\alpha + 1} \sqrt{\gamma} \beta^{\alpha - 1} \exp\left(-\frac{2\beta + \gamma(\delta - \mu)^2}{2\sigma^2}\right)}{\sigma \sqrt{2\pi} \Gamma(\alpha) \sigma^2}
$$

## MAP Solution

The MAP solution $\hat{\mu}, \hat{\sigma}^2$ is found by maximizing the logarithm of the cost function:

$$
\hat{\mu}, \hat{\sigma}^2 = \arg\max_{\mu, \sigma^2} \log[P(x_i | \mu, \sigma^2)] + \log[P(\mu, \sigma^2)]
$$

Substituting the expressions for the likelihood and prior, differentiating with respect to $\mu$ and $\sigma^2$, equating to zero, and rearranging, we get:

$$
\hat{\mu} = \frac{\sum_{i=1}^I x_i + \gamma\delta}{I + \gamma}
$$

and

$$
\hat{\sigma}^2 = \frac{1}{I + 3 + 2\alpha}
$$

## Interpretation of MAP Estimates

The MAP estimate of the mean $\hat{\mu}$ can be interpreted as a weighted sum of two terms: the data mean $x$ and the prior mean $\delta$. The weights are determined by the number of training examples $I$ and the hyperparameter $\gamma$, which controls the concentration of the prior.

With a large amount of data, the first term dominates, and the MAP estimate $\hat{\mu}$ is close to the data mean (and the ML estimate). With intermediate amounts of data, $\hat{\mu}$ is a weighted sum of the prediction from the data and the prediction from the prior. With no data, the estimate is governed by the prior.

Similarly, the MAP estimate of the variance $\hat{\sigma}^2$ is influenced by the prior and the number of data points.

## Conclusion

Maximum a posteriori estimation combines the likelihood of the data with the prior distribution over the parameters. The MAP solution is found at the peak of the posterior distribution, which is proportional to the product of the likelihood and the prior. The MAP estimates provide a balance between the data and the prior information, allowing for more robust parameter estimation, especially in cases with limited data.

The magenta cross indicates the peak of this prior distribution. The blue crosses are ﬁve samples randomly drawn from the distribution. b) The peak and the samples can be visualized by plotting the normal distri- butions that they represent.

The maximum a posteriori (MAP) solution (cyan cross) is found at the peak of the posterior distribu- tion.
```python
import random
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data from a univariate normal distribution
mu_true = 1
sigma_true = 1
num_samples = 1000
data = [random.gauss(mu_true, sigma_true) for _ in range(num_samples)]

# Maximum a Posteriori Estimation
def max_a_posteriori(data, alpha, beta, gamma, delta):
    n = len(data)
    sum_x = sum(data)
    sum_x_squared = sum(x**2 for x in data)
    mean = (sum_x + gamma * delta) / (n + gamma)
    variance = (sum_x_squared - (sum_x**2) / n) / (n + 3 + 2 * alpha)
    return mean, variance

# Prior parameters
alpha = 1
beta = 1
gamma = 1
delta = 0

# Estimate parameters using maximum a posteriori
mu_map, sigma_map = max_a_posteriori(data, alpha, beta, gamma, delta)

# Predictive distribution for MAP
x_values = [x for x in range(-30, 31, 1)]
pdf_map = norm.pdf(x_values, mu_map, sigma_map)

# Plotting
plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.hist(data, bins=20, density=True, alpha=0.7, label="Data")
plt.plot(x_values, pdf_map, color='red', label="MAP Predictive Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Maximum a Posteriori Estimation")

plt.subplot(2, 1, 2)
plt.plot(x_values, norm.pdf(x_values, mu_true, sigma_true), color='blue', label="True Distribution")
plt.plot(x_values, pdf_map, color='red', label="MAP Predictive Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Comparison with True Distribution")

plt.tight_layout()
plt.show()
```

```python
import random

# Generate data from a univariate normal distribution
mu_true = 1
sigma_true = 1
num_samples = 1000
data = [random.uniform(0, 2) for _ in range(num_samples)]

# Maximum a Posteriori Estimation
def max_a_posteriori(data, alpha, beta, gamma, delta):
    n = len(data)
    sum_x = sum(data)
    sum_x_squared = sum(x**2 for x in data)
    mean = (sum_x + gamma * delta) / (n + gamma)
    variance = (sum_x_squared - (sum_x**2) / n) / (n + 3 + 2 * alpha)
    return mean, variance

# Prior parameters
alpha = 1
beta = 1
gamma = 1
delta = 0

# Estimate parameters using maximum a posteriori
mu_map, sigma_map = max_a_posteriori(data, alpha, beta, gamma, delta)

# Predictive distribution for MAP
x_values = [x for x in range(-30, 31, 1)]
pdf_map = [1 / (sigma_map * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu_map) / sigma_map) ** 2) for x in x_values]

# Print the MAP estimates
print("MAP Mean:", mu_map)
print("MAP Variance:", sigma_map)

# Print the predictive distribution
for x, prob in zip(x_values, pdf_map):
    print(f"x = {x}, Probability = {prob}")
```

```python
import random
```
```python
import matplotlib.pyplot as plt

# Generate data from a univariate normal distribution
mu_true = 1
sigma_true = 1
num_samples = 1000
data = [random.uniform(0, 2) for _ in range(num_samples)]

# Maximum a Posteriori Estimation
def max_a_posteriori(data, alpha, beta, gamma, delta):
    n = len(data)
    sum_x = sum(data)
    sum_x_squared = sum(x**2 for x in data)
    mean = (sum_x + gamma * delta) / (n + gamma)
    variance = (sum_x_squared - (sum_x**2) / n) / (n + 3 + 2 * alpha)
    return mean, variance

# Prior parameters
alpha = 1
beta = 1
gamma = 1
delta = 0

# Estimate parameters using maximum a posteriori
mu_map, sigma_map = max_a_posteriori(data, alpha, beta, gamma, delta)

# Predictive distribution for MAP
x_values = [x for x in range(-30, 31, 1)]
pdf_map = [1 / (sigma_map * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu_map) / sigma_map) ** 2) for x in x_values]

# Plotting
plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.hist(data, bins=20, density=True, alpha=0.7, label="Data")
plt.plot(x_values, pdf_map, color='red', label="MAP Predictive Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Maximum a Posteriori Estimation")

plt.subplot(2, 1, 2)
plt.plot(x_values, [1 / (sigma_true * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu_true) / sigma_true) ** 2) for x in x_values], color='blue', label="True Distribution")
plt.plot(x_values, pdf_map, color='red', label="MAP Predictive Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Comparison with True Distribution")

plt.tight_layout()
plt.show()
```
