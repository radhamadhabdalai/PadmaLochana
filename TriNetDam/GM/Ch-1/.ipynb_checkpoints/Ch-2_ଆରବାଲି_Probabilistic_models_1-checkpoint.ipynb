##  Probabilistic Modeling: From Mixture Models to Probabilistic Circuits

### A Probabilistic Perspective on Modeling: Random Variables, Learning, Generalization, and Inference

#### Random Variables and Probability Distributions

Let us imagine cats. Most people like cats, and some people are crazy in love with cats. There are ginger cats, black cats, big cats, small cats, puffy cats, and furless cats. In fact, there are many different kinds of cats. However, when I say this word: *“a cat”*, everyone has some kind of a cat in their mind. One can close their eyes and generate a picture of a cat, either their own cat or a cat of a neighbor. 

Further, this generated cat is located somewhere, e.g., sleeping on a couch or in a garden chasing a fly, during the night or during the day, and so on. Probably, we can agree at this point that there are infinitely many possible scenarios of cats in some environments.

Why think of cats? Actually, we can think of any other object in the world. Formally, we can define a space of all objects in the world as $ \omega \in \Omega $. For simplicity, we consider only physical objects that an AI system (either embodied or not) could perceive through its sensors. Without a loss of generality, I would claim that even an AI-powered app deals with physical objects through, e.g., a camera, or images taken by a user through a camera.

The AI system processes the input signals through some functions that return, e.g., continuous values like analog signals, discrete values like text from a conversation, or bits like digital signals. These functions are called **random variables**. 

For instance, a robot with a camera can take an RGB photo (e.g., of a cat) of the height $ H $ and the width $ W $, which defines a random variable. Typically, a random variable is denoted by $ x(\omega) $ or $ x $ for short. In our example, for images taken by a digital camera, we get:

$$
x : \Omega \to \{0, \dots, 255\}^{3 \times H \times W}.
$$

In general, 

$$
x : \Omega \to E,
$$

where $ E $ is a $ D $-dimensional measurable space (in the sense of measure theory), e.g., 

$$
E = \mathbb{R}^D \quad \text{or} \quad E = \{0, 1\}^D.
$$

Random variables are called *random* because their outcomes are not necessarily predictable due to unobserved factors.


## A Side Note on Randomness

If we think about randomness, this simple introduction could very quickly turn into a deep philosophical discourse. If you, my curious reader, disagree with me, that is okay! But I hope that we can agree with the statements provided above and move on.

In other words, if one takes their cat and captures a series of photos, the resulting pixels could take various values due to factors such as lighting, the cat's breathing, a shaking hand, etc. These are factors that are neither controlled nor explicitly measured. This series of photos, taken at different times of the day, or even on different days, we call **observations**, denoted as:

$$
D = \{x_1, x_2, \dots, x_N\}.
$$

There is some randomness in the process of taking photos, but still, certain pixel values are highly improbable or even impossible. For instance, a cat has four legs, a single head, and its fur cannot be naturally pink. As a result, we can say that there exists a distribution of pixel values (values of random variables) that correspond to less likely or more likely outcomes.

Formally, we say that there exists some **probability distribution** of a random variable $ x $, denoted as $ p(x) $, such that:

- For the **discrete case**:
  $$
  \forall x \, p(x) \in [0, 1] \quad \text{and} \quad \sum_x p(x) = 1.
  $$

- For the **continuous case**:
  $$
  \int_x p(x) \, dx = 1.
  $$

### Properties of Probability Distributions

Probability distributions have several useful properties, two of which are especially important for modeling and inference. For simplicity, consider two random variables $ x = (x_1, x_2) $:

1. **The Sum Rule**:
   $$
   p(x_1) = \sum_{x_2} p(x_1, x_2).
   $$

2. **The Product Rule**:
   $$
   p(x_1, x_2) = p(x_1 \mid x_2) p(x_2) \quad \text{or} \quad p(x_1, x_2) = p(x_2 \mid x_1) p(x_1).
   $$

The **sum rule** allows us to "marginalize out" a random variable probabilistically. The **product rule** expresses a joint distribution in terms of conditional distributions. 

In the product rule, $ p(x_1 \mid x_2) $ is the **conditional distribution**, which is also a probability distribution. It sums to 1 and returns values in the range $[0, 1]$.

If two random variables are **independent**, e.g., $ x_1 $ is independent of $ x_2 $, then:

$$
p(x_1 \mid x_2) = p(x_1).
$$

In this case, the product rule simplifies to:

$$
p(x_1, x_2) = p(x_1)p(x_2).
$$

---

###  Modeling

In general, it is useful to think of some true probability distribution for a given random variable. However, this could also be misleading. Once again, this dry sentence could lead to a long discussion. Therefore, we will occasionally refer to some abstract **true distribution** when needed, but practically, we will work with some **empirical distribution** defined by observations $ D $.

Formally, the **empirical distribution** is defined as a sum of peaks:

### Empirical Distribution

The **empirical distribution** is defined as:

$$
p_{\text{data}}(x) = \frac{1}{N} \sum_{n=1}^N \delta(x - x_n),
$$

where $ \delta(x - x_n) $ is:

- $ 0 $ for $ x \neq x_n $,
- $ +\infty $ for $ x = x_n $ in the continuous case (Dirac’s delta), and
- $ 1 $ in the discrete case (Kronecker’s delta).

> **Note**: This is mathematics! It provides a clear and concise way to communicate and express ideas. It may seem complicated initially, but with time, these concepts become intuitive and useful.

---

### Identically and Independently Distributed (i.i.d.) Data

We often assume that observations come from the same **true distribution** and are achieved **independently**. This is referred to as **i.i.d. data**. These assumptions imply:

1. All $ x_n \in D $ follow the same distribution.
2. All $ x_n $'s are independent of each other, i.e.,

$$
p(x_1, x_2, \dots, x_N) = \prod_{n=1}^N p(x_n).
$$

---

### Probabilistic Modeling

Now, why introduce these concepts? When does this become interesting? These are the questions that intrigue us, aren't they?

The beauty lies in formulating a **model distribution** (or simply a model) of the world (or a specific aspect of it):

$$
p(x; \theta),
$$

where $ \theta $ denotes the **parameters** of the model.

---

### Choosing a Model

The form of the model depends on the nature of the observed data. Here are examples for different types of data:

#### 1. Real-Valued Data
For real-valued data, a suitable choice is the **Gaussian (Normal) Distribution**:

$$
p(x; \theta) = \frac{1}{\sqrt{(2\pi)^D |\Sigma|}} \exp\left(-\frac{1}{2}(x - \mu)^T \Sigma^{-1} (x - \mu)\right),
$$

or equivalently:

$$
p(x; \theta) = \mathcal{N}(x | \mu, \Sigma),
$$

where:

- $ \theta = (\mu, \Sigma) $,
- $ \mu $ denotes the **location parameters** (means),
- $ \Sigma $ denotes the **scale parameters** (covariance matrix with variances on the diagonal).

---

#### 2. Binary Data
For binary data $( x_d \in \{0, 1\} $), the **Bernoulli Distribution** is appropriate:

$$
p(x; \theta) = \prod_{d=1}^D \theta_d^{x_d} (1 - \theta_d)^{1 - x_d},
$$

or equivalently:

$$
p(x; \theta) = \text{Bern}(x | \theta).
$$

---

#### 3. Categorical Data
For categorical data (discrete and unordered, $ x_d \in \{0, 1, \dots, K-1\} $), the **Categorical Distribution** is used:

$$
p(x; \theta) = \prod_{d=1}^D \prod_{k=1}^K \theta_{dk}^{\mathbb{1}[x_d = k]},
$$

or equivalently:

$$
p(x; \theta) = \text{Cat}(x | \theta).
$$

    
    

```python
import math

def gaussian_pdf(x, mu, sigma):
    """
    Compute the probability density function of a Gaussian distribution.

    Args:
        x (float): The data point.
        mu (float): The mean of the distribution.
        sigma (float): The standard deviation of the distribution.

    Returns:
        float: The probability density.
    """
    coeff = 1 / (math.sqrt(2 * math.pi) * sigma)
    exponent = -0.5 * ((x - mu) / sigma) ** 2
    return coeff * math.exp(exponent)

def bernoulli_pmf(x, theta):
    """
    Compute the probability mass function of a Bernoulli distribution.

    Args:
        x (int): The data point (0 or 1).
        theta (float): The probability of success (1).

    Returns:
        float: The probability mass.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    return theta ** x * (1 - theta) ** (1 - x)

def categorical_pmf(x, theta):
    """
    Compute the probability mass function of a Categorical distribution.

    Args:
        x (int): The category (0 to K-1).
        theta (list of float): The probabilities for each category.

    Returns:
        float: The probability mass.
    """
    if x < 0 or x >= len(theta):
        raise ValueError("x must be in the range [0, len(theta) - 1].")
    if not math.isclose(sum(theta), 1.0):
        raise ValueError("Probabilities in theta must sum to 1.")
    return theta[x]

# Example Usage
if __name__ == "__main__":
    # Gaussian Example
    x = 5.0
    mu = 4.0
    sigma = 1.5
    print("Gaussian PDF:", gaussian_pdf(x, mu, sigma))

    # Bernoulli Example
    x = 1
    theta = 0.7
    print("Bernoulli PMF:", bernoulli_pmf(x, theta))

    # Categorical Example
    x = 2
    theta = [0.2, 0.3, 0.5]  # Probabilities for categories 0, 1, and 2
    print("Categorical PMF:", categorical_pmf(x, theta))
```

```python
import math
import matplotlib.pyplot as plt

# Gaussian PDF
def gaussian_pdf(x, mu, sigma):
    coefficient = 1 / (math.sqrt(2 * math.pi) * sigma)
    exponent = math.exp(-0.5 * ((x - mu) / sigma) ** 2)
    return coefficient * exponent

# Bernoulli PMF
def bernoulli_pmf(x, theta):
    return theta ** x * (1 - theta) ** (1 - x)

# Categorical PMF
def categorical_pmf(x, theta):
    if 0 <= x < len(theta):
        return theta[x]
    return 0

# Plot Gaussian PDF
def plot_gaussian(mu, sigma):
    x_values = [i * 0.1 for i in range(-50, 50)]
    y_values = [gaussian_pdf(x, mu, sigma) for x in x_values]

    plt.figure(figsize=(8, 5))
    plt.plot(x_values, y_values, label=f'Gaussian: $\mu={mu}$, $\sigma={sigma}$')
    plt.title('Gaussian Distribution')
    plt.xlabel('x')
    plt.ylabel('Density')
    plt.legend()
    plt.grid()
    plt.show()

# Plot Bernoulli PMF
def plot_bernoulli(theta):
    x_values = [0, 1]
    y_values = [bernoulli_pmf(x, theta) for x in x_values]

    plt.figure(figsize=(8, 5))
    plt.bar(x_values, y_values, width=0.3, color='skyblue', label=f'Bernoulli: $\Theta={theta}$')
    plt.title('Bernoulli Distribution')
    plt.xlabel('x')
    plt.ylabel('Probability')
    plt.xticks(x_values)
    plt.legend()
    plt.grid(axis='y')
    plt.show()

# Plot Categorical PMF
def plot_categorical(theta):
    x_values = list(range(len(theta)))
    y_values = [categorical_pmf(x, theta) for x in x_values]

    plt.figure(figsize=(8, 5))
    plt.bar(x_values, y_values, width=0.4, color='lightgreen', label=f'Categorical')
    plt.title('Categorical Distribution')
    plt.xlabel('Categories')
    plt.ylabel('Probability')
    plt.xticks(x_values)
    plt.legend()
    plt.grid(axis='y')
    plt.show()

# Main function for visualization
if __name__ == "__main__":
    # Gaussian
    mu, sigma = 0, 1
    plot_gaussian(mu, sigma)

    # Bernoulli
    theta = 0.7
    plot_bernoulli(theta)

    # Categorical
    theta_categorical = [0.2, 0.5, 0.3]
    plot_categorical(theta_categorical)
```

## Probabilistic Modeling: Learning
This notebook explores the concept of learning in probabilistic modeling, particularly through parameter estimation using the Kullback-Leibler (KL) divergence and maximum likelihood.

## KL Divergence and Log-Likelihood
The Kullback-Leibler divergence measures the distance between the empirical distribution $ p_{\text{data}}(x) $ and the model distribution $ p(x; \theta) $:
$$
\text{KL}[p_{\text{data}}(x) || p(x; \theta)] = \sum_x p_{\text{data}}(x) \log \frac{p_{\text{data}}(x)}{p(x; \theta)}.
$$

This can be rewritten as:
$$
\text{KL}[p_{\text{data}}(x) || p(x; \theta)] = C - \sum_x p_{\text{data}}(x) \log p(x; \theta),
$$
where $ C $ is the negative entropy of $ p_{\text{data}}(x) $, independent of $ \theta $.

For empirical data $ D = \{x_1, x_2, \ldots, x_N\} $, the log-likelihood is given by:
$$
\log p(D; \theta) = \sum_{n=1}^N \log p(x_n; \theta).
$$

This log-likelihood serves as the foundation for learning parameters $ \theta $.

---

## Implementation in Python
Below, we implement the KL divergence and log-likelihood calculation for a Gaussian model.



```python

import math

# Define the Gaussian PDF
def gaussian_pdf(x, mean, variance):
    return (1 / math.sqrt(2 * math.pi * variance)) * math.exp(-((x - mean) ** 2) / (2 * variance))

# KL Divergence
def kl_divergence(data, model_mean, model_variance):
    kl_sum = 0
    for x in data:
        p_data = 1 / len(data)  # Empirical probability for each data point
        p_model = gaussian_pdf(x, model_mean, model_variance)
        kl_sum += p_data * math.log(p_data / p_model)
    return kl_sum

# Log-Likelihood
def log_likelihood(data, model_mean, model_variance):
    log_likelihood_sum = 0
    for x in data:
        log_likelihood_sum += math.log(gaussian_pdf(x, model_mean, model_variance))
    return log_likelihood_sum

# Example Data
data = [1.0, 1.2, 0.8, 1.1, 1.3]  # Example dataset
model_mean = 1.0
model_variance = 0.1

# Compute KL Divergence and Log-Likelihood
kl = kl_divergence(data, model_mean, model_variance)
ll = log_likelihood(data, model_mean, model_variance)

print(f"KL Divergence: {kl}")
print(f"Log-Likelihood: {ll}")
```

# Probabilistic Modeling: Generalization, Inference, and Graphical Models

## 2.1.4 Generalization
George Box's famous quote reminds us:
> "All models are wrong, but some are useful."

### Example: Two-Moon Problem
- **Observed Data**: Sampled from a true distribution, as shown in Figure 2.1a.
- **Model Fit**: A probabilistic model's density estimation, shown in Figure 2.1b, generalizes to regions with little or no data.

### Key Concept: Generalization
Generalization refers to a model's ability to adapt to new, unseen data drawn from the same distribution:
- Models should assign probability mass to both observed and unobserved regions of the space.
- Probabilistic models play a crucial role in achieving generalization.

---

##  Inference
After fitting a model to data, we can use it for tasks such as:
1. **Generating New Data**: Sampling from the model distribution.
2. **Calculating Probabilities**: Computing the likelihood or evidence \( p(x) \).

### Probabilistic Queries
Inference involves answering probabilistic queries, such as:
- **EVI**: Evidence, \( p(x) \).
- **MAR**: Marginals, \( p(x_{-d}) = \int p(x) dx_d \).
- **MAP**: Most probable value, \( \arg\max_x p(x) \).
- Advanced queries about logical events.

### Note
Many contemporary deep generative models can perform EVI but struggle with MAR and MAP.

---

## 2.2 Probabilistic Graphical Models (PGMs)
PGMs represent probability distributions with a graph structure, illustrating stochastic dependencies.

### Types of PGMs
1. **Markov Networks**: Undirected graphs where edges imply dependencies.
2. **Bayesian Networks**: Directed acyclic graphs (DAGs) with:
   - Nodes as random variables.
   - Directed edges denoting conditional dependencies.

### Generative Modeling and Bayesian Networks
Bayesian networks are useful for generative processes due to their acyclic nature, ensuring proper ordering of dependencies.

---

## Example: Bayesian Network
Consider the Bayesian network in Figure 2.2:
- **Gray Node**: Observable variable.
- **White Node**: Latent variable.
- **Arrows**: Conditional dependencies.

### Joint Distribution
The network's joint distribution is:
$$
p(A, B, C) = p(C | A, B) \cdot p(A) \cdot p(B).
$$

### Marginal Distribution
To calculate the marginal over observable variables:
$$
p(A, C) = \sum_B p(A, B, C),
$$
which expands to:
$$
p(A, C) = \sum_B p(C | A, B) \cdot p(A) \cdot p(B).
$$

---

### Notes:
- Bayesian networks are limited in representing certain distributions.
- Some generative models go beyond Bayesian networks but serve as an excellent starting point.

---


```python
import numpy as np
from scipy.stats import norm, bernoulli, multinomial
import matplotlib.pyplot as plt
import networkx as nx

# Data: Two-moon problem (for generalization example)
from sklearn.datasets import make_moons

# 2.1.4 Generalization Example: Two-Moon Dataset
def plot_two_moon():
    X, _ = make_moons(n_samples=500, noise=0.1)
    plt.scatter(X[:, 0], X[:, 1], alpha=0.7, c='blue')
    plt.title("Observed Data: Two Moons")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.grid()
    plt.show()

# 2.1.5 Inference Example
def calculate_likelihood(data, mu, sigma):
    """
    Likelihood calculation for 1D Gaussian data.
    """
    likelihood = np.prod(norm.pdf(data, loc=mu, scale=sigma))
    return likelihood

def maximum_a_posteriori(data, mu_prior, sigma_prior, mu_likelihood, sigma_likelihood):
    """
    Example of MAP inference using Gaussian priors and likelihood.
    """
    posterior_mu = (mu_prior / sigma_prior**2 + np.sum(data) / sigma_likelihood**2) / \
                   (1 / sigma_prior**2 + len(data) / sigma_likelihood**2)
    posterior_sigma = np.sqrt(1 / (1 / sigma_prior**2 + len(data) / sigma_likelihood**2))
    return posterior_mu, posterior_sigma

# 2.2 Bayesian Network Example
def bayesian_network():
    """
    Construct and visualize a Bayesian network.
    """
    G = nx.DiGraph()
    G.add_edges_from([("A", "B"), ("A", "C"), ("B", "C")])

    pos = nx.spring_layout(G)  # Position nodes nicely
    nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=2000, font_size=15, font_weight="bold")
    plt.title("Bayesian Network")
    plt.show()

    # Joint probability example: P(A, B, C)
    A, B, C = 0.6, 0.3, 0.8  # Example probabilities
    P_A = A
    P_B = B
    P_C_given_A_B = lambda a, b: 0.5 * a + 0.5 * b  # Example dependency
    P_joint = P_A * P_B * P_C_given_A_B(A, B)
    print(f"Joint Probability P(A, B, C): {P_joint:.3f}")

# Generate Observed Data for Likelihood Example
data = np.random.normal(loc=0, scale=1, size=50)  # 50 samples from N(0, 1)

# Parameters for Likelihood and MAP
mu_prior = 0
sigma_prior = 1
mu_likelihood = np.mean(data)
sigma_likelihood = np.std(data)

# Calculate Likelihood
likelihood = calculate_likelihood(data, mu_likelihood, sigma_likelihood)

# Perform MAP Inference
posterior_mu, posterior_sigma = maximum_a_posteriori(
    data, mu_prior, sigma_prior, mu_likelihood, sigma_likelihood
)

# Results
print(f"Likelihood: {likelihood:.5e}")
print(f"Posterior Mean (MAP): {posterior_mu:.2f}, Posterior Std Dev: {posterior_sigma:.2f}")

# Visualization
plot_two_moon()
bayesian_network()
```

```python
# 2.1.4 Generalization: Two-Moon Example
def generate_two_moon_data(n_samples, noise):
    """
    Generate two-moon data manually.
    """
    data = []
    labels = []
    import math
    for i in range(n_samples):
        angle = i * math.pi / (n_samples // 2)
        radius = 1 + noise * (2 * random.random() - 1)
        if i < n_samples // 2:
            x = radius * math.cos(angle)
            y = radius * math.sin(angle)
            label = 0
        else:
            x = radius * math.cos(angle + math.pi) + 1
            y = radius * math.sin(angle + math.pi) - 0.5
            label = 1
        data.append((x, y))
        labels.append(label)
    return data, labels

def plot_two_moon(data):
    """
    Plot two-moon data without external libraries.
    """
    with open("two_moon_data.txt", "w") as file:
        for x, y in data:
            file.write(f"{x},{y}\n")

    print("Two-moon data saved to 'two_moon_data.txt'. You can visualize it using external tools.")

# 2.1.5 Inference: Gaussian Likelihood
def gaussian_pdf(x, mu, sigma):
    """
    Compute Gaussian PDF.
    """
    return (1 / (sigma * (2 * 3.14159) ** 0.5)) * math.exp(-0.5 * ((x - mu) / sigma) ** 2)

def calculate_likelihood(data, mu, sigma):
    """
    Calculate likelihood of data given Gaussian parameters.
    """
    likelihood = 1
    for x in data:
        likelihood *= gaussian_pdf(x, mu, sigma)
    return likelihood

def maximum_a_posteriori(data, mu_prior, sigma_prior, mu_likelihood, sigma_likelihood):
    """
    Perform MAP estimation manually for Gaussian prior and likelihood.
    """
    n = len(data)
    posterior_mu = (mu_prior / sigma_prior**2 + sum(data) / sigma_likelihood**2) / \
                   (1 / sigma_prior**2 + n / sigma_likelihood**2)
    posterior_sigma = (1 / (1 / sigma_prior**2 + n / sigma_likelihood**2))**0.5
    return posterior_mu, posterior_sigma

# 2.2 Bayesian Network
def bayesian_network_example():
    """
    Demonstrate Bayesian network with joint probability manually.
    """
    def p_c_given_a_b(a, b):
        return 0.5 * a + 0.5 * b  # Example conditional probability

    # Example probabilities
    p_a = 0.6
    p_b = 0.3
    p_c = p_c_given_a_b(p_a, p_b)
    joint_probability = p_a * p_b * p_c

    print(f"Joint Probability P(A, B, C): {joint_probability:.3f}")

# Main Execution
import math
import random

# Generate and Save Two-Moon Data
n_samples = 100
noise = 0.1
data, labels = generate_two_moon_data(n_samples, noise)
plot_two_moon(data)

# Perform Gaussian Likelihood and MAP
observed_data = [random.uniform(-1, 1) for _ in range(20)]  # Example observed data
mu_prior = 0
sigma_prior = 1
mu_likelihood = sum(observed_data) / len(observed_data)
sigma_likelihood = (sum((x - mu_likelihood) ** 2 for x in observed_data) / len(observed_data))**0.5

likelihood = calculate_likelihood(observed_data, mu_likelihood, sigma_likelihood)
posterior_mu, posterior_sigma = maximum_a_posteriori(observed_data, mu_prior, sigma_prior, mu_likelihood, sigma_likelihood)

print(f"Likelihood: {likelihood:.5e}")
print(f"Posterior Mean (MAP): {posterior_mu:.2f}, Posterior Std Dev: {posterior_sigma:.2f}")

# Bayesian Network Example
bayesian_network_example()
```

```python
# Define functions for Gaussian PDF and log-likelihood
def gaussian_pdf(x, mu, sigma):
    from math import exp, sqrt, pi
    return (1 / (sqrt(2 * pi) * sigma)) * exp(-0.5 * ((x - mu) / sigma) ** 2)

def log_likelihood(data, mu, sigma):
    from math import log
    likelihood = 0
    for x in data:
        likelihood += log(gaussian_pdf(x, mu, sigma))
    return likelihood

# Generate data for Gaussian curve
x_values = [-5 + i * 0.1 for i in range(101)]  # x in range [-5, 5] with step 0.1
mu = 0
sigma = 1
g_pdf = [gaussian_pdf(x, mu, sigma) for x in x_values]

# Generate data for likelihood curve
data = [1.5, 2.0, 2.5]  # Observed data points
mu_values = [-3 + i * 0.1 for i in range(61)]  # mu in range [-3, 3] with step 0.1
likelihood_values = [log_likelihood(data, mu, sigma) for mu in mu_values]

# Plot graphs without external libraries
import turtle

def plot_graph(x, y, title, x_label, y_label):
    screen = turtle.Screen()
    screen.setworldcoordinates(-10, -10, 10, 10)
    screen.title(title)

    pen = turtle.Turtle()
    pen.speed(0)
    pen.hideturtle()

    # Draw axes
    pen.penup()
    pen.goto(-10, 0)
    pen.pendown()
    pen.goto(10, 0)

    pen.penup()
    pen.goto(0, -10)
    pen.pendown()
    pen.goto(0, 10)

    # Plot points
    pen.penup()
    scale_x = 200 / (max(x) - min(x))
    scale_y = 200 / (max(y) - min(y))
    for i in range(len(x)):
        screen_x = scale_x * (x[i] - min(x)) - 100
        screen_y = scale_y * (y[i] - min(y)) - 100
        pen.goto(screen_x, screen_y)
        pen.dot(3, "blue")

    # Display labels
    screen.textinput("Graph", f"{title}\nX-axis: {x_label}\nY-axis: {y_label}")
    turtle.bye()

# Plot Gaussian curve
plot_graph(x_values, g_pdf, "Gaussian PDF", "x", "p(x)")

# Plot log-likelihood curve
plot_graph(mu_values, likelihood_values, "Log-Likelihood Curve", "mu", "Log-Likelihood")
```

## 2.3 Mixture Models
### 2.3.1 Modeling with Mixture of Gaussians (MoG)

Take a deep breath. A lot of information so far, I am aware of that, my curious reader. Take a short break if needed, and make a coffee. And buckle up, we are going to get all the information together and propose one of the simplest yet very effective generative models: a mixture model. Here we go!

---

### Bayesian Network Example

Consider a probabilistic graphical model (PGM) as shown in **Fig. 2.2** for three random variables:

- **A**, **B**: Observable variables (gray nodes)
- **C**: Latent variable (white node)

The corresponding joint distribution is given by:

$$
p(A, B, C) = p(C| A, B) p(A) p(B).
$$

To compute the marginal distribution over observable variables:

$$
p(A, C) = \sum_B p(A, B, C).
$$

---

### Mixture of Gaussians for Cat Weights

Let us consider the following situation:

- **Observable random variable**: $ x $, representing the weight of cats $( x \in \mathbb{R} $).
- **Latent variable**: $ z \in \{0, 1\} $, representing the sex of the cat:
  - $ z = 0 $: Female cats
  - $ z = 1 $: Male cats

The probabilistic graphical model (PGM) is shown in **Fig.3**, with $ z $ influencing $ x $.

#### Joint Distribution
The joint distribution can be written as:

$$
p(x, z) = p(x|z) p(z).
$$

Using a Gaussian distribution for $ x $ (real-valued) and a Bernoulli distribution for $ z $ (binary):

$$
p(x, z) = \mathcal{N}(x | \mu(z), \sigma^2(z)) \cdot \text{Bern}(z | \omega).
$$

Here:
- $ \mu(z) $ and $ \sigma^2(z) $ depend on $ z $.
- $ \omega $: Probability parameter for the Bernoulli distribution.

#### Marginal Distribution
Since $ z $ is unobserved, we marginalize it out:

$$
p(x) = \sum_z p(x, z)
$$

Substituting the joint distribution:

$$
p(x) = (1 - \omega) \mathcal{N}(x | \mu_0, \sigma_0^2) + \omega \mathcal{N}(x | \mu_1, \sigma_1^2),
$$

where:
- $ \mu_i \equiv \mu(z = i) $
- $ \sigma_i^2 \equiv \sigma^2(z = i) $

#### Example: Mixture of Two Gaussians
**Fig. 2.4a**: Two Gaussian components representing weights of:
- Female cats (blue, left)
- Male cats (orange, right)

**Fig.4b**: The marginal distribution $ p(x) $ combines the two components. Overlapping regions are summed.

#### General Definition
For $ K $ components, a Mixture of Gaussians (MoG) is defined as:

$$
p(x) = \sum_{k=1}^K \pi_k \mathcal{N}(x | \mu_k, \sigma_k^2),
$$

where:
- $ \pi_k $: Mixing coefficient $( \pi_k > 0 $, $ \sum_k \pi_k = 1 $).
- $ \mathcal{N}(x | \mu_k, \sigma_k^2) $: Gaussian component.

---

### Visualizing the Mixture of Gaussians
Let’s plot the individual components and the combined marginal distribution.

#### Implementation


```python

import matplotlib.pyplot as plt
import numpy as np

# Parameters for the two Gaussians
mu0, sigma0, pi0 = 4, 1, 0.6  # Female cats
mu1, sigma1, pi1 = 7, 1.5, 0.4  # Male cats

# Range for x
x = np.linspace(0, 12, 500)

# Gaussian distributions
p0 = pi0 * (1 / (np.sqrt(2 * np.pi) * sigma0)) * np.exp(-0.5 * ((x - mu0) / sigma0) ** 2)
p1 = pi1 * (1 / (np.sqrt(2 * np.pi) * sigma1)) * np.exp(-0.5 * ((x - mu1) / sigma1) ** 2)

# Mixture of Gaussians
p_mixture = p0 + p1

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, p0, label="Female Cats (Component 1)", color="blue", linestyle="--")
plt.plot(x, p1, label="Male Cats (Component 2)", color="orange", linestyle="--")
plt.plot(x, p_mixture, label="Mixture Distribution", color="green")
plt.title("Mixture of Gaussians for Cat Weights")
plt.xlabel("Weight")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()
```
