# Flow-Based Models

## 4.1 Flows for Continuous Random Variables

### 4.1.1 Introduction

So far, we have discussed a class of deep generative models that model the distribution \( p(x) \) directly in an autoregressive manner. The main advantage of ARMs is that they can learn long-range statistics and, as a consequence, are powerful density estimators. However, their drawback is that they are parameterized in an autoregressive manner, hence sampling is a slow process. Moreover, they lack a latent representation, making it less clear how to manipulate their internal data representation. This makes them less appealing for tasks like compression or metric learning.

In this chapter, we present a different approach to directly modeling \( p(x) \). However, before we begin, let's consider a simple example.

### Example

Let us take a random variable \( z \in \mathbb{R} \) with \( \pi(z) = \mathcal{N}(z|0, 1) \). Now, we consider a new random variable after applying some linear transformation to \( z \), namely, \( x = 0.75z + 1 \). The question is the following: What is the distribution of \( x \), \( p(x) \)?

We can guess the solution by using properties of Gaussians, or recall the change of variables formula to calculate this distribution:

$$
p(x) = \pi_z \left( f^{-1}(x) \right) \left| \frac{\partial f^{-1}(x)}{\partial x} \right|
$$

where \( f \) is an invertible function (a bijection). What does it mean? It means that the function maps one point to another distinctive point, and we can always invert the function to obtain the original point. 

In Figure 4.1, we have an example of a bijection. Notice that the volumes of the domains do not need to be the same! Keep it in mind and think about it in the context of \( \left| \frac{\partial f}{\partial x} \right| \).

Coming back to our example, we have:

$$
f(z) = 0.75z + 1 \quad \text{(4.2)}
$$

The inverse of \( f \) is:

$$
f^{-1}(x) = \frac{x - 1}{0.75} \quad \text{(4.3)}
$$

The derivative of the change of volume is:

$$
\left| \frac{\partial f^{-1}(x)}{\partial x} \right| = \frac{1}{0.75} = \frac{4}{3} \quad \text{(4.4)}
$$

Putting all the information together yields:

$$
p(x) = \pi_z \left( f^{-1}(x) \right) \left| \frac{\partial f^{-1}(x)}{\partial x} \right| = \frac{1}{0.75} \sqrt{2 \pi} \exp \left( -\frac{(x - 1)^2}{2 \cdot 0.75^2} \right)
$$

This simplifies to:

$$
p(x) = \mathcal{N}(x | 1, 0.75) \quad \text{(4.6)}
$$

Moreover, we see that the term \( \left| \frac{\partial f^{-1}(x)}{\partial x} \right| \) is responsible for normalizing the distribution \( \pi(z) \) after applying the transformation \( f \). In other words, \( \left| \frac{\partial f^{-1}(x)}{\partial x} \right| \) counteracts a possible change of volume caused by \( f \).

## Flow-Based Models

###  Flows for Continuous Random Variables (continued)

###  Example Continued

First of all, this example indicates that we can calculate a new distribution of a continuous random variable by applying a known bijective transformation \( f \) to a random variable with a known distribution \( z \sim p(z) \). The same holds for multiple variables \( x, z \in \mathbb{R}^D \):

$$
p(x) = p(z = f^{-1}(x)) \left| \frac{\partial f^{-1}(x)}{\partial x} \right|
$$
where:
$$
\left| \frac{\partial f^{-1}(x)}{\partial x} \right| = \det J_{f^{-1}}(x)
$$

The Jacobian matrix \( J_{f^{-1}} \) is defined as follows:

$$
J_{f^{-1}}(x) = 
\begin{bmatrix}
\frac{\partial f_1^{-1}}{\partial x_1} & \cdots & \frac{\partial f_1^{-1}}{\partial x_D} \\
\vdots & \ddots & \vdots \\
\frac{\partial f_D^{-1}}{\partial x_1} & \cdots & \frac{\partial f_D^{-1}}{\partial x_D}
\end{bmatrix}
$$

Moreover, we can also use the inverse function theorem that yields:

$$
\left| J_{f^{-1}}(x) \right| = \left| J_f(z) \right|^{-1}
$$

Since \( f \) is invertible, we can rewrite the equation for \( p(x) \) as follows:

$$
p(x) = p(z = f^{-1}(x)) \left| J_f(z) \right|
$$

### Role of the Jacobian Determinant

To get some insight into the role of the Jacobian determinant, consider the following three cases of invertible transformations, each applied to a uniform distribution defined over a square. The transformation cases are illustrated in Fig. 4.2:

1. **Volume-Preserving Transformation**: In the top case, the transformation turns the square into a rhombus without changing its volume. As a result, the Jacobian determinant of this transformation is 1. Such transformations are called volume-preserving. Notice that the resulting distribution is still uniform, and since there is no change in volume, it is defined over the same volume as the original one, maintaining the same color.

2. **Shrinking Transformation**: In the middle case, the transformation shrinks the volume. The resulting uniform distribution becomes "denser" (represented by a darker color). The Jacobian determinant is smaller than 1.

3. **Enlarging Transformation**: In the bottom case, the transformation enlarges the volume. Consequently, the uniform distribution is defined over a larger area (represented by a lighter color). The Jacobian determinant is larger than 1.

In the case of a **shifting operator**, the transformation is volume-preserving. Imagine adding an arbitrary value (e.g., 5) to all points of the square. Does it change the volume? Not at all! Hence, the Jacobian determinant equals 1.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Define the original random variable z ~ N(0, 1)
z = np.random.normal(0, 1, 10000)

# Step 2: Define the bijective transformation f(z) = 0.75z + 1
def f(z):
    return 0.75 * z + 1

# Inverse of f: f^(-1)(x) = (x - 1) / 0.75
def f_inv(x):
    return (x - 1) / 0.75

# Step 3: Jacobian of the inverse transformation |J_f_inv(x)|
# Since f_inv(x) = (x - 1) / 0.75, the derivative w.r.t x is 1 / 0.75
def jacobian_inv(x):
    return 1 / 0.75

# Step 4: Calculate p(x) = p(z = f^(-1)(x)) * |J_f_inv(x)|
# p(z) is the PDF of N(0, 1), and we want to transform it to p(x)
p_z = norm.pdf(z)  # Original PDF p(z) = N(0, 1)
x = f(z)            # Apply the transformation
p_x = norm.pdf(f_inv(x)) * jacobian_inv(x)  # Calculate the transformed PDF

# Step 5: Plot the original and transformed distributions

# Plotting p(z)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(z, bins=50, density=True, alpha=0.6, color='b', label=r'$p(z) \sim N(0, 1)$')
x_vals = np.linspace(-4, 4, 1000)
plt.plot(x_vals, norm.pdf(x_vals), 'r', label=r'$\mathcal{N}(0, 1)$')
plt.title("Original Distribution p(z)")
plt.legend()

# Plotting p(x)
plt.subplot(1, 2, 2)
plt.hist(x, bins=50, density=True, alpha=0.6, color='g', label=r'$p(x) \sim N(1, 0.75)$')
plt.plot(x_vals, norm.pdf(x_vals, loc=1, scale=0.75), 'r', label=r'$\mathcal{N}(1, 0.75)$')
plt.title("Transformed Distribution p(x)")
plt.legend()

plt.tight_layout()
plt.show()
````

````python
import random
import math
import matplotlib.pyplot as plt

# Step 1: Define the original random variable z ~ N(0, 1) using Box-Muller Transform
def box_muller():
    u1 = random.random()
    u2 = random.random()
    z0 = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
    return z0

# Generate 10000 samples of z ~ N(0, 1)
z = [box_muller() for _ in range(10000)]

# Step 2: Define the bijective transformation f(z) = 0.75z + 1
def f(z):
    return 0.75 * z + 1

# Inverse of f: f^(-1)(x) = (x - 1) / 0.75
def f_inv(x):
    return (x - 1) / 0.75

# Step 3: Jacobian of the inverse transformation |J_f_inv(x)|
# Since f_inv(x) = (x - 1) / 0.75, the derivative w.r.t x is 1 / 0.75
def jacobian_inv(x):
    return 1 / 0.75

# Step 4: Manually calculate the PDF of the normal distribution p(z)
def normal_pdf(x, mean=0, std=1):
    return (1 / (std * math.sqrt(2 * math.pi))) * math.exp(-0.5 * ((x - mean) / std) ** 2)

# Step 5: Calculate p(x) = p(z = f^(-1)(x)) * |J_f_inv(x)|
# p(z) is the PDF of N(0, 1), and we want to transform it to p(x)
p_z = [normal_pdf(z_i) for z_i in z]  # Original PDF p(z) = N(0, 1)
x = [f(z_i) for z_i in z]             # Apply the transformation
p_x = [normal_pdf(f_inv(x_i)) * jacobian_inv(x_i) for x_i in x]  # Calculate the transformed PDF

# Step 6: Plot the original and transformed distributions

# Plotting p(z)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(z, bins=50, density=True, alpha=0.6, color='b', label=r'$p(z) \sim N(0, 1)$')
x_vals = [i / 100 for i in range(-400, 400)]  # Range for plotting
plt.plot(x_vals, [normal_pdf(x_i) for x_i in x_vals], 'r', label=r'$\mathcal{N}(0, 1)$')
plt.title("Original Distribution p(z)")
plt.legend()

# Plotting p(x)
plt.subplot(1, 2, 2)
plt.hist(x, bins=50, density=True, alpha=0.6, color='g', label=r'$p(x) \sim N(1, 0.75)$')
plt.plot(x_vals, [normal_pdf(x_i, mean=1, std=0.75) for x_i in x_vals], 'r', label=r'$\mathcal{N}(1, 0.75)$')
plt.title("Transformed Distribution p(x)")
plt.legend()

plt.tight_layout()
plt.show()
````

## Change of Variables for Deep Generative Modeling

A natural question is whether we can utilize the idea of the **change of variables** to model a complex and high-dimensional distribution over images, audio, or other data sources. Let us consider a **hierarchical model**, or, equivalently, a sequence of invertible transformations $ f_k : \mathbb{R}^D \rightarrow \mathbb{R}^D $. 

We start with a known distribution $ \pi(z_0) = \mathcal{N}(z_0 | 0, I) $, where $ z_0 $ is a latent variable with a simple Gaussian distribution. Then, we can sequentially apply invertible transformations to obtain a flexible distribution $ p(x) $ as follows:

$$
p(x) = \pi(z_0) \left| \prod_{i=1}^{K} \frac{\partial f_i(z_{i-1})}{\partial z_{i-1}} \right|^{-1}, \quad \text{where} \quad x = f_K \circ f_{K-1} \circ \cdots \circ f_1(z_0)
$$

Alternatively, using the notation of a Jacobian for the $ i $-th transformation, we can express the density as:

$$
p(x) = \pi(z_0 = f^{-1}(x)) \prod_{i=1}^{K} \left| J f_i(z_{i-1}) \right|^{-1}
$$

where $ J f_i(z_{i-1}) $ denotes the Jacobian matrix of the transformation $ f_i $.

An example of transforming a **unimodal base distribution** like Gaussian into a **multimodal distribution** through invertible transformations is presented in the following figure. 

In principle, we should be able to get almost any arbitrarily complex distribution and revert to a simple one.

### Logarithmic Formulation

The logarithm of $ p(x) $ is given by:

$$
\ln p(x) = \ln \mathcal{N}(z_0 = f^{-1}(x) | 0, I) - \sum_{i=1}^K \ln \left| J f_i(z_{i-1}) \right|
$$

Interestingly, we see that the first part, $ \ln \mathcal{N}(z_0 = f^{-1}(x) | 0, I) $, corresponds to the **Mean Squared Error (MSE)** loss function between 0 and $ f^{-1}(x) $, plus a constant. The second part,

$$
\sum_{i=1}^K \ln \left| J f_i(z_{i-1}) \right|
$$

ensures that the distribution is properly normalized. However, since it penalizes the change of volume (as seen in the example above), we can think of it as a **regularizer** for the invertible transformations $ \{f_i\} $.

### Two Main Questions

Once we have laid down the foundations of the change of variables for expressing density functions, we must face two questions:

1. **How to model the invertible transformations?**
2. **What is the difficulty here?**

The answer to the first question could be **neural networks**, as they are flexible and easy to train. However, we cannot take any neural network because of two reasons:

- First, the transformation must be **invertible**; thus, we must pick an invertible neural network.
- Second, even if a neural network is invertible, we face the problem of calculating the second part of the equation above:

$$
\sum_{i=1}^K \ln \left| J f_i(z_{i-1}) \right|
$$

This term is nontrivial and computationally intractable for an arbitrary sequence of invertible transformations.

### Normalizing Flows

As a result, we seek neural networks that are both invertible and for which the logarithm of the Jacobian determinant is (relatively) easy to calculate. The resulting model, which consists of invertible transformations (neural networks) with tractable Jacobian determinants, is referred to as **normalizing flows** or **flow-based models**.

There are various possible invertible neural networks with tractable Jacobian determinants, such as:

- **Planar normalizing flows** [1]
- **Sylvester normalizing flows** [3]
- **Residual flows** [4, 5]
- **Invertible DenseNets** [6]

However, here we focus on a very important class of models: **RealNVP**, Real-valued Non-Volume Preserving flows [7], which serve as a starting point for many other flow-based generative models (e.g., GLOW [8]).

````python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Define a simple invertible neural network (Coupling Layer)
class RealNVP_CouplingLayer(nn.Module):
    def __init__(self, in_features, hidden_size):
        super(RealNVP_CouplingLayer, self).__init__()
        # Define the neural network that will be applied to half of the input
        self.network = nn.Sequential(
            nn.Linear(in_features // 2, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, in_features // 2)
        )

    def forward(self, x, reverse=False):
        # Split the input into two halves
        x1, x2 = x.chunk(2, dim=-1)

        if reverse:
            # Apply inverse transformation
            h = self.network(x2)
            x1 = x1 - h
            return torch.cat([x1, x2], dim=-1)
        else:
            # Apply forward transformation
            h = self.network(x2)
            x1 = x1 + h
            return torch.cat([x1, x2], dim=-1), h

    def log_jacobian_det(self, x):
        # Calculate the log of the absolute value of the determinant of the Jacobian
        x1, x2 = x.chunk(2, dim=-1)
        h = self.network(x2)
        log_det = torch.sum(h, dim=-1)  # Sum the element-wise Jacobian contributions
        return log_det

# Define the normalizing flow model
class RealNVP(nn.Module):
    def __init__(self, in_features, hidden_size, num_layers):
        super(RealNVP, self).__init__()
        self.layers = nn.ModuleList([
            RealNVP_CouplingLayer(in_features, hidden_size) for _ in range(num_layers)
        ])

    def forward(self, x, reverse=False):
        log_det_total = 0
        for layer in self.layers:
            x, log_det = layer(x, reverse=reverse)
            log_det_total += log_det
        return x, log_det_total

    def sample(self, num_samples):
        # Sampling from the standard normal distribution
        z = torch.randn(num_samples, 2, 2)  # Sample from standard normal (2x2)
        return self.forward(z, reverse=True)[0]

# Hyperparameters
in_features = 4  # Number of input features (2x2)
hidden_size = 128
num_layers = 4
num_epochs = 1000
learning_rate = 1e-3

# Create the model
model = RealNVP(in_features, hidden_size, num_layers)

# Define the optimizer
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Training Loop
for epoch in range(num_epochs):
    optimizer.zero_grad()

    # Sample data from standard normal distribution (latent space)
    z = torch.randn(64, in_features)  # Batch of 64 samples

    # Forward pass (get transformed data and log determinant)
    x, log_det = model(z)
    # Compute the negative log-likelihood (simplified loss)
    log_p_z = -0.5 * torch.sum(z ** 2, dim=-1)  # Gaussian log-probability
    loss = -torch.mean(log_p_z + log_det)

    # Backpropagation and optimization
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(f'Epoch [{epoch}/{num_epochs}], Loss: {loss.item():.4f}')

# Sampling from the model (generated data)
generated_samples = model.sample(1000).detach().numpy()

# Plot the generated samples
plt.scatter(generated_samples[:, 0], generated_samples[:, 1], alpha=0.5, color='r')
plt.title("Generated Samples from RealNVP")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define a simple invertible coupling layer
def coupling_layer(x, network_output, reverse=False):
    # Split the input into two parts
    x1, x2 = np.split(x, 2, axis=-1)
    
    if reverse:
        # Apply inverse transformation
        return np.concatenate([x1 - network_output(x2), x2], axis=-1)
    else:
        # Apply forward transformation
        return np.concatenate([x1 + network_output(x2), x2], axis=-1)

# Define a simple feedforward network (this will act as the function applied to one part of the input)
def simple_network(x, W1, b1, W2, b2):
    # A simple two-layer neural network
    hidden = np.maximum(0, np.dot(x, W1) + b1)  # ReLU activation
    output = np.dot(hidden, W2) + b2  # Output layer
    return output

# Define the log Jacobian determinant of the coupling layer
# Define the log Jacobian determinant of the coupling layer
# Define the log Jacobian determinant of the coupling layer
def log_jacobian(x, network_output, reverse=False):
    # For the additive coupling layer, the Jacobian determinant is simply the sum of the outputs
    # of the network function applied to one part of the input.
    return np.sum(network_output(x), axis=-1)

# Modify the forward method to correctly calculate the log-Jacobian without recursion
# Modify the RealNVP class to correctly handle the input dimensions
class RealNVP:
    def __init__(self, input_dim, hidden_dim):
        # Initialize the parameters for the neural network (input_dim is split in half)
        self.W1, self.b1, self.W2, self.b2 = init_params(input_dim // 2, hidden_dim, input_dim // 2)
        self.params = [self.W1, self.b1, self.W2, self.b2]

    def forward(self, x, reverse=False):
        # Define the network output as a function
        def network_output(x):  
            return simple_network(x, self.W1, self.b1, self.W2, self.b2)
        
        # Apply the coupling layer to the data
        transformed_x = coupling_layer(x, network_output, reverse)
        log_det = log_jacobian(x, network_output, reverse)  # Directly compute the log-det
        return transformed_x, log_det

    def sample(self, num_samples):
        # Sample from a standard normal distribution and apply the inverse transformation
        z = np.random.randn(num_samples, 2, 2)  # Ensure correct input shape
        return self.forward(z, reverse=True)[0]




# Initialize parameters for the neural network
def init_params(input_dim, hidden_dim, output_dim):
    W1 = np.random.randn(input_dim, hidden_dim)
    b1 = np.random.randn(hidden_dim)
    W2 = np.random.randn(hidden_dim, output_dim)
    b2 = np.random.randn(output_dim)
    return W1, b1, W2, b2

# Training loop
def train(model, x_train, num_epochs=1000, learning_rate=1e-3):
    for epoch in range(num_epochs):
        # Forward pass
        z = np.random.randn(*x_train.shape)  # Latent space (standard normal)
        
        # Compute transformed samples
        x, log_det = model.forward(z)  # Corrected line to call forward() method
        
        # Compute log-likelihood and loss (negative log-likelihood)
        log_p_z = -0.5 * np.sum(z ** 2, axis=-1)  # Log probability of Gaussian distribution
        loss = -np.mean(log_p_z + log_det)
        
        # Compute gradients
        grad_loss = -2 * (z / x_train.shape[0])  # Simple gradient calculation
        grad_loss = grad_loss + log_det # Include Jacobian contribution in gradient

        # Update model parameters
        for param in model.params:
            param -= learning_rate * grad_loss

        if epoch % 100 == 0:
            print(f'Epoch [{epoch}/{num_epochs}], Loss: {loss:.4f}')

# Model class


# Initialize the RealNVP model with the correct input dimension
model = RealNVP(input_dim=4, hidden_dim=128)

# Generate some random data for training
x_train = np.random.randn(64, 4)  # 64 samples, 4-dimensional data

# Train the model
train(model, x_train)


# Sampling from the trained model
generated_samples = model.sample(1000)

# Visualize the generated samples
plt.scatter(generated_samples[:, 0], generated_samples[:, 1], alpha=0.5, color='r')
plt.title("Generated Samples from RealNVP")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
````

````python
import random
import math

# Implement basic matrix operations
def matmul(A, B):
    """Matrix multiplication: A * B."""
    return [[sum(a * b for a, b in zip(A_row, B_col)) for B_col in zip(*B)] for A_row in A]

def add_bias(A, b):
    """Add bias to the matrix."""
    return [[a + bi for a, bi in zip(A_row, b)] for A_row in A]

def relu(x):
    """ReLU activation function."""
    return [[max(0, val) for val in row] for row in x]

def mat_transpose(A):
    """Transpose of matrix A."""
    return list(zip(*A))

# Initialize parameters for the neural network
def init_params(input_dim, hidden_dim, output_dim):
    W1 = [[random.gauss(0, 1) for _ in range(hidden_dim)] for _ in range(input_dim)]
    b1 = [random.gauss(0, 1) for _ in range(hidden_dim)]
    W2 = [[random.gauss(0, 1) for _ in range(output_dim)] for _ in range(hidden_dim)]
    b2 = [random.gauss(0, 1) for _ in range(output_dim)]
    return W1, b1, W2, b2

# Simple feedforward neural network
def simple_network(x, W1, b1, W2, b2):
    # Compute hidden layer
    hidden = matmul(x, W1)
    hidden = add_bias(hidden, b1)
    hidden = relu(hidden)

    # Output layer
    output = matmul(hidden, W2)
    output = add_bias(output, b2)
    return output

def coupling_layer(x, network_output, reverse=False):
    # Split input into two parts
    mid = len(x[0]) // 2
    x1, x2 = [row[:mid] for row in x], [row[mid:] for row in x]
    
    if reverse:
        # Apply inverse transformation
        return [x1_row + [x2_row[i] - network_output([x1_row])[0][i] for i in range(len(x2_row))] for x1_row, x2_row in zip(x1, x2)]
    else:
        # Apply forward transformation
        return [x1_row + [x2_row[i] + network_output([x1_row])[0][i] for i in range(len(x2_row))] for x1_row, x2_row in zip(x1, x2)]
# Simple coupling layer



# Log Jacobian determinant
def log_jacobian(x, network_output, reverse=False):
    # For simplicity, assume diagonal Jacobian, so just sum the network output
    return sum([sum(network_output([row])[0]) for row in x])

# RealNVP class
class RealNVP:
    def __init__(self, input_dim, hidden_dim):
        self.W1, self.b1, self.W2, self.b2 = init_params(input_dim // 2, hidden_dim, input_dim // 2)
        self.params = [self.W1, self.b1, self.W2, self.b2]

    def forward(self, x, reverse=False):
        def network_output(x):
            return simple_network(x, self.W1, self.b1, self.W2, self.b2)

        transformed_x = coupling_layer(x, network_output, reverse)
        log_det = log_jacobian(x, network_output, reverse)
        return transformed_x, log_det

    def sample(self, num_samples):
        # Sample from standard normal distribution (using random.gauss for simplicity)
        z = [[random.gauss(0, 1) for _ in range(2)] for _ in range(num_samples)]
        return self.forward(z, reverse=True)[0]

# Training function (simplified for illustration)
def train(model, x_train, num_epochs=1000, learning_rate=1e-3):
    for epoch in range(num_epochs):
        # Forward pass
        z = [[random.gauss(0, 1) for _ in range(len(x_train[0]))] for _ in range(len(x_train))]  # Latent space (standard normal)

        # Compute transformed samples
        x, log_det = model.forward(z)

        # Compute log-likelihood and loss (negative log-likelihood)
        log_p_z = [-0.5 * sum(val ** 2 for val in row) for row in z]  # Log probability of Gaussian distribution
        loss = -sum(log_p_z) / len(x_train) - log_det / len(x_train)

        # Print loss for illustration
        if epoch % 100 == 0:
            print(f'Epoch [{epoch}/{num_epochs}], Loss: {loss:.4f}')


# Initialize and train the model
model = RealNVP(input_dim=4, hidden_dim=128)

# Generate some random data for training (64 samples, 4-dimensional data)
x_train = [[random.gauss(0, 1) for _ in range(4)] for _ in range(64)]

# Train the model
train(model, x_train)

# Sample from the trained model
generated_samples = model.sample(1000)

# Visualize the generated samples (you can use matplotlib or another plotting library here)
import matplotlib.pyplot as plt

plt.scatter([sample[0] for sample in generated_samples], [sample[1] for sample in generated_samples], alpha=0.5, color='r')
plt.title("Generated Samples from RealNVP")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
````

###  Building Blocks of RealNVP

####  Coupling Layers

The main component of RealNVP is a coupling layer. The idea behind this transformation is as follows. Let us consider an input to the layer that is divided into two parts:  
$$ \mathbf{x} = [\mathbf{x}_a, \mathbf{x}_b] $$  
The division into two parts could be done by splitting the vector $\mathbf{x}$ into $\mathbf{x}_{1:d}$ and $\mathbf{x}_{d+1:D}$ or according to a more sophisticated manner, e.g., a checkerboard pattern. Then, the transformation is defined as follows:

$$ \mathbf{y}_a = \mathbf{x}_a $$  
$$ \mathbf{y}_b = \exp(s(\mathbf{x}_a)) \odot \mathbf{x}_b + t(\mathbf{x}_a) $$

where $s(\cdot)$ and $t(\cdot)$ are arbitrary neural networks called scaling and translation, respectively. This transformation is invertible by design, namely:

$$ \mathbf{x}_b = (\mathbf{y}_b - t(\mathbf{y}_a)) \odot \exp(-s(\mathbf{y}_a)) $$  
$$ \mathbf{x}_a = \mathbf{y}_a $$

Importantly, the logarithm of the Jacobian determinant is easy to calculate, because:

$$ \mathbf{J} = \begin{bmatrix} 
I_{d \times d} & 0_{d \times (D-d)} \\
\frac{\partial \mathbf{x}_a}{\partial \mathbf{y}_a} & \text{diag}(\exp(s(\mathbf{x}_a))) 
\end{bmatrix} $$

which leads to:

$$ \det(\mathbf{J}) = \prod_{j=1}^{D-d} \exp(s(\mathbf{x}_a))_j $$

Hence, the coupling layers are flexible and powerful transformations with tractable Jacobian determinants.

####  Permutation Layers

A simple yet effective transformation that could be combined with a coupling layer is a permutation layer. Since permutation is volume-preserving, i.e., its Jacobian determinant is equal to 1, we can apply it each time after the coupling layer. For instance, we can reverse the order of variables. An example of an invertible block, i.e., a combination of a coupling layer with a permutation layer, is schematically presented in Figure 4.4.

####  Dequantization

Flow-based models assume that $\mathbf{x}$ is a vector of real-valued random variables. However, in practice, many objects are discrete. For instance, images are typically represented as integers taking values in $\{0, 1, \dots, 255\}^D$. 

To handle discrete data, we use **uniform dequantization**. In this process, we add uniform noise, $\mathbf{u} \in [-0.5, 0.5]^D$, to the original data, $\mathbf{y} \in \{0, 1, \dots, 255\}^D$, to obtain a continuous space:

$$ \mathbf{x} = \mathbf{y} + \mathbf{u} $$

This procedure allows us to apply density estimation to $\mathbf{x}$.

####  Flows in Action!



Fig.5 A schematic representation of the uniform dequantization for two binary random variables: (left) the probability mass is assigned to points, (right) after the uniform dequantization, the probability mass is assigned to square areas. Colors correspond to probability values.

Let us now discuss the log-likelihood function (i.e., the learning objective) and how the mathematical formulas correspond to the code.

The learning objective is the log-likelihood, defined as:

$$ \ln p(\mathbf{x}) = \sum_{k=1}^{K} \ln \mathcal{N}(\mathbf{z}_0 = f^{-1}(\mathbf{x}) | 0, I) $$

where $\mathbf{s}_k$ is the scale network in the $k$-th coupling layer, and $\mathbf{x}_k$ denotes the input to the $k$-th coupling layer.

We can express this objective in terms of the log-likelihood of the standard Gaussian distribution:

$$ \ln \mathcal{N}(\mathbf{z}_0 = f^{-1}(\mathbf{x}) | 0, I) = -\text{const} - \frac{1}{2} \|f^{-1}(\mathbf{x})\|^2 $$

where $\text{const} = D \cdot \ln(2\pi)$ is the normalizing constant of the standard Gaussian distribution.

The second part of the objective involves calculating the log-Jacobian determinants, where we sum over transformations. For each coupling layer, we consider only the outputs of the scale nets.

Finally, during the implementation of the coupling layers, we must return not only the output but also the outcome of the scale layer.


## RealNVP Implementation

The following is an implementation of the RealNVP (Real-valued Non-Volume Preserving) model using PyTorch. It consists of coupling layers, permutation layers, and the necessary forward and inverse passes to transform data and calculate the log-likelihood.

````python
import math

class RealNVP:
    def __init__(self, nets, nett, num_flows, prior, D=2, dequantization=True):
        # Initialize parameters and components for RealNVP
        self.dequantization = dequantization  # Whether to dequantize data
        self.prior = prior  # The prior distribution
        self.num_flows = num_flows  # Number of flows (transformations)
        self.D = D  # Dimensionality of input
        self.t = [nett() for _ in range(num_flows)]  # List of translation networks
        self.s = [nets() for _ in range(num_flows)]  # List of scale networks
    
    def coupling(self, x, index, forward=True):
        # Split input into two parts
        mid = len(x) // 2
        xa, xb = x[:mid], x[mid:]
        
        # Compute scale and translation using simple dummy functions (as lists)
        s = self.s[index](xa)
        t = self.t[index](xa)
        
        # Convert s and t to lists of floats
        s = [float(i) for i in s]
        t = [float(i) for i in t]
        
        # Forward pass
        if forward:
            yb = [(xb_i - t_i) * math.exp(-s_i) for xb_i, t_i, s_i in zip(xb, t, s)]  # Exp(-s(xa))
        else:
            yb = [math.exp(s_i) * xb_i + t_i for xb_i, t_i, s_i in zip(xb, t, s)]  # Exp(s(xa)) * xb + t(xa)
        
        return xa + yb, s
    
    def permute(self, x):
        # Permutation of input (reversing the order)
        return x[::-1]
    
    def f(self, x):
        # Compute forward pass through all layers
        log_det_J, z = 0, x
        for i in range(self.num_flows):
            z, s = self.coupling(z, i, forward=True)
            z = self.permute(z)
            log_det_J -= sum(s)  # Accumulate the Jacobian determinant
        return z, log_det_J
    
    def f_inv(self, z):
        # Inverse pass (from z to x)
        x = z
        for i in reversed(range(self.num_flows)):
            x = self.permute(x)
            x, _ = self.coupling(x, i, forward=False)
        return x
    
    def forward(self, x, reduction='avg'):
        # Forward function (used during training)
        z, log_det_J = self.f(x)
        log_prob = self.prior.log_prob(z)
        if reduction == 'sum':
            return -(log_prob + log_det_J).sum()
        else:
            return -(log_prob + log_det_J).mean()
    
    def sample(self, batch_size):
        # Generate samples from the prior
        z = self.prior.sample((batch_size, self.D))
        return self.f_inv(z).reshape(batch_size, self.D)


# Define networks for scale and translation (using simple dummy functions here)
def nets():
    # Simple scale network (dummy function for demonstration)
    return lambda x: [0.1 for _ in range(len(x) // 2)]

def nett():
    # Simple translation network (dummy function for demonstration)
    return lambda x: [0.5 for _ in range(len(x) // 2)]


# Prior distribution (e.g., standard normal)
class Prior:
    def log_prob(self, z):
        # Dummy log-probability function for a standard normal distribution
        return -0.5 * sum([i**2 for i in z])


    def sample(self, shape):
        # Dummy sampling function for standard normal distribution
        return [[0.0 for _ in range(shape[1])] for _ in range(shape[0])]


# Example setup
num_flows = 8  # Number of flows (transformations)
D = 2  # Dimensionality of the data
prior = Prior()  # The prior (standard normal distribution)
model = RealNVP(nets, nett, num_flows, prior, D=D)

# Dummy training loop (loss calculation only)
losses = []

# Simulate training
for epoch in range(100):
    # Dummy input (just random values for demonstration)
    x = [0.0 for _ in range(D)]  # Dummy data for input
    
    # Compute loss (forward pass)
    loss = model.forward(x, reduction='mean')
    losses.append(loss)
    
    # Plot every 10 epochs
    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}")

# Simulate generation of samples
generated_samples = model.sample(5)  # Generate 5 samples

# Text-based loss plot
def plot_training_loss_text(losses):
    max_loss = max(losses)
    min_loss = min(losses)
    
    print("\nTraining Loss Visualization:")
    for i, loss in enumerate(losses):
        bar_length = int((loss - min_loss) / (max_loss - min_loss) * 50)
        bar = "#" * bar_length
        print(f"Epoch {i+1}: {bar} ({loss:.4f})")

# Plot the training loss (simple text-based graph)
plot_training_loss_text(losses)

# Print generated samples
def plot_generated_samples_text(samples):
    print("\nGenerated Samples:")
    for sample in samples:
        print(f"x1: {sample[0]:.2f}, x2: {sample[1]:.2f}")

plot_generated_samples_text(generated_samples)
````
