# The Normal Distribution

## Introduction

The multivariate normal distribution is a common representation of uncertainty in machine vision. It has two parameters: the mean $\mu$ and the covariance $\Sigma$. The mean $\mu$ is a $D \times 1$ vector that describes the position of the distribution, while the covariance $\Sigma$ is a symmetric $D \times D$ positive definite matrix that describes the shape of the distribution.

The probability density function of the multivariate normal distribution is given by:

$$
P(x) = \frac{1}{(2\pi)^{D/2} |\Sigma|^{1/2}} \exp \left(-\frac{1}{2} (x - \mu)^T \Sigma^{-1} (x - \mu) \right)
$$

or, for short:

$$
P(x) = \text{Norm}(x | \mu, \Sigma)
$$

## Types of Covariance Matrix

Covariance matrices in multivariate normals take three forms: spherical, diagonal, and full covariances. For the two-dimensional (bivariate) case, these are:

$$
\Sigma_{\text{spher}} = \begin{bmatrix}
\sigma^2 & 0 \\
0 & \sigma^2
\end{bmatrix}, \quad
\Sigma_{\text{diag}} = \begin{bmatrix}
\sigma_1^2 & 0 \\
0 & \sigma_2^2
\end{bmatrix}, \quad
\Sigma_{\text{full}} = \begin{bmatrix}
\sigma_1^2 & \sigma_{12} \\
\sigma_{12} & \sigma_2^2
\end{bmatrix}
$$

- **Spherical Covariance**: The covariance matrix is a positive multiple of the identity matrix, with the same value on all diagonal elements and zeros elsewhere. The variables are independent, and the iso-probability surfaces are hyperspheres.
- **Diagonal Covariance**: Each value on the diagonal has a different positive value, and the rest of the elements are zeros. The variables are independent but scaled differently, and the iso-probability surfaces are hyper-ellipsoids (ellipses in 2D) aligned with the coordinate axes.
- **Full Covariance**: The covariance matrix can have non-zero elements everywhere, but it must be symmetric and positive definite. The variables are dependent, and the iso-probability surfaces are ellipsoids that can take an arbitrary orientation.

## Properties of Covariance Matrices

- **Spherical Covariance**: In $D$ dimensions, spherical covariances produce iso-contours that are $D$-spheres.
- **Diagonal Covariance**: In $D$ dimensions, diagonal covariances produce iso-contours that are $D$-dimensional ellipsoids aligned with the coordinate axes.
- **Full Covariance**: In $D$ dimensions, full covariances produce iso-contours that are $n$-dimensional ellipsoids in general position.

## Independence of Variables

When the covariance is spherical or diagonal, the individual variables are independent. For example, in the bivariate diagonal case with zero mean:

$$
P(x_1, x_2) = P(x_1) P(x_2)
$$

## Decomposition of Full Covariance

For any bivariate normal distribution with full covariance, there exists a coordinate system with variables $x_1'$ and $x_2'$ where the covariance is diagonal. This means that the ellipsoidal iso-contours align with the coordinate axes in this canonical coordinate frame. The two frames of reference are related by a rotation matrix $R$ that maps $(x_1', x_2')$ to $(x_1, x_2)$. This decomposition of full covariance into the product of a rotation matrix and a diagonal covariance matrix is a fundamental property of multivariate normal distributions.

```python
import numpy as np
import matplotlib.pyplot as plt

def multivariate_normal(mean, cov, size=1):
    """
    Generate random samples from a multivariate normal distribution.
    
    Parameters:
    - mean: Mean vector of the distribution.
    - cov: Covariance matrix of the distribution.
    - size: Number of samples to generate.
    
    Returns:
    - samples: Random samples from the multivariate normal distribution.
    """
    D = len(mean)
    L = np.linalg.cholesky(cov)
    samples = np.random.normal(size=(size, D)) @ L.T + mean
    return samples

def pdf_multivariate_normal(x, mean, cov):
    """
    Calculate the probability density function of a multivariate normal distribution.
    
    Parameters:
    - x: Input vector.
    - mean: Mean vector of the distribution.
    - cov: Covariance matrix of the distribution.
    
    Returns:
    - pdf: Probability density function value at x.
    """
    D = len(mean)
    det_cov = np.linalg.det(cov)
    inv_cov = np.linalg.inv(cov)
    exponent = -0.5 * (x - mean) @ inv_cov @ (x - mean)
    pdf = (1 / (np.power(2 * np.pi, D / 2) * np.sqrt(det_cov))) * np.exp(exponent)
    return pdf

# Example mean and covariance matrix
mean = [0, 0]
cov = [[1, 0], [0, 1]]

# Generate random samples
samples = multivariate_normal(mean, cov, size=1000)

# Calculate the probability density function at a specific point
x = [1, 1]
pdf = pdf_multivariate_normal(x, mean, cov)
print("Probability Density Function at x =", x, ":", pdf)

# Plotting the samples and the probability density function
plt.figure(figsize=(10, 6))
plt.scatter(samples[:, 0], samples[:, 1], label="Samples")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()

# Plotting the probability density function contour
x1, x2 = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
positions = np.vstack([x1.ravel(), x2.ravel()])
values = np.vstack([x.ravel() for x in [x1, x2]])
pdf_values = pdf_multivariate_normal(values, mean, cov)
pdf_values = pdf_values.reshape(x1.shape)
plt.contour(x1, x2, pdf_values, levels=10, colors='black', linestyles='dashed')

plt.title("Multivariate Normal Distribution")
plt.show()
```

```python
import math
import random
import matplotlib.pyplot as plt

def multivariate_normal(mean, cov, size=1):
    """
    Generate random samples from a multivariate normal distribution.
    
    Parameters:
    - mean: Mean vector of the distribution.
    - cov: Covariance matrix of the distribution.
    - size: Number of samples to generate.
    
    Returns:
    - samples: Random samples from the multivariate normal distribution.
    """
    D = len(mean)
    samples = []
    for _ in range(size):
        sample = [random.gauss(mean[i], math.sqrt(cov[i][i])) for i in range(D)]
        samples.append(sample)
    return samples

def pdf_multivariate_normal(x, mean, cov):
    """
    Calculate the probability density function of a multivariate normal distribution.
    
    Parameters:
    - x: Input vector.
    - mean: Mean vector of the distribution.
    - cov: Covariance matrix of the distribution.
    
    Returns:
    - pdf: Probability density function value at x.
    """
    D = len(mean)
    det_cov = math.prod([cov[i][i] for i in range(D)])
    inv_cov = [[0] * D for _ in range(D)]
    for i in range(D):
        for j in range(D):
            inv_cov[i][j] = cov[j][i] / det_cov
    exponent = -0.5 * sum([(x[i] - mean[i]) * inv_cov[i][j] * (x[j] - mean[j]) for i in range(D) for j in range(D)])
    pdf = (1 / (math.pow(2 * math.pi, D / 2) * math.sqrt(det_cov))) * math.exp(exponent)
    return pdf

# Example mean and covariance matrix
mean = [0, 0]
cov = [[1, 0], [0, 1]]

# Generate random samples
samples = multivariate_normal(mean, cov, size=1000)

# Calculate the probability density function at a specific point
x = [1, 1]
pdf = pdf_multivariate_normal(x, mean, cov)
print("Probability Density Function at x =", x, ":", pdf)

# Plotting the samples and the probability density function
plt.figure(figsize=(10, 6))
plt.scatter([sample[0] for sample in samples], [sample[1] for sample in samples], label="Samples")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()

# Plotting the probability density function contour
x1, x2 = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
positions = np.vstack([x1.ravel(), x2.ravel()])
values = np.vstack([x.ravel() for x in [x1, x2]])
pdf_values = pdf_multivariate_normal(values, mean, cov)
pdf_values = pdf_values.reshape(x1.shape)
plt.contour(x1, x2, pdf_values, levels=10, colors='black', linestyles='dashed')

plt.title("Multivariate Normal Distribution")
plt.show()
```

# Decomposition of Covariance

## Introduction

We can use the geometrical intuitions from the previous section to decompose a full covariance matrix $\Sigma_{\text{full}}$. Given a normal distribution with mean zero and a full covariance matrix, we know that the iso-contours take an ellipsoidal form with the major and minor axes at arbitrary orientations.

## Coordinate Transformation

Consider viewing the distribution in a new coordinate frame where the axes are aligned with the axes of the normal (Figure 5.2). In this new frame of reference, the covariance matrix $\Sigma_{\text{diag}}$ will be diagonal. We denote the data vector in the new coordinate system as $x' = [x'_1, x'_2]^T$, where the frames of reference are related by $x' = Rx$.

The probability distribution over $x'$ can be written as:

$$
P(x') = \frac{1}{(2\pi)^{D/2} |\Sigma_{\text{diag}}|^{1/2}} \exp \left(-\frac{1}{2} x'^T \Sigma_{\text{diag}}^{-1} x' \right)
$$

Converting back to the original axes by substituting $x' = Rx$, we get:

$$
P(x) = \frac{1}{(2\pi)^{D/2} |R \Sigma_{\text{diag}} R^T|^{1/2}} \exp \left(-\frac{1}{2} x^T (R \Sigma_{\text{diag}} R^T)^{-1} x \right)
$$

This equation represents a multivariate normal distribution with covariance matrix $\Sigma_{\text{full}} = R \Sigma_{\text{diag}} R^T$.

## Decomposition of Full Covariance

We conclude that full covariance matrices can be expressed as a product of this form, involving a rotation matrix $R$ and a diagonal covariance matrix $\Sigma_{\text{diag}}$. This decomposition allows us to retrieve the principal directions of the ellipsoid (contained in the columns of $R$) and the variances along each of these axes (encoded in the diagonal of $\Sigma_{\text{diag}}$).

## Singular Value Decomposition

The decomposition of a full covariance matrix can be achieved using the singular value decomposition (SVD). The SVD of a matrix $A$ is given by $A = U \Sigma V^T$, where $U$ and $V$ are orthogonal matrices, and $\Sigma$ is a diagonal matrix containing the singular values of $A$.

## Linear Transformations of Variables

The form of the multivariate normal distribution is preserved under linear transformations of the form $y = Ax + b$ (Figure 5.3). If the original distribution is $P(x) = \text{Norm}(x | \mu, \Sigma)$, then the transformed variable $y$ is distributed as:

$$
P(y) = \text{Norm}(y | A\mu + b, A\Sigma A^T)
$$

This relationship provides a simple method to draw samples from a normal distribution with mean $\mu$ and covariance $\Sigma$. We can first draw a sample $x$ from a standard normal distribution and then apply the transform $y = \Sigma^{1/2} x + \mu$.

## Marginal Distributions

If we marginalize over any subset of random variables in a multivariate normal distribution, the resulting marginal distribution is also a multivariate normal. The mean and covariance of the marginal distribution depend on the original mean and covariance, as well as the subset of variables being marginalized over.




Fig.4 The marginal distribution of any subset of variables in a normal distribution is also normally distributed. In other words, if we sum over the distribution in any direction, the remaining quantity is also nor- mally distributed. To ﬁnd the mean and the covariance of the new distri- bution, we can simply extract the rel- evant entries from the original mean and covariance matrix.

```python
import random
import math

def decompose_covariance(cov):
    """
    Decompose a full covariance matrix into a rotation matrix and a diagonal covariance matrix.
    
    Parameters:
    - cov: Full covariance matrix.
    
    Returns:
    - R: Rotation matrix.
    - diag_cov: Diagonal covariance matrix.
    """
    D = len(cov)
    U, S, Vt = svd(cov)
    diag_cov = [[0] * D for _ in range(D)]
    for i in range(D):
        diag_cov[i][i] = S[i]
    R = Vt
    return R, diag_cov

def svd(matrix):
    """
    Perform singular value decomposition on a matrix.
    
    Parameters:
    - matrix: Input matrix.
    
    Returns:
    - U: Left singular vectors.
    - S: Singular values.
    - Vt: Transpose of right singular vectors.
    """
    D = len(matrix)
    U = [[0] * D for _ in range(D)]
    S = [0] * D
    Vt = [[0] * D for _ in range(D)]
    
    for i in range(D):
        x = [random.random() for _ in range(D)]
        for j in range(D):
            x[j] -= matrix[j][i] * matrix[j][i]
        norm_x = math.sqrt(sum(x[k]**2 for k in range(D)))
        x = [x[k] / norm_x for k in range(D)]
        
        for j in range(D):
            U[i][j] = sum(matrix[j][k] * x[k] for k in range(D))
        
        for j in range(D):
            y = [random.random() for _ in range(D)]
            for k in range(D):
                y[k] -= U[i][k] * U[i][k]
            norm_y = math.sqrt(sum(y[k]**2 for k in range(D)))
            y = [y[k] / norm_y for k in range(D)]
            
            S[i] += sum(matrix[j][k] * x[k] * y[j] for k in range(D))
            
            for j in range(D):
                Vt[i][j] = sum(matrix[j][k] * x[k] * y[k] for k in range(D))
    
    return U, S, Vt

# Example full covariance matrix
cov = [[1, 0], [0, 1]]

# Decompose the covariance matrix
R, diag_cov = decompose_covariance(cov)

print("Rotation Matrix:")
for row in R:
    print(row)

print("\nDiagonal Covariance Matrix:")
for row in diag_cov:
    print(row)
```

```python
import numpy as np

def decompose_covariance(cov):
    """
    Decompose a full covariance matrix into a rotation matrix and a diagonal covariance matrix.
    
    Parameters:
    - cov: Full covariance matrix.
    
    Returns:
    - R: Rotation matrix.
    - diag_cov: Diagonal covariance matrix.
    """
    U, S, Vt = np.linalg.svd(cov)
    diag_cov = np.diag(S)
    R = Vt
    return R, diag_cov

# Example full covariance matrix
cov = np.array([[1, 0], [0, 1]])

# Decompose the covariance matrix
R, diag_cov = decompose_covariance(cov)

print("Rotation Matrix:")
print(R)

print("\nDiagonal Covariance Matrix:")
print(diag_cov)
```

# Marginal Distributions

## Introduction

The marginal distribution of any subset of variables in a normal distribution is also normally distributed. In other words, if we sum over the distribution in any direction, the remaining quantity is also normally distributed. This property allows us to easily find the mean and covariance of the marginal distribution by extracting the relevant entries from the original mean and covariance matrix.

## Marginal Distribution of a Subset of Variables

Consider a normal distribution with mean $\mu$ and covariance matrix $\Sigma$, and partition the random variable into two parts: $x = [x_1^T, x_2^T]^T$. The marginal distribution of $x_1$ and $x_2$ can be written as:

$$
P(x_1) = \text{Norm}(x_1 | \mu_1, \Sigma_{11}) \quad \text{and} \quad P(x_2) = \text{Norm}(x_2 | \mu_2, \Sigma_{22})
$$

where:
- $\mu_1$ and $\mu_2$ are the mean vectors of $x_1$ and $x_2$, respectively.
- $\Sigma_{11}$ and $\Sigma_{22}$ are the covariance matrices of $x_1$ and $x_2$, respectively.

## Extracting Mean and Covariance

To find the mean and covariance of the marginal distribution of a subset of variables, we can simply extract the relevant entries from the original mean and covariance matrix.

- The mean vector of the marginal distribution is given by the corresponding entries of the original mean vector $\mu$.
- The covariance matrix of the marginal distribution is given by the corresponding submatrix of the original covariance matrix $\Sigma$.

## Example

Let's consider a normal distribution with mean $\mu = [1, 2, 3]^T$ and covariance matrix:

$$
\Sigma = \begin{bmatrix}
1 & 0.5 & 0.3 \\
0.5 & 2 & 0.2 \\
0.3 & 0.2 & 1
\end{bmatrix}
$$

If we want to find the marginal distribution of $x_1$ and $x_2$, we can extract the relevant entries:

- Mean vector of $x_1$: $\mu_1 = [1, 2]^T$
- Mean vector of $x_2$: $\mu_2 = [3]$
- Covariance matrix of $x_1$: $\Sigma_{11} = \begin{bmatrix} 1 & 0.5 \\ 0.5 & 2 \end{bmatrix}$
- Covariance matrix of $x_2$: $\Sigma_{22} = [1]$

So, the marginal distributions are:

$$
P(x_1) = \text{Norm}(x_1 | [1, 2]^T, \begin{bmatrix} 1 & 0.5 \\ 0.5 & 2 \end{bmatrix}) \quad \text{and} \quad P(x_2) = \text{Norm}(x_2 | [3], [1])
$$


```python
import numpy as np

def marginal_distribution(mean, cov, indices):
    """
    Find the marginal distribution of a subset of variables in a normal distribution.
    
    Parameters:
    - mean: Mean vector of the original distribution.
    - cov: Covariance matrix of the original distribution.
    - indices: Indices of the variables to consider for the marginal distribution.
    
    Returns:
    - marginal_mean: Mean vector of the marginal distribution.
    - marginal_cov: Covariance matrix of the marginal distribution.
    """
    marginal_mean = mean[indices]
    marginal_cov = cov[np.ix_(indices, indices)]
    return marginal_mean, marginal_cov

# Example mean and covariance matrix
mean = np.array([1, 2, 3])
cov = np.array([[1, 0.5, 0.3], [0.5, 2, 0.2], [0.3, 0.2, 1]])

# Indices of the variables to consider for the marginal distribution
indices = [0, 1]

# Find the marginal distribution
marginal_mean, marginal_cov = marginal_distribution(mean, cov, indices)

print("Marginal Mean:", marginal_mean)
print("Marginal Covariance Matrix:")
print(marginal_cov)
```

```python
def marginal_distribution(mean, cov, indices):
    """
    Find the marginal distribution of a subset of variables in a normal distribution.
    
    Parameters:
    - mean: Mean vector of the original distribution.
    - cov: Covariance matrix of the original distribution.
    - indices: Indices of the variables to consider for the marginal distribution.
    
    Returns:
    - marginal_mean: Mean vector of the marginal distribution.
    - marginal_cov: Covariance matrix of the marginal distribution.
    """
    D = len(mean)
    marginal_mean = [mean[i] for i in indices]
    marginal_cov = [[0] * len(indices) for _ in range(len(indices))]
    for i in range(len(indices)):
        for j in range(len(indices)):
            marginal_cov[i][j] = cov[indices[i]][indices[j]]
    return marginal_mean, marginal_cov

# Example mean and covariance matrix
mean = [1, 2, 3]
cov = [[1, 0.5, 0.3], [0.5, 2, 0.2], [0.3, 0.2, 1]]

# Indices of the variables to consider for the marginal distribution
indices = [0, 1]

# Find the marginal distribution
marginal_mean, marginal_cov = marginal_distribution(mean, cov, indices)

print("Marginal Mean:", marginal_mean)
print("Marginal Covariance Matrix:")
for row in marginal_cov:
    print(row)
```

# Conditional Distributions

## Introduction

If the variable $x$ is distributed as a multivariate normal, then the conditional distribution of a subset of variables $x_1$ given known values for the remaining variables $x_2$ is also distributed as a multivariate normal (Figure 5.5).

## Conditional Distributions Formula



Fig.5 Conditional distributions of multivariate normal. a) If we take any multivariate normal distribution, ﬁx a subset of the variables, and look at the distribution of the remaining variables, this distribution will also take the form of a normal. The mean of this new normal depends on the values that we ﬁxed the subset to, but the covariance is always the same. b) If the original multivariate normal has spherical or diagonal covariance, both the mean and covariance of the resulting normal distributions are the same, regardless of the value we conditioned on: these forms of covariance matrix imply independence between the constituent variables.

Given a multivariate normal distribution with mean $\mu$ and covariance matrix $\Sigma$, and partitioning the random variable into two parts: $x = [x_1^T, x_2^T]^T$, the conditional distributions are:

$$
P(x_1 | x_2 = x_2^*) = \text{Norm}(x_1 | \mu_1 + \Sigma_{12} \Sigma_{22}^{-1} (x_2^* - \mu_2), \Sigma_{11} - \Sigma_{12} \Sigma_{22}^{-1} \Sigma_{21})
$$

$$
P(x_2 | x_1 = x_1^*) = \text{Norm}(x_2 | \mu_2 + \Sigma_{21} \Sigma_{11}^{-1} (x_1^* - \mu_1), \Sigma_{22} - \Sigma_{21} \Sigma_{11}^{-1} \Sigma_{12})
$$

## Product of Two Normals

The product of two normal distributions is proportional to a third normal distribution (Figure 5.6). If the two original distributions have means $a$ and $b$ and covariances $A$ and $B$, respectively, then:

$$
\text{Norm}(x | a, A) \cdot \text{Norm}(x | b, B) \propto \text{Norm}(x | A^{-1} + B^{-1} (A^{-1} a + B^{-1} b), A^{-1} + B^{-1})
$$

The constant of proportionality is itself a normal distribution:

$$
\text{Norm}(a | b, A + B) = \text{Norm}(b | a, A + B)
$$



Fig.6 The product of any two normals N1 and N2 is proportional to a third normal distribution, with a mean between the two original means and a variance that is smaller than ei- ther of the original distributions.

## Self-Conjugacy

The preceding property can be used to demonstrate that the normal distribution is self-conjugate with respect to its mean $\mu$. Consider taking a product of a normal distribution over data $x$ and a second normal distribution over the mean vector $\mu$ of the first distribution. It can be shown that:

$$
\text{Norm}(x | \mu, \Sigma) \cdot \text{Norm}(\mu | \mu_p, \Sigma_p) = \kappa \cdot \text{Norm}(\mu | \tilde{\mu}, \tilde{\Sigma})
$$

where $\kappa$ is a normal distribution, and $\tilde{\mu}$ and $\tilde{\Sigma}$ are determined from the product of normals formula. This demonstrates the self-conjugacy of the normal distribution.


```python
import numpy as np

def conditional_distribution(mean, cov, indices1, indices2, x2_value):
    """
    Find the conditional distribution of a subset of variables in a multivariate normal distribution.
    
    Parameters:
    - mean: Mean vector of the original distribution.
    - cov: Covariance matrix of the original distribution.
    - indices1: Indices of the variables to consider for the first subset.
    - indices2: Indices of the variables to consider for the second subset.
    - x2_value: Known values for the second subset of variables.
    
    Returns:
    - conditional_mean: Mean vector of the conditional distribution.
    - conditional_cov: Covariance matrix of the conditional distribution.
    """
    D = len(mean)
    indices = indices1 + indices2
    marginal_mean = mean[indices]
    marginal_cov = cov[np.ix_(indices, indices)]
    
    # Extract the submatrices for the conditional distribution
    Sigma_11 = marginal_cov[np.ix_(indices1, indices1)]
    Sigma_12 = marginal_cov[np.ix_(indices1, indices2)]
    Sigma_21 = marginal_cov[np.ix_(indices2, indices1)]
    Sigma_22 = marginal_cov[np.ix_(indices2, indices2)]
    
    # Calculate the conditional mean and covariance
    conditional_mean = marginal_mean[indices1] + Sigma_12 @ np.linalg.inv(Sigma_22) @ (x2_value - marginal_mean[indices2])
    conditional_cov = Sigma_11 - Sigma_12 @ np.linalg.inv(Sigma_22) @ Sigma_21
    
    return conditional_mean, conditional_cov

# Example mean and covariance matrix
mean = np.array([1, 2, 3, 4, 5])
cov = np.array([[1, 0.5, 0.3, 0.2, 0.1], [0.5, 2, 0.2, 0.1, 0.05], [0.3, 0.2, 1, 0.08, 0.03], [0.2, 0.1, 0.08, 1, 0.02], [0.1, 0.05, 0.03, 0.02, 1]])

# Indices of the variables to consider for the first and second subsets
indices1 = [0, 1]
indices2 = [2, 3, 4]

# Known values for the second subset of variables
x2_value = np.array([3.5, 4.2, 5.1])

# Find the conditional distribution
conditional_mean, conditional_cov = conditional_distribution(mean, cov, indices1, indices2, x2_value)

print("Conditional Mean:", conditional_mean)
print("Conditional Covariance Matrix:")
print(conditional_cov)
```

```python
import math
import random
import matplotlib.pyplot as plt

def conditional_distribution(mean, cov, indices1, indices2, x2_value):
    """
    Find the conditional distribution of a subset of variables in a multivariate normal distribution.
    
    Parameters:
    - mean: Mean vector of the original distribution.
    - cov: Covariance matrix of the original distribution.
    - indices1: Indices of the variables to consider for the first subset.
    - indices2: Indices of the variables to consider for the second subset.
    - x2_value: Known values for the second subset of variables.
    
    Returns:
    - conditional_mean: Mean vector of the conditional distribution.
    - conditional_cov: Covariance matrix of the conditional distribution.
    """
    D = len(mean)
    indices = indices1 + indices2
    marginal_mean = [mean[i] for i in indices]
    marginal_cov = [[0] * len(indices) for _ in range(len(indices))]
    
    for i in range(len(indices)):
        for j in range(len(indices)):
            marginal_cov[i][j] = cov[indices[i]][indices[j]]
    
    # Extract the submatrices for the conditional distribution
    Sigma_11 = [[marginal_cov[i][j] for j in range(len(indices1))] for i in range(len(indices1))]
    Sigma_12 = [[marginal_cov[i][j] for j in range(len(indices1), len(indices))] for i in range(len(indices1))]
    Sigma_21 = [[marginal_cov[i][j] for j in range(len(indices1))] for i in range(len(indices1), len(indices))]
    Sigma_22 = [[marginal_cov[i][j] for j in range(len(indices1), len(indices))] for i in range(len(indices1), len(indices))]
    
    # Calculate the conditional mean and covariance
    conditional_mean = [marginal_mean[i] for i in indices1] + [Sigma_12[i][j] * math.pow(Sigma_22[j][j], -1) * (x2_value[j] - marginal_mean[indices2[j]]) for i in range(len(indices1)) for j in range(len(indices2))]
    conditional_cov = [[Sigma_11[i][j] - Sigma_12[i][k] * math.pow(Sigma_22[k][k], -1) * Sigma_21[k][j] for j in range(len(indices1))] for i in range(len(indices1))]
    
    return conditional_mean, conditional_cov

# Example mean and covariance matrix
mean = [1, 2, 3, 4, 5]
cov = [[1, 0.5, 0.3, 0.2, 0.1], [0.5, 2, 0.2, 0.1, 0.05], [0.3, 0.2, 1, 0.08, 0.03], [0.2, 0.1, 0.08, 1, 0.02], [0.1, 0.05, 0.03, 0.02, 1]]

# Indices of the variables to consider for the first and second subsets
indices1 = [0, 1]
indices2 = [2, 3, 4]

# Known values for the second subset of variables
x2_value = [3.5, 4.2, 5.1]

# Find the conditional distribution
conditional_mean, conditional_cov = conditional_distribution(mean, cov, indices1, indices2, x2_value)

print("Conditional Mean:", conditional_mean)
print("Conditional Covariance Matrix:")
for row in conditional_cov:
    print(row)

# Plotting the original distribution and the conditional distribution
D = len(mean)
x = [random.gauss(mean[i], math.sqrt(cov[i][i])) for i in range(D)]

plt.figure(figsize=(10, 6))
plt.scatter(x[indices1[0]], x[indices1[1]], label="Original Distribution")
plt.scatter(conditional_mean[0], conditional_mean[1], color='red', label="Conditional Distribution")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.title("Original and Conditional Distributions")
plt.show()
```

# Change of Variable

## Introduction

Consider a normal distribution in variable $x$ whose mean is a linear function $Ay + b$ of a second variable $y$. We can re-express this in terms of a normal distribution in $y$ which is a linear function $A^{-1} x + b^{-1}$ of $x$. This relationship is given by:

$$
\text{Norm}(x | Ay + b, \Sigma) = \kappa \cdot \text{Norm}(y | A^{-1} x + b^{-1}, \Sigma^{-1})
$$

where $\kappa$ is a constant, and the new parameters are:

$$
\Sigma^{-1} = (A^T \Sigma^{-1} A)^{-1} \quad A^{-1} = (A^T \Sigma^{-1} A)^{-1} A^T \Sigma^{-1} b \quad b^{-1} = -(A^T \Sigma^{-1} A)^{-1} A^T \Sigma^{-1} b
$$

## Visual Interpretation

This relationship is mathematically opaque, but it is easy to understand visually when $x$ and $y$ are scalars (Figure 5.7). It is often used in the context of Bayes' rule, where our goal is to move from $P(x|y)$ to $P(y|x)$.

## Example

Let's consider a normal distribution in $x$ with mean $Ay + b$ and covariance matrix $\Sigma$, where:

$$
A = \begin{bmatrix}
a & 0 \\
0 & 1
\end{bmatrix}, \quad
b = \begin{bmatrix}
b_1 \\
b_2
\end{bmatrix}, \quad
\Sigma = \begin{bmatrix}
\sigma_1^2 & 0 \\
0 & \sigma_2^2
\end{bmatrix}
$$

Using the change of variable formula, we can express this distribution in terms of $y$:

$$
\text{Norm}(y | A^{-1} x + b^{-1}, \Sigma^{-1})
$$

where:

$$
A^{-1} = \begin{bmatrix}
a^{-1} & 0 \\
0 & 1
\end{bmatrix}, \quad
b^{-1} = \begin{bmatrix}
b_1^{-1} \\
b_2^{-1}
\end{bmatrix}, \quad
\Sigma^{-1} = \begin{bmatrix}
\sigma_1^{-2} & 0 \\
0 & \sigma_2^{-2}
\end{bmatrix}
$$

## Visual Representation

Fig.7 illustrates this change of variable visually. It shows how a normal distribution in $x$ with a linear mean function $Ay + b$ is equivalent to a constant $\kappa$ times a normal distribution in $y$ with a linear mean function $A^{-1} x + b^{-1}$.


```python
import numpy as np

def change_of_variable(mean, cov, A, b):
    """
    Apply the change of variable formula to a multivariate normal distribution.
    
    Parameters:
    - mean: Mean vector of the original distribution.
    - cov: Covariance matrix of the original distribution.
    - A: Linear transformation matrix.
    - b: Linear transformation vector.
    
    Returns:
    - new_mean: Mean vector of the new distribution.
    - new_cov: Covariance matrix of the new distribution.
    """
    D = len(mean)
    new_mean = A @ mean + b
    new_cov = np.linalg.inv(A.T @ np.linalg.inv(cov) @ A)
    return new_mean, new_cov

# Example mean and covariance matrix
mean = np.array([1, 2, 3])
cov = np.array([[1, 0.5, 0.3], [0.5, 2, 0.2], [0.3, 0.2, 1]])

# Linear transformation matrix and vector
A = np.array([[2, 0, 0], [0, 1, 0], [0, 0, 1]])
b = np.array([1, 0, 0])

# Apply the change of variable formula
new_mean, new_cov = change_of_variable(mean, cov, A, b)

print("New Mean:", new_mean)
print("New Covariance Matrix:")
print(new_cov)
```

```python
def change_of_variable(mean, cov, A, b):
    """
    Apply the change of variable formula to a multivariate normal distribution.
    
    Parameters:
    - mean: Mean vector of the original distribution.
    - cov: Covariance matrix of the original distribution.
    - A: Linear transformation matrix.
    - b: Linear transformation vector.
    
    Returns:
    - new_mean: Mean vector of the new distribution.
    - new_cov: Covariance matrix of the new distribution.
    """
    D = len(mean)
    new_mean = [sum(A[i][j] * mean[j] for j in range(D)) + b[i] for i in range(D)]
    new_cov = [[0] * D for _ in range(D)]
    
    for i in range(D):
        for j in range(D):
            new_cov[i][j] = sum(A[i][k] * cov[k][l] * A[j][l] for k in range(D) for l in range(D))
    
    return new_mean, new_cov

# Example mean and covariance matrix
mean = [1, 2, 3]
cov = [[1, 0.5, 0.3], [0.5, 2, 0.2], [0.3, 0.2, 1]]

# Linear transformation matrix and vector
A = [[2, 0, 0], [0, 1, 0], [0, 0, 1]]
b = [1, 0, 0]

# Apply the change of variable formula
new_mean, new_cov = change_of_variable(mean, cov, A, b)

print("New Mean:", new_mean)
print("New Covariance Matrix:")
for row in new_cov:
    print(row)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def change_of_variable(mean, cov, A, b):
    """
    Apply the change of variable formula to a multivariate normal distribution.
    
    Parameters:
    - mean: Mean vector of the original distribution.
    - cov: Covariance matrix of the original distribution.
    - A: Linear transformation matrix.
    - b: Linear transformation vector.
    
    Returns:
    - new_mean: Mean vector of the new distribution.
    - new_cov: Covariance matrix of the new distribution.
    """
    D = len(mean)
    new_mean = A @ mean + b
    new_cov = np.linalg.inv(A.T @ np.linalg.inv(cov) @ A)
    return new_mean, new_cov

# Example mean and covariance matrix
mean = np.array([1, 2, 3])
cov = np.array([[1, 0.5, 0.3], [0.5, 2, 0.2], [0.3, 0.2, 1]])

# Linear transformation matrix and vector
A = np.array([[2, 0, 0], [0, 1, 0], [0, 0, 1]])
b = np.array([1, 0, 0])

# Apply the change of variable formula
new_mean, new_cov = change_of_variable(mean, cov, A, b)

print("New Mean:", new_mean)
print("New Covariance Matrix:")
print(new_cov)

# Plotting the original distribution
x = np.random.multivariate_normal(mean, cov, size=1000)
plt.figure(figsize=(10, 6))
plt.scatter(x[:, 0], x[:, 1], label="Original Distribution")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()

# Plotting the new distribution
new_x = np.random.multivariate_normal(new_mean, new_cov, size=1000)
plt.scatter(new_x[:, 0], new_x[:, 1], color='red', label="New Distribution")
plt.title("Original and New Distributions")
plt.show()
```
