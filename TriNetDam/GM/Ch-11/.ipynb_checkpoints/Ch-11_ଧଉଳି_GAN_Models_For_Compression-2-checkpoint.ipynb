## Generative AI Systems

###  Introduction

Large Language Models (LLMs) have revolutionized the field of artificial intelligence by enabling natural language communication with machines. This technological advancement has not only changed the way we interact with AI but also our understanding of it. Recent developments in GenAI, such as the GPT family (e.g., GPT-4V) and Google-based models (e.g., Gemini), have brought multimodal learning back into the spotlight and shown great potential for using LLMs as multimodal systems.

Current LLMs are capable of processing various data modalities, including text, images, audio, and video, and generating different data types. As a result, it is more accurate to refer to these systems as GenAI Systems rather than solely LLMs. Text, or natural language, serves as a communication medium, replacing predefined formal protocols and modality encoders that act as I/O interfaces for processing diverse data sources.

Furthermore, GenAI Systems can be enhanced by attaching databases, knowledge graphs, or external specialized tools. For example, a tool-augmented Seq2Seq model proposed by [44] utilizes a fine-tuned T5 LLM as a backbone to interact with external tools for input and output processing. Another system for chemistry [45] integrates external databases and tools to provide reliable answers and draw molecules.

In the following sections, we will explore various Generative AI Systems (GenAISys) that utilize GenAI modules. It is interesting to discuss these systems at a high level, as we are still far from fully understanding their building blocks, such as neural networks and their simplest units. While the community has a good understanding of the compositionality of deep learning [46, 47, 59], many claims remain hypotheses. Nonetheless, we can delve into the discussion of compositionality at a higher level, similar to control theory and systems science and engineering [48].

## GenAISys: A General Architecture

When discussing systems, we often think of multiple modules and components combined to achieve specific goals. The definition of a system can be blurry at times, but we can consider a system as a composition of distinguished parts or simpler components. Category theory [49, 59] provides an interesting perspective on systems and compositionality.

A GenAISys consists of the following key parts:

- **Data Encoders (DEs)**: All raw input data are processed using models that encode them into tensors. These encoders are typically pre-trained and kept non-trainable (frozen). Foundation models like BERT for text encoding or ConvNeXT for image encoding can be used as DEs. Encoders can also be composed of other modules, such as transforming speech into text using Whisper and then encoding the text with BERT. Data encoders are crucial in formulating GenAISys and have catalyzed the development of these systems.

- **GenAI Model (GeM)**: The central unit of a GenAISys is the GenAI model, which can be an LLM or another generative model. It processes the encoded input data, communicates with external memory (database) and tools, and generates an output. GeM is the core component of the architecture and can have its own short-term memory (cache) and built-in instructions for interacting with other components. It may also include submodels for specific tasks, such as named entity recognition (NER) for detecting instructions in inputs.

- **Retrieval/Storage Module (R/S)**: This module assists the GeM by retrieving facts from long-term memory and utilizing inputs processed by DEs. It can also have its own models, such as BERT for embedding input text, and sets of instructions, like routing algorithms. The R/S module plays a vital role in accessing external tools and databases.

Figure 11.4 presents a general architecture of a Generative AI System, highlighting the main components: data encoders for various modalities, a retrieval/storage module, and a central generative AI model producing new content (output). The snowflake icon represents that a module is "frozen," indicating it is already trained.

## Training

Training a GenAISys is non-trivial due to the trainability of all its components. End-to-end training is possible but infeasible with current hardware. Instead, foundation models are often pre-trained separately to formulate DEs and an R/S, and only the GeM is trained. With advancements in hardware and training schemes, GenAISys will likely be trained more effectively, utilizing multiple data modalities simultaneously.

## A Side Note

The general architecture of GenAISys resembles the functional map of the human brain. Applying a coordinate descent algorithm to optimize the system one modality or block at a time could keep the complexity manageable. However, the challenge lies in preventing the system from forgetting previously learned representations. While continual learning methods are being developed [50, 51], solving this issue will be a significant breakthrough in AI.

```python

import torch
from torch import nn
from torch.nn import functional as F
from torchvision import transforms

# Define the text encoder
class TextEncoder(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super(TextEncoder, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)

    def forward(self, text):
        embedded = self.embedding(text)
        output, _ = self.lstm(embedded)
        return output

# Define the diffusion model
class DiffusionModel(nn.Module):
    def __init__(self, in_channels, out_channels, hidden_dim):
        super(DiffusionModel, self).__init__()
        self.conv1 = nn.Conv2d(in_channels, hidden_dim, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(hidden_dim, hidden_dim, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(hidden_dim, out_channels, kernel_size=3, padding=1)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = self.conv3(x)
        return x

# Initialize the text encoder and diffusion model
text_encoder = TextEncoder(vocab_size=10000, embedding_dim=300, hidden_dim=512)
diffusion_model = DiffusionModel(in_channels=3, out_channels=3, hidden_dim=128)

# Define the LVM system
class LVM:
    def __init__(self, text_encoder, diffusion_model):
        self.text_encoder = text_encoder
        self.diffusion_model = diffusion_model

    def generate_image(self, text):
        # Tokenize and encode the text
        encoded_text = self.text_encoder(text)

        # Generate the image using the diffusion model
        image = self.diffusion_model(encoded_text)

        return image

# Example usage
text = "A beautiful sunset over the ocean"
lvm = LVM(text_encoder, diffusion_model)
generated_image = lvm.generate_image(text)

# Save or display the generated image
save_image(generated_image, "generated_image.png")
```

```python
class TextEncoder:
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        self.vocab_size = vocab_size
        self.embedding_dim = embedding_dim
        self.hidden_dim = hidden_dim
        self.embeddings = [0] * vocab_size
        self.hidden_state = [0] * hidden_dim

    def encode(self, text):
        encoded_text = []
        for word in text:
            if word in self.embeddings:
                encoded_text.append(self.embeddings[word])
            else:
                # OOV word, you can handle it as needed (e.g., use a default embedding)
                encoded_text.append(self.embeddings[0])
        return encoded_text

class DiffusionModel:
    def __init__(self, in_channels, out_channels, hidden_dim):
        self.in_channels = in_channels
        self.out_channels = out_channels
        self.hidden_dim = hidden_dim
        self.weights = [0] * (in_channels * hidden_dim)
        self.biases = [0] * hidden_dim
        self.output_weights = [0] * (hidden_dim * out_channels)
        self.output_biases = [0] * out_channels

    def forward(self, x):
        # Simple forward pass using matrix multiplication and bias addition
        hidden = [0] * self.hidden_dim
        for i in range(self.hidden_dim):
            for j in range(self.in_channels):
                hidden[i] += x[j] * self.weights[i * self.in_channels + j]
        for i in range(self.hidden_dim):
            hidden[i] += self.biases[i]
        output = [0] * self.out_channels
        for i in range(self.out_channels):
            for j in range(self.hidden_dim):
                output[i] += hidden[j] * self.output_weights[i * self.hidden_dim + j]
        for i in range(self.out_channels):
            output[i] += self.output_biases[i]
        return output

# Initialize the text encoder and diffusion model
text_encoder = TextEncoder(vocab_size=10000, embedding_dim=300, hidden_dim=512)
diffusion_model = DiffusionModel(in_channels=3, out_channels=3, hidden_dim=128)

# Define the LVM system
class LVM:
    def __init__(self, text_encoder, diffusion_model):
        self.text_encoder = text_encoder
        self.diffusion_model = diffusion_model

    def generate_image(self, text):
        # Encode the text
        encoded_text = self.text_encoder.encode(text)

        # Generate the image using the diffusion model
        image = self.diffusion_model.forward(encoded_text)

        return image

# Example usage
text = "A beautiful sunset over the ocean"
lvm = LVM(text_encoder, diffusion_model)
generated_image = lvm.generate_image(text)

# Save or display the generated image (you'll need to implement this part)
# You can use libraries like PIL or OpenCV for image manipulation
```
